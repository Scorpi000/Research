# 基础概念

* $\boldsymbol{\iota}=(1,\ldots,1)'$: 全 1 向量；$\boldsymbol{\iota}_k$：第 k 个元素为 1，其他元素为 0 的向量。
* $\mathbf{r} = (r_1,\ldots,r_N)'$：风险资产的收益率向量，$r_n$ 表示风险资产 $n$ 上的收益率，通常定义为随机变量；
* $\mathbf{h} = (h_1,\ldots,h_N)'$: 风险资产头寸的权重向量，也称为投资组合，$h_n$ 表示组合在风险资产 $n$ 上的投资权重；
* $\mathbf{f} = \mathbb{E}\mathbf{r} = (f_1,f_2,\ldots,f_N)'$: 风险资产的预期超额收益率向量，$f_n$ 表示风险资产 $n$ 的预期超额收益率，$f_p = \mathbf{f}'\mathbf{h}_p$ 表示组合 p 的预期超额收益率；
* $\mathbf{V} = \operatorname{Var}(\mathbf{r}) = \mathbb{E}\left[(\mathbf{r}-\mathbb{E}\mathbf{r})(\mathbf{r}-\mathbb{E}\mathbf{r})'\right] = (\sigma^2_{n,m})_{N\times N}$: 风险资产超额收益率的协方差矩阵，假设为正定矩阵，$\sigma_p^2=\mathbf{h}'_p\mathbf{V}\mathbf{h}_p$ 为组合 p 的方差；

* $\mathbf{h}_B$：基准组合 B 风险资产权重向量；
* $\boldsymbol{\beta} = \frac{\operatorname{Cov}(\mathbf{r}, r_B)}{\sigma_B^2} = (\beta_1,\ldots,\beta_N)'$: 风险资产相对于基准组合 B 的 beta 值，$r_B$ 是基准组合 B 的收益率，$\sigma_B$ 是基准组合的波动率，$\beta_n$ 表示风险资产 $n$ 的 beta 值，$\beta_p = \boldsymbol{\beta}'\mathbf{h}_p$ 为组合 p 的 beta，基准组合 B 的 beta 为 1：$\beta_B = \boldsymbol{\beta}'\mathbf{h}_B = 1$；
* $\boldsymbol{\theta} = \mathbf{r} - \boldsymbol{\beta}\cdot r_B$：风险资产的残差收益率向量，随机变量，自由度为 N-1，因为 $\boldsymbol{\theta}'\mathbf{h}_B=0$；
* $\boldsymbol{\alpha} = \mathbb{E}\boldsymbol{\theta} = \mathbf{f} - \boldsymbol{\beta}f_B$：风险资产的 alpha 向量，$\alpha_p = \boldsymbol{\alpha}'\mathbf{h}_p$ 为组合 p 的 alpha，基准组合 B 的 alpha 为 0：$\alpha_B = \boldsymbol{\alpha}'\mathbf{h}_B = 0$；

* $\boldsymbol{\mu} = \boldsymbol{\beta}^M\mu_M = (\mu_1,\mu_2,\ldots,\mu_N)'$: 一致预期超额收益率（或可认为是 CAPM 模型下风险资产的预期超额收益率），$\mu_n$ 表示风险资产 $n$ 的一致预期超额收益率，$\boldsymbol{\beta}^M$ 是风险资产相对于市场组合 M 的 beta 值，$\mu_M$ 是市场组合 M 的一致预期超额收益率，当 CAPM 成立时有 $\mathbf{f}=\boldsymbol{\mu}$；

## 基本假设

* A1. 存在一个无风险资产 F，其收益率为 $i_F$；
* A2. 所有一阶和二阶矩均存在；
* A3. 不存在零风险的全额投资组合；
* A4. 最小方差的全额投资组合的预期超额收益率是严格正值

## 变量分解

选定基准组合 B $\mathbf{h}_B$，可以将权重向量，预期超额收益率，组合方差分解成基准（Benchmark）和残差（Residual）部分：

**权重向量分解**：
$$
\mathbf{h}_p = \beta_p\cdot\mathbf{h}_B + \mathbf{h}_{pR} = \mathbf{h}_B + \beta_{pA}\cdot\mathbf{h}_B + \mathbf{h}_{pR}
$$
其中，$\beta_p = \boldsymbol{\beta}'\mathbf{h}_p$ 是组合 p 的 beta，$\beta_{pA} = \beta_p - 1$ 为组合 p 的 **主动 beta（Active beta）**，$\mathbf{h}_{pR}$ 为残差组合权重向量，其与 beta 正交，即残差组合的 beta 为 0：
$$
\beta_{pR} = \boldsymbol{\beta}'\mathbf{h}_{pR} = 0
$$
可以建立组合权重空间 $\left\{\mathbf{h}_p\in\mathbb{R}^N\right\}$ 与空间 $\left\{\left(\beta_{pA},\mathbf{h}_{pR}\right)\in\mathbb{R}^{N+1} \big| \beta_{pA}\in\mathbb{R},\ \boldsymbol{\beta}'\mathbf{h}_{pR}=0,\mathbf{h}_{pR}\in\mathbb{R}^N\right\}$ 的一一对应：
$$
\left\{\begin{aligned}
& \beta_{pA} = \boldsymbol{\beta}'\mathbf{h}_p - 1\\
& \mathbf{h}_{pR} = \mathbf{h}_p - (1+\beta_{pA})\cdot\mathbf{h}_B
\end{aligned}
\right.,\ \mathbf{h}_p = (1+\beta_{pA})\cdot\mathbf{h}_B + \mathbf{h}_{pR}
$$
后者是 $\mathbb{R}^{N+1}$ 中的线性子空间，其将决策行为分成独立的两部分：基准择时 $\beta_{pA}$ 和资产选择 $\mathbf{h}_{pR}$。

**预期超额收益率分解**：
$$
f_p = \beta_p\cdot f_B + \alpha_p = f_B + \beta_{pA}\cdot\mu_B + \beta_{pA}\cdot\Delta f_B + \alpha_p
$$
其中，$\Delta f_B = f_B - \mu_B$ 为基准的预期超常收益率（Exceptional Return），$\alpha_p$ 为组合 p 的 alpha 或者预期残差收益率，其可以表达为：
$$
\alpha_p = \boldsymbol{\alpha}'\cdot\mathbf{h}_p = \mathbf{f}'\cdot\mathbf{h}_{pR} = \boldsymbol{\alpha}'\cdot\mathbf{h}_{pR}
$$
预期超额收益率由四部分组成：
* $f_B$：基准的预期超额收益率，与预测（Forecast）有关，与行为（Action）无关；
* $\beta_{pA}\cdot\mu_B$：主动 beta 与基准的一致预期产生的收益率，与预测无关，与行为有关；
* $\beta_{pA}\cdot\Delta f_B$：主动 beta 与基准的超常预测产生的收益率，与预测和行为均有关；
* $\alpha_p$：资产 alpha 与资产选择产生的收益率，与预测和行为均有关；

**协方差矩阵分解**：
$$
\mathbf{V} = \sigma_B^2\boldsymbol{\beta}\boldsymbol{\beta}' + \mathbf{VR}
$$
其中，$\mathbf{VR}$ 为残差协方差矩阵，其为半正定矩阵，秩为 $N-1$。

**组合方差分解**：
$$
\sigma_p^2 = \beta_p^2\cdot\sigma_B^2 + \omega_p^2 = \sigma_B^2 + 2\cdot\beta_{pA}\cdot\sigma_B^2 + \beta^2_{pA}\cdot\sigma_B^2 + \omega_p^2
$$
其中，$\omega_p^2$ 为残差方差，$\omega_p$ 称为残差风险，其可以表达为：
$$
\omega_p^2 = \mathbf{h}'_p\cdot\mathbf{VR}\cdot\mathbf{h}_p = \mathbf{h}'_{pR}\cdot\mathbf{V}\cdot\mathbf{h}_{pR} = \mathbf{h}'_{pR}\cdot\mathbf{VR}\cdot\mathbf{h}_{pR}
$$
组合方差也由四部分组成：
* $\sigma_B$：基准的方差；
* $2\cdot\beta_{pA}\cdot\sigma_B^2$：主动 beta 产生的协方差；
* $\beta^2_{pA}\cdot\sigma_B^2$：主动 beta 产生的方差；
* $\omega_p^2$：资产选择产生的方差。

主动组合管理人主要关心组合的主动成份，所以还可以将上述变量分解为基准（Benchmark）和主动（Active）部分，其中主动部分包括了基准择时和资产选择两部分。

**权重向量分解**：
$$
\mathbf{h}_p = \mathbf{h}_B + \mathbf{h}_{pA}
$$
其中，$\mathbf{h}_{pA} = \beta_{pA}\cdot\mathbf{h}_B + \mathbf{h}_{pR}$ 为主动组合权重向量。

**预期超额收益率分解**：
$$
f_p = f_B + f_{pA}
$$
其中，
$$
f_{pA} = \mathbf{f}'\mathbf{h}_{pA} = \beta_{pA}\cdot\mu_B + \beta_{pA}\cdot\Delta f_B + \alpha_p
$$
为预期的超常收益率。

**组合方差分解**：
$$
\sigma_p^2 = \sigma^2_B + 2\cdot\beta_{pA}\cdot\sigma_B^2 + \psi_p^2
$$
其中，
$$
\psi_p^2 = \mathbf{h}'_{pA}\cdot\mathbf{V}\cdot\mathbf{h}_{pA} = \beta^2_{pA}\cdot\sigma_B^2 + \omega_p^2
$$
为主动方差，$\psi_p$ 称为主动风险或者跟踪误差。

## 效用函数

均值方差分析中使用的效用函数通常为二次效用函数，Merton 认为在较短的时间尺度上，可以用二次效用函数近似代表一般的效用函数。

组合 p 的预期总体效用为：
$$
\operatorname{U}(\mathbf{h}_p) = f_p - \lambda_T\cdot\sigma_p^2 = \mathbf{f}'\mathbf{h}_p - \lambda_T\cdot\mathbf{h}'_p\mathbf{V}\mathbf{h}_p
$$
其中，$\lambda_T$ 为总体风险厌恶系数。

如果对于不同的风险有不同的风险厌恶系数，则可以将效用分解：
* 基准部分：$f_B - \lambda_T\cdot\sigma_B^2$，与预测（Forecast）有关，与行为（Action）无关；
* 交叉项：$\beta_{pA}\left(\mu_B - 2\lambda_T\cdot\sigma_B^2\right)$，与预测无关，与行为有关；
* 基准择时：$\beta_{pA}\cdot\Delta f_B - \lambda_{BT}\cdot\beta^2_{pA}\cdot\sigma_B^2$，与预测与行为均有关；
* 资产选择：$\alpha_{p} - \lambda_R\cdot\omega_p^2$，与预测与行为均有关；
其中，$\lambda_T$ 为总体风险厌恶系数，$\lambda_{BT}$ 为基准择时风险厌恶系数，$\lambda_R$ 为残差风险厌恶系数。

[1] 中证明，当预期超额收益率等于一致预期 $\mathbf{f}=\boldsymbol{\mu}$ 时，投资者应该选择基准组合 B，从而总体风险厌恶系数的取值应为：
$$
\lambda_T = \frac{\mu_B}{2\sigma_B^2}
$$
从而，交叉项恒等于 0。基准部分不受主动组合管理人影响，从而主动组合管理人的预期效用或者 **附加值（Value Added, VA）** 定义为：
$$
\operatorname{VA}(\mathbf{h}_p) = \left\{\beta_{pA}\cdot\Delta f_B - \lambda_{BT}\cdot\beta^2_{pA}\cdot\sigma_B^2\right\} + \left\{\alpha_{p} - \lambda_R\cdot\omega_p^2\right\}
$$

# 特征组合

资产拥有许多**特征（Attribute）**，例如贝塔、预期收益率、盈市率（Earnings-to-Price Ratio, EP）、市值、所属行业板块等。用向量 $\mathbf{a}=(a_1,a_2,\ldots,a_N)'$ 表示资产的任意一个特征，投资组合 $\mathbf{h}_p$ 对特征 $\mathbf{a}$ 的**暴露（Exposure）**定义为：
$$
a_p=\sum\limits_{n=1}^{N}a_nh_{n,p}=\mathbf{a}'\mathbf{h}_p
$$

定义特征 $\mathbf{a}$ 的**特征组合（Characteristic Portfolio）**为对 $\mathbf{a}$ 具有单位暴露度的最小风险投资组合：
$$
\begin{aligned}
\mathbf{h}_a = \ & \mathop{\arg\min}_{\mathbf{h}}\mathbf{h}'\mathbf{V}\mathbf{h}\\
& \rm{s.t.}\ \mathbf{a}'\mathbf{h}=1
\end{aligned}
$$
由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint)可得：
$$
\mathbf{h}_a = \frac{\mathbf{V}^{-1}\bf{a}}{\bf{a}'\mathbf{V}^{-1}\bf{a}}
$$
特征组合并不一定是全额投资组合，它可以包含多头头寸和空头头寸，并带有杠杆。特征组合可以充分并且唯一地代表定义它的特征。换言之，这个特征组合能够最有效地表达该特征。

这个过程是可逆的。可以从一个组合开始，找到以其作为特征组合的那个特征。对于特征 $\mathbf{a}$ 的特征组合 $\mathbf{h}_a$，全体资产对于该组合的 beta 即为特征 $\mathbf{a}$：
$$
\boldsymbol{\beta}_a=\frac{\mathbf{V}\mathbf{h}_a}{\sigma^2_a}=\frac{\mathbf{V}\mathbf{h}_a}{\mathbf{h}_a'\mathbf{V}\mathbf{h}_a}=\bf{a}
$$

特征和投资组合之间具有一一对应关系：
$$
\begin{aligned}
& \mathbf{h}_a = \frac{\mathbf{V}^{-1}\mathbf{a}}{\mathbf{a}'\mathbf{V}^{-1}\mathbf{a}}\\
& \mathbf{a} = \frac{\mathbf{V}\mathbf{h}_a}{\mathbf{h}_a'\mathbf{V}\mathbf{h}_a}\\
& \mathbf{a}'\mathbf{h}_a = 1
\end{aligned}
$$

特征组合有如下性质：
1. 特征组合的方差为：
$$
\sigma^2_a=\mathbf{h}'_a\mathbf{V}\mathbf{h}_a=\frac{1}{\mathbf{a}'\mathbf{V}^{-1}\mathbf{a}}
$$

2. 对于两个特征 $\mathbf{a},\mathbf{b}$，其对应的特征组合分别为 $\mathbf{h}_a,\mathbf{h}_b$，则这两个特征组合的协方差为：
$$
\sigma^2_{a,b} = \mathbf{h}'_a\mathbf{V}\mathbf{h}_b=\frac{\mathbf{a}'\mathbf{V}^{-1}\mathbf{b}}{\mathbf{a}'\mathbf{V}^{-1}\mathbf{a}\cdot\mathbf{b}'\mathbf{V}^{-1}\mathbf{b}}=a_b\sigma_a^2 = b_a\sigma_b^2
$$
相关系数为：
$$
\rho_{a,b} = \frac{\mathbf{h}'_a\mathbf{V}\mathbf{h}_b}{\sqrt{\mathbf{h}'_a\mathbf{V}\mathbf{h}_a}\cdot\sqrt{\mathbf{h}'_b\mathbf{V}\mathbf{h}_b}} = \frac{\mathbf{a}'\mathbf{V}^{-1}\mathbf{b}}{\sqrt{\mathbf{a}'\mathbf{V}^{-1}\mathbf{a}}\cdot\sqrt{\mathbf{b}'\mathbf{V}^{-1}\mathbf{b}}}
$$
其中，$a_b=\mathbf{a}'\mathbf{h}_b,\ b_a=\mathbf{b}'\mathbf{h}_a$ 分别为特征组合 $\mathbf{h}_b$ 对特征 $\mathbf{a}$ 的暴露和特征组合 $\mathbf{h}_a$ 对特征 $\mathbf{b}$ 的暴露。

3. 对于任意的实数 $\kappa$，特征 $\kappa\mathbf{a}$ 的特征组合为 $\frac{\mathbf{h}_a}{\kappa}$。

4. 两个特征 $\mathbf{a},\mathbf{b}$ 的线性组合特征 $\mathbf{d} = \kappa_a\mathbf{a}+\kappa_b\mathbf{b}$ 的特征组合为：
$$
\frac{\mathbf{h}_d}{\sigma_d^2} = \kappa_a\frac{\mathbf{h}_a}{\sigma_a^2} + \kappa_b\frac{\mathbf{h}_b}{\sigma_b^2}
$$
其中，
$$
\frac{1}{\sigma_d^2} = \left(\frac{\kappa_a}{\sigma_a}\right)^2 + \left(\frac{\kappa_b}{\sigma_b}\right)^2 + 2\rho_{a,b}\frac{\kappa_a}{\sigma_a}\frac{\kappa_b}{\sigma_b} = \frac{\kappa_ad_a}{\sigma^2_a} + \frac{\kappa_bd_b}{\sigma^2_b}
$$

5. 两个特征 $\mathbf{a},\mathbf{b}$ 的特征组合为 $\mathbf{h}_a,\mathbf{h}_b$，则对于特征 $\mathbf{b}$，特征 $\mathbf{a}$ 可以分解成两部分：
$$
\mathbf{a} = a_b\mathbf{b} + \mathbf{d}
$$
特征 $\mathbf{d}$ 与 $\mathbf{b}$ 不相关（即对应的特征组合不相关）,并且有：
$$
d_b=\mathbf{d}'\mathbf{h}_b=0,\ b_d=\mathbf{b}'\mathbf{h}_d=0
$$

特征组合的机制使得能够通过计算一个组合与特征组合的协方差，来确定该组合对相应特征的暴露。一旦建立了特征和投资组合之间的一一对应关系，CAPM 就变为一个关于特征——“预期超额收益率”的特征组合的经济命题。

## 最小风险组合 C
组合 p 在风险资产上的投资比例等于其对特征全 1 向量 $\boldsymbol{\iota}$ 的暴露：$\iota_p = \boldsymbol{\iota}'\mathbf{h}_p$

取特征为全 1 向量 $\boldsymbol{\iota}$，则对应的特征组合为最小风险组合 C，即具有最小风险的全额投资组合：
$$
\mathbf{h}_C = \frac{\mathbf{V}^{-1}\boldsymbol{\iota}}{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}
$$

组合 C 具有性质：
1. 组合 C 具有全局最小方差：
$$
\sigma_C^2=\mathbf{h}'_C\mathbf{V}\mathbf{h}_C=\frac{1}{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}
$$

2. 任一组合 $\mathbf{h}_p$ 和最小方差组合的协方差为：
$$
\sigma^2_{p,C} = \iota_p\sigma_C^2
$$
从而，全额投资组合与最小方差组合的协方差恒等于 $\sigma_C^2$。

3. 全体资产对于组合 C 的 beta 均为 1：
$$
\boldsymbol{\beta}_C=\frac{\mathbf{V}\mathbf{h}_C}{\sigma^2_C}=\boldsymbol{\iota}
$$

## 基准组合 B
资产对于基准组合 B 的 **beta** 定义为资产的超额收益率与基准超额收益率的协方差与基准收益率方差的比：
$$
\boldsymbol{\beta}=\frac{\mathbf{V}\mathbf{h}_B}{\sigma_B^2} = \frac{\mathbf{V}\mathbf{h}_B}{\mathbf{h}'_B\mathbf{V}\mathbf{h}_B}
$$

取特征为全体资产对于基准组合 B 的 beta，则其对应的特征组合恰好为基准组合 B：
$$
\mathbf{h}_B = \frac{\mathbf{V}^{-1}\boldsymbol{\beta}}{\boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta}}
$$
从而，基准组合就是 beta 为 1 的最小方差组合。

组合 B 具有性质：
1. 基准组合 B 的方差：
$$
\sigma_B^2 = \mathbf{h}'_B\mathbf{V}\mathbf{h}_B = \frac{1}{\boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta}}
$$

2. 任一组合 $\mathbf{h}_p$ 和基准组合 B 的协方差为：
$$
\sigma^2_{p,B} = \beta_p\sigma_B^2
$$

## 预期超额收益特征组合 q

取特征为全体资产的预期超额收益率 $\mathbf{f}$，则其对应的特征组合为：
$$
\mathbf{h}_q = \frac{\mathbf{V}^{-1}\mathbf{f}}{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}}
$$

组合 q 有如下性质：
1. 组合 q 的方差为：
$$
\sigma_q^2 = \mathbf{h}'_q\mathbf{V}\mathbf{h}_q = \frac{1}{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}}
$$

2. 任一组合 $\mathbf{h}_p$ 和组合 q 的协方差为：
$$
\sigma^2_{p,q} = f_p\sigma_q^2
$$

3. 组合 q 具有最大夏普率：
$$
SR_q = \max\{SR_p|p\} = \sqrt{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}}
$$
其中，投资组合 p 的 **夏普率（Sharpe Ratio）** 定义为组合的预期超额收益率与组合波动率的比：
$$
SR_p = \frac{f_p}{\sigma_p}
$$

4. 组合 p 具有最大夏普率当且仅当 $\mathbf{h}_p$ 可以表示为：
$$
\mathbf{h}_p = f_p\mathbf{h}_q,\ f_p>0
$$

4. 假设组合 p 具有最大夏普率，即 $SR_p=SR_q$，则组合 p 可表示为：
$$
\mathbf{h}_p = f_p\mathbf{h}_q
$$

5. 假设组合 p 和组合 q 收益率的相关系数 $\operatorname{Corr}(r_p, r_q)$ 则组合 p 的夏普率可以表示为：
$$
SR_p = SR_q\cdot\operatorname{Corr}(r_p, r_q)
$$

6. 组合 q 在风险资产上的投资比例为：
$$
\iota_q = \frac{f_C\sigma_q^2}{\sigma_C^2}
$$

## alpha 特征组合 A
取特征为全体资产的 alpha 向量 $\boldsymbol{\alpha}$，则其对应的特征组合为：
$$
\mathbf{h}_A = \frac{\mathbf{V}^{-1}\boldsymbol{\alpha}}{\boldsymbol{\alpha}'\mathbf{V}^{-1}\boldsymbol{\alpha}}
$$

alpha 特征组合 A 有如下性质：
1. 组合 A 的方差和残差方差相等，均为：
$$
\sigma_A^2 = \omega^2_A = \mathbf{h}'_A\mathbf{V}\mathbf{h}_A = \frac{1}{\boldsymbol{\alpha}'\mathbf{V}^{-1}\boldsymbol{\alpha}}
$$

2. 任一组合 $\mathbf{h}_p$ 和组合 A 的协方差为：
$$
\sigma^2_{p,A} = \alpha_p\sigma_A^2
$$

3. 组合 A 的 beta 为 0: $\beta_A=0$，组合 A 与基准组合 B 不相关：
$$
\sigma^2_{A, B} = \alpha_B\sigma_A^2 = \beta_A\sigma_B^2 = 0
$$

4. 组合 A 可以由组合 q 和基准组合 B 来表达：
$$
\begin{aligned}
& \mathbf{h}_A = \frac{\sigma_A^2}{\sigma_q^2}\mathbf{h}_q - f_B\frac{\sigma_A^2}{\sigma_B^2}\mathbf{h}_B\\
& \frac{1}{\sigma^2_A} = \frac{1}{\sigma_q^2}+\frac{f_B^2}{\sigma_B^2} - 2\rho_{q,B}\frac{1}{\sigma_q}\frac{f_B}{\sigma_B}
\end{aligned}
$$

5. 组合 A 具有最大信息率：
$$
IR_A = \max\{IR_p|p\} = \sqrt{\boldsymbol{\alpha}'\mathbf{V}^{-1}\boldsymbol{\alpha}}
$$
其中，投资组合 p 的 **信息率（Information Ratio）** 定义为组合的 alpha 与残差风险的比：
$$
IR_p = \frac{\alpha_p}{\omega_p}
$$

6. 组合 p 具有最大信息率当且仅当 $\mathbf{h}_p$ 可以表示为基准组合 B 与 alpha 特征组合 A 的投资组合：
$$
\mathbf{h}_p = \beta_p\mathbf{h}_B + \alpha_p\mathbf{h}_A,\ \alpha_p>0
$$
或者其残差组合与 alpha 特征组合 A 成比例：
$$
\mathbf{h}_{pR} = \alpha_p\mathbf{h}_A,\ \alpha_p>0
$$

7. 假设组合 p 和组合 A 残差收益率的相关系数为 $\operatorname{Corr}(\theta_p, \theta_A)$ 则组合 p 的信息率可以表示为：
$$
IR_p = IR_A\cdot\operatorname{Corr}(\theta_p, \theta_A)
$$

8. 组合 A 在风险资产上的投资比例为：
$$
\iota_A = \frac{\alpha_C\omega_A^2}{\sigma_C^2}
$$

9. 资产的 alpha $\boldsymbol{\alpha}$ 可以表示为：
$$
\boldsymbol{\alpha} = IR_A\cdot\left(\frac{\mathbf{V}\mathbf{h}_A}{\omega_A}\right)
$$

性质 3 是特征组合性质 5 的一个特例，考虑特征: 预期超额收益率 $\mathbf{f}$ 和相对于基准组合 B 的 beta $\boldsymbol{\beta}$，则预期超额收益率可以分解为：
$$
\mathbf{f} = f_B\boldsymbol{\beta} + \boldsymbol{\alpha}
$$
则 $\boldsymbol{\beta}$ 和 $\boldsymbol{\alpha}$ 不相关，即基准组合 B 与组合 A 不相关，并且有：$\alpha_B=0,\ \beta_A=0$。

性质 4 是特征组合性质 4 的一个特例。

## 组合 Q
假设最小方差组合的预期超额收益率大于 0：$f_C>0$，此时组合 q 是净多头（即在风险资产上的投资比例大于 0）：
$$
\iota_q=\frac{f_C \sigma_q^2}{\sigma_C^2}>0
$$

取特征为：$\iota_q\mathbf{f}$，则其对应的特征组合定义为组合 Q：
$$
\mathbf{h}_Q = \frac{1}{\iota_q}\frac{\mathbf{V}^{-1}\mathbf{f}}{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}}=\frac{\mathbf{h}_q}{\iota_q}
$$
组合 Q 有如下性质：
1. 组合 Q 的方差为：
$$
\sigma_Q^2=\frac{\sigma_q^2}{\iota_q^2}
$$

2. 任一组合 $\mathbf{h}_p$ 和组合 Q 的协方差为：
$$
\sigma_{p,Q}^2 = \frac{f_p}{f_Q}\sigma_Q^2
$$

3. 组合 Q 的预期超额收益率：
$$
f_Q = \frac{f_q}{\iota_q} = \frac{1}{\iota_q}
$$

4. 组合 Q 是全额投资组合：
$$
\iota_Q = 1
$$

5. 组合 Q 具有最大夏普率：
$$
SR_Q = SR_q
$$

6. 假设组合 p 和组合 Q 收益率的相关系数为 $\operatorname{Corr}(r_p, r_Q)$ 则组合 p 的夏普率可以表示为：
$$
SR_p = SR_Q\cdot\operatorname{Corr}(r_p, r_Q)
$$

7. 组合 Q 具有最大信息率：
$$
IR_Q = IR_A
$$
且：
$$
\begin{aligned}
& \mathbf{h}_Q = \beta_Q\mathbf{h}_B + \alpha_Q\mathbf{h}_A\\
& \beta_Q = \frac{f_B\cdot\sigma_Q^2}{f_Q\cdot\sigma_B^2}\\
& \alpha_Q = \frac{\sigma_Q^2}{f_Q\omega_A^2}
\end{aligned}
$$

8. 假设组合 p 和组合 Q 残差收益率的相关系数为 $\operatorname{Corr}(\theta_p, \theta_Q)$ 则组合 p 的信息率可以表示为：
$$
IR_p = IR_Q\cdot\operatorname{Corr}(\theta_p, \theta_Q)
$$

9. 组合 Q 和组合 C 有相同的收益方差比：
$$
\frac{f_Q}{\sigma_Q^2} = \frac{f_C}{\sigma_C^2}
$$

10. 资产的预期超额收益率 $\mathbf{f}$ 可以完全由组合 Q 来解释：
$$
\mathbf{f} = f_Q\cdot\boldsymbol{\beta}^Q
$$
其中：$\boldsymbol{\beta}^Q$ 为所有资产相对于组合 Q 的 beta。

11. 资产的 alpha $\boldsymbol{\alpha}$ 可以完全由组合 Q 来解释：
$$
\boldsymbol{\alpha} = IR_Q\cdot MCRR_Q
$$
其中，$MCRR_Q$ 为资产对组合 Q 残差风险的边际贡献：
$$
MCRR_Q = \frac{\partial \omega_Q}{\partial \mathbf{h}_{QR}}
$$

最大夏普率和最大信息率有如下关系：
$$
\begin{aligned}
& IR = SR\cdot\frac{\omega_Q}{\sigma_Q}\\
& SR^2 = SR_B^2 + IR^2
\end{aligned}
$$

## 多特征组合

对于 K 个特征 $\mathbf{A} = (\mathbf{a}_1,\ldots,\mathbf{a}_K)$，也可以类似定义其特征组合为对 K 个特征均具有单位暴露的最小方差组合：
$$
\begin{aligned}
\mathbf{h}_{\mathbf{A}} = \ & \mathop{\arg\min}_{\mathbf{h}}\mathbf{h}'\mathbf{V}\mathbf{h}\\
& \rm{s.t.}\ \mathbf{A}'\mathbf{h} = \boldsymbol{\iota}
\end{aligned}
$$
由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint)可得：
$$
\mathbf{h}_{\mathbf{A}} = \mathbf{V}^{-1}\mathbf{A}(\mathbf{A}'\mathbf{V}^{-1}\mathbf{A})^{-1}\boldsymbol{\iota}
$$
$\mathbf{h}_{\mathbf{A}}$ 还可以表达为 K 个单特征组合的线性组合：
$$
\mathbf{h}_{\mathbf{A}} = \sum\limits_{k=1}^{K}\lambda_k\mathbf{V}^{-1}\mathbf{a}_k = \sum\limits_{k=1}^{K}\frac{\lambda_k}{\sigma_{\mathbf{a}_k}^2}\mathbf{h}_{\mathbf{a}_k}
$$
其中，$\boldsymbol{\lambda} = (\lambda_1,\ldots,\lambda_K)'$ 是 Lagrange 乘子：
$$
\boldsymbol{\lambda} = (\mathbf{A}'\mathbf{V}^{-1}\mathbf{A})^{-1}\boldsymbol{\iota} = \begin{pmatrix} \mathbf{a}'_1\mathbf{V}^{-1}\mathbf{a}_1 & \cdots & \mathbf{a}'_1\mathbf{V}^{-1}\mathbf{a}_K \\ \vdots & \ddots & \vdots \\ \mathbf{a}'_K\mathbf{V}^{-1}\mathbf{a}_1 & \cdots & \mathbf{a}'_K\mathbf{V}^{-1}\mathbf{a}_K \end{pmatrix}^{-1}\boldsymbol{\iota}
$$

组合的方差为：
$$
\sigma^2_{\mathbf{A}} = \boldsymbol{\iota}'(\mathbf{A}'\mathbf{V}^{-1}\mathbf{A})^{-1}\boldsymbol{\iota}
$$

# 有效前沿组合

## 总收益风险的有效前沿

总收益风险的有效前沿是在预期超额收益率和总风险框架下考虑最优投资问题。

### 无约束有效前沿

无约束 **有效前沿组合（Efficient Frontier Portfolio）** 定义为给定预期组合收益下的最小方差组合，所有的有效前沿组合组成 **有效前沿（Efficient Frontier）**。无约束有效前沿组合 $\mathbf{h}_p$ 是如下优化问题的解：
$$
\begin{aligned}
\mathbf{h}_p = \ & \mathop{\arg\min}_{\mathbf{h}}\mathbf{h}'\mathbf{V}\mathbf{h}\\
\rm{s.t.}\ &\mathbf{f}'\mathbf{h}=f_p
\end{aligned}
$$

由此可见无约束有效前沿组合是特征 $\frac{\mathbf{f}}{f_p}$ 的特征组合，即：
$$
\mathbf{h}_p = f_p\frac{\mathbf{V}^{-1}\mathbf{f}}{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}} = f_p\mathbf{h}_q = \frac{f_p}{f_Q}\mathbf{h}_Q
$$

无约束有效前沿组合也可以定义为给定风险厌恶系数 $\lambda_p$ 的最大二次效用组合，即如下优化问题的解：
$$
\mathbf{h}_p = \ \mathop{\arg\max}_{\mathbf{h}}\mathbf{f}'\mathbf{h}-\lambda_p\mathbf{h}'\mathbf{V}\mathbf{h}
$$
解得：
$$
\mathbf{h}_p = \frac{1}{2\lambda_p}\mathbf{V}^{-1}\mathbf{f} = \frac{1}{2\lambda_p\sigma_q^2}\mathbf{h}_q = \frac{f_Q}{2\lambda_p\sigma_Q^2}\mathbf{h}_Q
$$
其对应的预期组合收益率为：
$$
f_p = \frac{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}}{2\lambda_p} = \frac{1}{2\lambda_p\sigma_q^2} = \frac{1}{2\lambda_p}\left(\frac{f_Q}{\sigma_Q}\right)^2 = \frac{SR_Q^2}{2\lambda_p}
$$

无约束有效前沿组合有如下性质：
1. 无约束有效前沿组合在风险资产上的投资比例为：
$$
\iota_p = f_p\iota_q = \frac{f_p}{f_Q}
$$

2. 组合 Q 是无约束有效前沿组合。

3. 无约束有效前沿组合的方差为：
$$
\sigma_p^2 = \frac{f_p^2}{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}} = f_p^2\sigma_q^2 = \left(\frac{f_p}{f_Q}\right)^2\sigma_Q^2
$$

无约束有效前沿在总风险-预期超额收益率平面 $(\sigma_p, f_p)$ 上的图像是两条过原点上下对称的直线：
$$
|f_p| = \frac{\sigma_p}{\sigma_q} = SR_q\sigma_p
$$
即有效前沿的斜率为最大夏普率。

### 全额投资有效前沿

全额投资有效前沿组合定义为给定预期组合收益下的全额投资最小方差组合，即如下优化问题的解：
$$
\begin{aligned}
\mathbf{h}_p = \ & \mathop{\arg\min}_{\mathbf{h}}\mathbf{h}'\mathbf{V}\mathbf{h}\\
\rm{s.t.}\ &\boldsymbol{\iota}'\mathbf{h}=1\\
& \mathbf{f}'\mathbf{h}=f_p
\end{aligned}
$$
由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint) 或者多特征组合的表达式可解得：
$$
\begin{aligned}
\mathbf{h}_p &= \mathbf{V}^{-1}\begin{pmatrix} \boldsymbol{\iota} & \mathbf{f}\end{pmatrix}\mathbf{M}^{-1}\begin{pmatrix}1 \\ f_p\end{pmatrix}\\
&= \frac{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f} - (\mathbf{f}'\mathbf{V}^{-1}\boldsymbol{\iota})f_p}{|\mathbf{M}|}\mathbf{V}^{-1}\boldsymbol{\iota} + \frac{(\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota})f_p - \boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{|\mathbf{M}|}\mathbf{V}^{-1}\mathbf{f}
\end{aligned}
$$
其中，
$$
\begin{aligned}
& \mathbf{M} = \begin{pmatrix} \boldsymbol{\iota}' \\ \mathbf{f}' \end{pmatrix} \mathbf{V}^{-1} \begin{pmatrix} \boldsymbol{\iota} & \mathbf{f} \end{pmatrix} = \begin{pmatrix} \boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota} & \boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f} \\ \mathbf{f}'\mathbf{V}^{-1}\boldsymbol{\iota} & \mathbf{f}'\mathbf{V}^{-1}\mathbf{f} \end{pmatrix}\\
& \mathbf{M}^{-1} = \frac{1}{|\mathbf{M}|}\begin{pmatrix} \mathbf{f}'\mathbf{V}^{-1}\mathbf{f} & -\mathbf{f}'\mathbf{V}^{-1}\boldsymbol{\iota} \\ -\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f} & \boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota} \end{pmatrix}\\
& |\mathbf{M}| = (\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota})\cdot(\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}) - (\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f})^2
\end{aligned}
$$

全额投资有效前沿组合也可以定义为给定风险厌恶系数 $\lambda_p$ 的全额投资最大二次效用组合，即如下优化问题的解：
$$
\begin{aligned}
\mathbf{h}_p = \ & \mathop{\arg\max}_{\mathbf{h}}\mathbf{f}'\mathbf{h}-\lambda_p\mathbf{h}'\mathbf{V}\mathbf{h}\\
\rm{s.t.}\ &\boldsymbol{\iota}'\mathbf{h}=1
\end{aligned}
$$
由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint)可解得：
$$
\mathbf{h}_p = \frac{1}{2\lambda_p}\mathbf{V}^{-1}\mathbf{f} + \left(1-\frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{2\lambda_p}\right)\frac{\mathbf{V}^{-1}\boldsymbol{\iota}}{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}
$$
其对应的预期组合收益率为：
$$
f_p=\frac{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}}{2\lambda_p} + \left(1-\frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{2\lambda_p}\right)\frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}
$$

全额投资有效前沿组合有如下性质：
1. 最小风险组合 C，组合 Q 均为全额投资有效前沿组合。

2. 任意选定两个不同的全额投资有效前沿组合，则其他的全额投资有效前沿组合可以表示成选定组合的线性组合。如果选择最小方差组合 C 和组合 Q 来生成全额投资有效前沿组合，则：
$$
\mathbf{h}_p = \frac{f_Q-f_p}{f_Q-f_C}\mathbf{h}_C + \frac{f_p-f_C}{f_Q-f_C}\mathbf{h}_Q = \left(1-\frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{2\lambda_p}\right)\mathbf{h}_C + \frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{2\lambda_p}\mathbf{h}_Q
$$
从上式可知：
$$
\frac{f_p-f_C}{f_Q-f_C} = \frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{2\lambda_p}
$$

3. 全额投资有效前沿组合的全额投资组合仍然为全额投资有效前沿组合。

4. 全额投资有效前沿组合的内蕴特征为：
$$
\mathbf{a}_p = \frac{\sigma_C^2}{\sigma_p^2}\frac{f_Q-f_p}{f_Q-f_C}\boldsymbol{\iota} + \frac{\sigma_Q^2}{\sigma_p^2}\frac{f_p-f_C}{f_Q-f_C}\iota_q\mathbf{f}
$$

5. 全额投资有效前沿组合的方差为：
$$
\sigma_p^2=\left[1-\left(\frac{f_p-f_C}{f_Q-f_C}\right)^2\right]\sigma_C^2+\left(\frac{f_p-f_C}{f_Q-f_C}\right)^2\sigma_Q^2
$$

6. 任意两个全额投资有效前沿组合 p 和组合 r 的协方差为：
$$
\begin{aligned}
\sigma_{p,r}^2 &= \begin{pmatrix} 1 & f_p \end{pmatrix} \mathbf{M}^{-1} \begin{pmatrix} 1 \\ f_r \end{pmatrix}\\
&= \frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}{|\mathbf{M}|}\left(f_p-\frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}\right)\left(f_r-\frac{\boldsymbol{\iota}'\mathbf{V}^{-1}\mathbf{f}}{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}\right)+\frac{1}{\boldsymbol{\iota}'\mathbf{V}^{-1}\boldsymbol{\iota}}\\
&= \frac{(f_p-f_C)(f_r-f_C)}{\sigma^2_C|\mathbf{M}|}+\sigma_C^2
\end{aligned}
$$

7. 对于任意的全额投资有效前沿组合 $p$（非最小方差组合），存在唯一的全额投资有效前沿组合 $op$，使得组合 $p$ 和组合 $op$ 不相关，组合 $op$ 称为组合 $p$ 的 **零 beta 组合（Zero-Beta Portfolio）**。

8. 给定全额投资有效前沿组合 $p$（非最小方差组合），其零 beta 组合为 $op$，则对任意的全额投资有效前沿组合 s，有回归方程：
$$
R_s = \beta^p_sR_p + (1-\beta^p_s)R_{op}+\varepsilon
$$
另外：
$$
f_s = \beta^p_sf_p + (1-\beta^p_s)f_{op}
$$


全额投资有效前沿在风险-预期超额收益率平面 $(\sigma_p,f_p)$ 上的图像是双曲线（右支）：
$$
\frac{\sigma^2_p}{\sigma_C^2} - \frac{(f_p-f_C)^2}{\sigma_C^2\frac{(f_Q-f_C)^2}{\sigma_Q^2-\sigma_C^2}} = 1
$$
其顶点即是最小方差组合 C：$(\sigma_C, f_C)$，过原点做双曲线的切线，其切点即为组合 Q：$(\sigma_Q, f_Q)$，即无约束有效前沿与全额投资有效前沿相切于组合 Q $(\sigma_Q, f_Q)$。

### 单位 beta 有效前沿

单位 beta 有效前沿组合定义为给定预期组合收益下 beta 等于 1 的最小方差组合，即如下优化问题的解：
$$
\begin{aligned}
\mathbf{h}_p = \ & \mathop{\arg\min}_{\mathbf{h}}\mathbf{h}'\mathbf{V}\mathbf{h}\\
\rm{s.t.}\ &\boldsymbol{\beta}'\mathbf{h} = 1\\
& \mathbf{f}'\mathbf{h} = f_p
\end{aligned}
$$
由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint) 或者多特征组合的表达式可解得：
$$
\begin{aligned}
\mathbf{h}_p &= \mathbf{V}^{-1}\begin{pmatrix} \boldsymbol{\beta} & \mathbf{f} \end{pmatrix}\mathbf{M}^{-1}\begin{pmatrix} 1 \\ f_p \end{pmatrix}\\
&= \frac{\mathbf{f}'\mathbf{V}^{-1}\mathbf{f} - (\mathbf{f}'\mathbf{V}^{-1}\boldsymbol{\beta})f_p}{|\mathbf{M}|}\mathbf{V}^{-1}\boldsymbol{\beta} + \frac{(\boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta})f_p - \boldsymbol{\beta}'\mathbf{V}^{-1}\mathbf{f}}{|\mathbf{M}|}\mathbf{V}^{-1}\mathbf{f}
\end{aligned}
$$
其中，
$$
\begin{aligned}
& \mathbf{M} = \begin{pmatrix} \boldsymbol{\beta}' \\ \mathbf{f}' \end{pmatrix} \mathbf{V}^{-1} \begin{pmatrix} \boldsymbol{\beta} & \mathbf{f} \end{pmatrix} = \begin{pmatrix} \boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta} & \boldsymbol{\beta}'\mathbf{V}^{-1}\mathbf{f} \\ \mathbf{f}'\mathbf{V}^{-1}\boldsymbol{\beta} & \mathbf{f}'\mathbf{V}^{-1}\mathbf{f} \end{pmatrix}\\
& \mathbf{M}^{-1} = \frac{1}{|\mathbf{M}|}\begin{pmatrix} \mathbf{f}'\mathbf{V}^{-1}\mathbf{f} & -\mathbf{f}'\mathbf{V}^{-1}\boldsymbol{\beta} \\ -\boldsymbol{\beta}'\mathbf{V}^{-1}\mathbf{f} & \boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta} \end{pmatrix}\\
& |\mathbf{M}| = (\boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta})\cdot(\mathbf{f}'\mathbf{V}^{-1}\mathbf{f}) - (\boldsymbol{\beta}'\mathbf{V}^{-1}\mathbf{f})^2
\end{aligned}
$$

单位 beta 有效前沿组合也可以定义为给定风险厌恶系数 $\lambda_p$ 的单位 beta 最大二次效用组合，即如下优化问题的解：
$$
\begin{aligned}
\mathbf{h}_p = \ & \mathop{\arg\max}_{\mathbf{h}}\mathbf{f}'\mathbf{h}-\lambda_p\mathbf{h}'\mathbf{V}\mathbf{h}\\
\rm{s.t.}\ &\boldsymbol{\beta}'\mathbf{h} = 1
\end{aligned}
$$
由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint)可解得：
$$
\mathbf{h}_p = \frac{\mathbf{V}^{-1}\boldsymbol{\beta}}{\boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta}} + \frac{1}{2\lambda_p}\mathbf{V}^{-1}\left(\mathbf{f} - \frac{\mathbf{f}'\mathbf{V}^{-1}\boldsymbol{\beta}}{\boldsymbol{\beta}'\mathbf{V}^{-1}\boldsymbol{\beta}}\boldsymbol{\beta}\right) = \mathbf{h}_B + \frac{IR^2_A}{2\lambda_p}\mathbf{h}_A
$$
从上式可知，单位 beta 有效前沿组合是最大信息率组合，其对应的预期组合收益率为：
$$
f_p = f_B + \frac{IR_A^2}{2\lambda_p}
$$
组合的 alpha 为：
$$
\alpha_p = \frac{IR_A^2}{2\lambda_p}
$$
从而：
$$
\mathbf{h}_p = \mathbf{h}_B + \alpha_p\mathbf{h}_A
$$

单位 beta 有效前沿组合有如下性质：
1. 基准组合 B 为单位 beta 有效前沿组合。

2. 单位 beta 有效前沿组合的方差为：
$$
\sigma_p^2 = \sigma_B^2 + \alpha_p^2\omega_A^2 = \sigma_B^2 + \frac{IR_A^4}{4\lambda_p^2}\omega_A^2
$$
残差方差为：
$$
\omega_p^2 = \alpha_p^2\omega_A^2 = \frac{IR_A^4}{4\lambda_p^2}\omega_A^2
$$

3. 任意两个单位 beta 有效前沿组合 p 和组合 r 的协方差为：
$$
\begin{aligned}
\sigma_{p,r}^2 &= \begin{pmatrix} 1 & f_p \end{pmatrix} \mathbf{M}^{-1} \begin{pmatrix} 1 \\ f_r \end{pmatrix}\\
&= \sigma_B^2 + \alpha_p\alpha_r\omega_A^2\\
&= \sigma_B^2 + \frac{IR_A^4}{4\lambda_p\lambda_r}\omega_A^2
\end{aligned}
$$

4. 单位 beta 有效前沿组合的内蕴特征为：
$$
\mathbf{a}_p = \frac{\sigma_B^2}{\sigma_p^2}\boldsymbol{\beta} + \frac{\alpha_p\omega_A^2}{\sigma_p^2}\boldsymbol{\alpha}
$$

单位 beta 有效前沿在总风险-预期超额收益率平面 $(\sigma_p,f_p)$ 上的图像是双曲线（右支）：
$$
\frac{\sigma^2_p}{\sigma_B^2} - \frac{(f_p-f_B)^2}{\sigma_B^2IR_A^2} = 1
$$
其顶点即是基准组合 B：$(\sigma_B, f_B)$。

### 资本资产定价模型（CAPM）

假设：
1. 所有资产均纳入模型；
2. 所有投资者有一致的预期超额收益率：$\mathbf{f}$；
3. 所有投资者有同样的协方差矩阵：$\mathbf{V}$；
4. 所有投资者具有均值方差偏好，即选择有效前沿组合作为投资组合
5. 没有交易成本和税负
在上述假设下，一致预期超额收益率可以表示为：
$$
\mathbf{f} = \boldsymbol{\beta}^Qf_Q = \boldsymbol{\beta}\mu_M
$$
即组合 Q 为市场组合。

\begin{proof}
设投资者 j 的财富为 $T_j$，则社会总财富为：$\sum\limits_{j}T_j$。设投资者 j 的投资组合为 $\mathbf{h}_j$，那么市场组合为:

$$
\mathbf{h}_M = \frac{\sum\limits_jT_j\mathbf{h}_j}{\sum\limits_jT_j}
$$

由假设 2~5 可知：$\mathbf{h}_j = \iota_j\mathbf{h}_Q$，其中，$\iota_j=\boldsymbol{\iota}'\mathbf{h}_j$ 为投资者在风险资产上的投资比例，现金持有比例为：$1-\iota_j$，如果 $1-\iota_j\geq0$ 表示投资者 j 持有现金：$T_j(1-\iota_j)$；如果 $1-\iota_j<0$ 表示投资者 j 借入现金：$T_j(\iota_j-1)$。

由假设 1 可知，全社会借贷平衡：
$$
\sum\limits_j(1-\iota_j)T_j=0
$$
从而，$\sum\limits_{j}T_j=\sum\limits_j\iota_jT_j$，从而：
$$
\mathbf{h}_M = \frac{\sum\limits_jT_j\mathbf{h}_j}{\sum\limits_jT_j} = \frac{\sum\limits_jT_j\iota_j\mathbf{h}_Q}{\sum\limits_jT_j} = \mathbf{h}_Q
$$
\end{proof}


## 残差收益风险的有效前沿

残差收益风险的有效前沿是在 alpha 和残差风险框架下考虑最优投资问题。

### 无约束有效前沿

无约束有效前沿组合定义为给定 alpha 下的最小残差风险组合。无约束有效前沿组合 $\mathbf{h}_p$ 是如下优化问题的解：
$$
\begin{aligned}
\mathbf{h}_p = \ & \mathop{\arg\min}_{\mathbf{h}}\omega^2 = \mathbf{h}'\cdot\mathbf{VR}\cdot\mathbf{h}\\
\rm{s.t.}\ &\boldsymbol{\alpha}'\mathbf{h} = \alpha_p
\end{aligned}
$$
由于 $\mathbf{VR}$ 是减秩的半正定矩阵，所以做变量代换 $\mathbf{h}_p\mapsto(\beta_{pA}, \mathbf{h}_{pR})$：
$$
\begin{aligned}
(\beta_{pA}, \mathbf{h}_{pR}) = \ & \mathop{\arg\min}_{\beta_A,\mathbf{h}_R}\omega^2 = \mathbf{h}_R'\cdot\mathbf{V}\cdot\mathbf{h}_R\\
\rm{s.t.}\ &\boldsymbol{\beta}'\mathbf{h}_R = 0\\
\ &\boldsymbol{\alpha}'\mathbf{h}_R = \alpha_p
\end{aligned}
$$
从上式可以看出该优化问题与主动 beta 无关，由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint)可解得：
$$
\mathbf{h}_{pR} = \alpha_p\frac{\mathbf{V}^{-1}\boldsymbol{\alpha}}{\boldsymbol{\alpha}'\mathbf{V}^{-1}\boldsymbol{\alpha}} = \alpha_p\mathbf{h}_A
$$
还原为原变量即为：
$$
\mathbf{h}_p = \beta_{p}\mathbf{h}_B + \alpha_p\mathbf{h}_A
$$
其中，$\beta_{p}$ 可以取任意值，无约束的有效前沿就是最大信息率组合。

无约束有效前沿组合也可以定义为给定风险厌恶系数 $\lambda_R$ 的最大二次效用组合，或者最大残差附加值组合，即如下优化问题的解：
$$
\mathbf{h}_p = \ \mathop{\arg\max}_{\mathbf{h}}\boldsymbol{\alpha}'\mathbf{h}-\lambda_R\mathbf{h}'\cdot\mathbf{VR}\cdot\mathbf{h}
$$
解得：
$$
\mathbf{h}_p = \beta_{p}\mathbf{h}_B + \frac{\boldsymbol{\alpha}'\mathbf{V}^{-1}\boldsymbol{\alpha}}{2\lambda_R}\mathbf{h}_A = \beta_{p}\mathbf{h}_B + \frac{IR^2_A}{2\lambda_R}\mathbf{h}_A
$$
从而：
$$
\alpha_p = \frac{IR^2_A}{2\lambda_R}
$$

无约束有效前沿组合有如下性质：
1. 无约束有效前沿组合在风险资产上的投资比例为：
$$
\iota_p = \alpha_p\iota_A
$$

2. 组合 Q 是无约束有效前沿组合。

3. 无约束有效前沿组合的残差风险为：
$$
\omega_p = \frac{IR_A}{2\lambda_R}
$$

4. 最大附加值为：
$$
VA = \frac{IR_A^2}{4\lambda_R}
$$

无约束有效前沿在残差风险-alpha 平面 $(\omega_p, \alpha_p)$ 上的图像是过原点的直线：
$$
\alpha_p = IR_A\omega_p
$$
其中，有效前沿的斜率为最大信息率，原点代表无风险资产或者基准组合 B。

如果限定组合的 beta 为 1，则在总风险-超额收益率平面 $(\sigma_p, f_p)$ 上无约束残差收益风险有效前沿即是单位 beta 总收益风险有效前沿。
