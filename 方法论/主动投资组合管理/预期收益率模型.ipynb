# 预测基本公式

定义随机向量 $\mathbf{g}$ 的标准差对角阵为：
$$
\operatorname{Std}(\mathbf{g}) = \begin{pmatrix} \sigma_{g_1} & & \\ & \ddots & \\ & & \sigma_{g_K} \end{pmatrix}
$$
则对于 $\mathbf{g}$ 的协方差矩阵有分解：
$$
\operatorname{Var}(\mathbf{g}) = \operatorname{Std}(\mathbf{g})\cdot\operatorname{Corr}(\mathbf{g})\cdot\operatorname{Std}(\mathbf{g})
$$
其中，$\operatorname{Corr}(\mathbf{g})$ 为 $\mathbf{g}$ 的相关系数矩阵：
$$
\operatorname{Corr}(\mathbf{g}) = \operatorname{Std}^{-1}(\mathbf{g})\cdot\operatorname{Var}(\mathbf{g})\cdot\operatorname{Std}^{-1}(\mathbf{g})
$$

## 信息系数（Information Coefficient，IC）

**原始预测（Raw Forecast）**：对收益率有预测作用的信号。通常用 $K\times 1$ 向量表示，记为 $\mathbf{g}$。

预测信号的 **信息系数（Information Coefficient，IC）** 定义为信号和收益率的相关系数：
$$
\mathbf{IC} = \operatorname{Corr}(\mathbf{g},\mathbf{r})
$$

## 标准 z 分值

预测信号 **标准 z 分值（z-score）** 定义为：
$$
\mathbf{z}_g = \operatorname{Std}^{-1}(\mathbf{g})\cdot\left(\mathbf{g}-\mathbb{E}(\mathbf{g})\right)
$$
则 z 分值的期望为 $\mathbf{0}$，标准差对角阵为 $\mathbf{I}$，其协方差矩阵和相关系数矩阵均为原始变量的相关系数矩阵：
$$
\operatorname{Var}(\mathbf{z}_g) = \operatorname{Corr}(\mathbf{z}_g) = \operatorname{Corr}(\mathbf{g})
$$
信号 z 分值对收益率的信息系数等于原始信号的信息系数，即 $\mathbf{IC}_z = \mathbf{IC}_g$。

如果信号是线性无关的，即 $\operatorname{Var}(\mathbf{g})$ 是正定阵，则对信号的相关系数矩阵进行特征分解：
$$
\operatorname{Corr}(\mathbf{g}) = \mathbf{Q}\cdot\boldsymbol{\Lambda}\cdot\mathbf{Q}'
$$
其中，$\mathbf{Q}$ 是正交阵。对原始信号的 z 分值可以进行正交化，得到一组正交信号：
$$
\mathbf{z}^*_g = \boldsymbol{\Lambda}^{-\frac{1}{2}}\mathbf{Q}'\mathbf{z}_g
$$
正交信号 $\mathbf{z}^*_g$ 的期望为 $\mathbf{0}$，协方差矩阵、相关系数矩阵以及标准差对角阵均为单位阵 $\mathbf{I}$，这组正交信号对收益率的信息系数 IC 为：
$$
\mathbf{IC}_{\mathbf{z}^*_g} = \boldsymbol{\Lambda}^{-\frac{1}{2}}\mathbf{Q}'\mathbf{IC}
$$

## 精炼预测

**预测基本公式**：给定预测信号或原始预测 $\mathbf{g}$，对于收益率 $\mathbf{r}$ 调整后的预测量为：
$$
\begin{aligned}
\hat{\mathbf{r}} &= \mathbb{E}(\mathbf{r}) + \operatorname{Cov}(\mathbf{r},\mathbf{g})\cdot\operatorname{Var}^{-1}(\mathbf{g})\cdot\left(\mathbf{g}-\mathbb{E}(\mathbf{g})\right)\\
&= \mathbb{E}(\mathbf{r}) + \operatorname{Std}(\mathbf{r})\cdot\operatorname{Corr}(\mathbf{r},\mathbf{g})\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{z}_g
\end{aligned}
$$
其中，$\mathbb{E}(\mathbf{r})$ 称为 **朴素预测（Naive Forecast）**，通常对于给定的基准组合 B，朴素预测为：$\mathbb{E}(\mathbf{r}) = \boldsymbol{\beta}\cdot\mu_B$。

对于收益率协方差矩阵 $\operatorname{Var}(\mathbf{r})$ 调整后的预测量为：
$$
\begin{aligned}
\widehat{\operatorname{Var}}(\mathbf{r}) &= \operatorname{Var}(\mathbf{r}) - \operatorname{Cov}(\mathbf{r},\mathbf{g})\cdot\operatorname{Var}^{-1}(\mathbf{g})\cdot\operatorname{Cov}(\mathbf{g},\mathbf{r})\\
&= \operatorname{Var}(\mathbf{r}) - \operatorname{Std}(\mathbf{r})\cdot\operatorname{Corr}(\mathbf{r},\mathbf{g})\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\operatorname{Corr}(\mathbf{g},\mathbf{r})\cdot\operatorname{Std}(\mathbf{r})
\end{aligned}
$$

定义 **精炼预测（Refined Forecast）** 为预测修正量：
$$
\Phi = \hat{\mathbf{r}} - \mathbb{E}(\mathbf{r}) = \operatorname{Std}(\mathbf{r})\cdot\operatorname{Corr}(\mathbf{r},\mathbf{g})\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{z}_g
$$

如果记调整 IC 为 $\mathbf{IC}^* = \operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{IC}$，则上述的精炼预测可表示为：
$$
\Phi = \operatorname{Std}(\mathbf{r})\cdot\mathbf{IC}'\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{z}_g = \operatorname{Std}(\mathbf{r})\cdot(\mathbf{IC}^*)'\cdot\mathbf{z}_g
$$

也可以将精练预测表达为正交化信号的形式：
$$
\Phi = \operatorname{Std}(\mathbf{r})\cdot\mathbf{IC}'_{\mathbf{z}^*_g}\cdot\mathbf{z}^*_g
$$

预测基本公式将精练预测表示为：波动率 * IC * 标准分值。

预测基本公式有如下性质：
* $\hat{\mathbf{r}}$ 是 $\mathbf{r}$ 的最优线性无偏估计（BLUE）。
* 如果 $\mathbf{r},\mathbf{g}$ 服从联合正态分布，则 $\hat{\mathbf{r}}$ 是已知 $\mathbf{g}$ 下的 $\mathbf{r}$ 的最大似然估计，并且为最小方差无偏估计（MVUE），更进一步有 $\hat{\mathbf{r}} = \mathbb{E}(\mathbf{r}|\mathbf{g})$，即 $\hat{\mathbf{r}}$ 是已知 $\mathbf{g}$ 下的 $\mathbf{r}$ 的条件期望。
* 如果 $\mathbf{r},\mathbf{g}$ 服从联合正态分布，则 $\widehat{\operatorname{Var}}(\mathbf{r}) = \operatorname{Var}(\mathbf{r}|\mathbf{g})$，即 $\widehat{\operatorname{Var}}(\mathbf{r})$ 是已知 $\mathbf{g}$ 下的 $\mathbf{r}$ 的条件方差。
* 精炼预测的协方差矩阵为：
$$
\operatorname{Var}(\Phi) = \operatorname{Std}(\mathbf{r})\cdot\operatorname{Corr}(\mathbf{r},\mathbf{g})\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\operatorname{Corr}(\mathbf{g},\mathbf{r})\cdot\operatorname{Std}(\mathbf{r})
$$
从而有：
$$
\widehat{\operatorname{Var}}(\mathbf{r}) = \operatorname{Var}(\mathbf{r}) - \operatorname{Var}(\Phi)
$$

## 单资产情形

考虑只有一个资产的情形，记资产的收益率为 $r$，资产的波动率为 $\sigma$，共有 K 个预测信号 $\mathbf{g}=\left(g_1,\ldots,g_K\right)'$，此时预测基本公式为：
$$
\Phi = \sigma\cdot\mathbf{IC}'\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{z}_g
$$
其中，$\mathbf{IC}$ 是 $K\times 1$ 的相关系数向量，$\mathbf{z}_g$ 也是 $K\times 1$ 为标准 z 分值向量。

如果记修正 IC 为 $\mathbf{IC}^* = \operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{IC}$，则精练预测为：
$$
\Phi = \sigma\cdot(\mathbf{IC}^*)'\cdot\mathbf{z}_g = \sigma\cdot\sum\limits_{k=1}^{K}IC^*_k\cdot z_k
$$
即精炼预测等于以调整 IC 加权的原始预测标准分值的线性组合。

精练预测的标准差为：
$$
\sigma_\Phi = \sigma\cdot\sqrt{\mathbf{IC}'\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{IC}}
$$

精练预测作为多个信号的组合信号对于收益率的信息系数 IC 为：
$$
IC_\Phi = \operatorname{Corr}(\Phi, r)= \sqrt{\mathbf{IC}'\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\mathbf{IC}}
$$

### 单信号

最简单的情形是考察一只资产并且只有一个预测信号的情形，此时预测基本公式为：
$$
\Phi = \sigma\cdot IC\cdot z_g
$$

### 双信号

对于一只资产两个预测信号 $g_1,g_2$ 的情形，记两个信号的相关系数为 $\rho = \operatorname{Corr}(g_1,g_2)$，预测基本公式为：
$$
\Phi = \sigma\cdot\begin{pmatrix} IC_{g_1} & IC_{g_2} \end{pmatrix} \cdot \begin{pmatrix} 1 & \rho \\ \rho & 1 \end{pmatrix}^{-1} \cdot \begin{pmatrix} z_{g_1} \\ z_{g_2} \end{pmatrix}
$$

一种方法是直接计算信号相关系数矩阵的逆为：
$$
\begin{pmatrix} 1 & \rho \\ \rho & 1 \end{pmatrix}^{-1} = \frac{1}{1-\rho^2}\cdot \begin{pmatrix} 1 & -\rho \\ -\rho & 1 \end{pmatrix}
$$
调整 IC 为：
$$
\begin{aligned}
\begin{pmatrix} IC^{*}_{g_1} \\ IC^{*}_{g_2} \end{pmatrix} &= \frac{1}{1-\rho^2}\cdot \begin{pmatrix} 1 & -\rho \\ -\rho & 1 \end{pmatrix}\cdot\begin{pmatrix} IC_{g_1} \\ IC_{g_2} \end{pmatrix}\\
&= \begin{pmatrix} \frac{IC_{g_1} - \rho\cdot IC_{g_2}}{1-\rho^2} \\ \frac{IC_{g_2} - \rho\cdot IC_{g_1}}{1-\rho^2} \end{pmatrix}
\end{aligned}
$$
从而预测基本公式为：
$$
\Phi = \sigma\cdot\begin{pmatrix} IC^{*}_{g_1} & IC^{*}_{g_2} \end{pmatrix}\cdot\begin{pmatrix} z_{g_1} \\ z_{g_2} \end{pmatrix} = \sigma\cdot IC^*_{g_1}\cdot z_{g_1} + \sigma\cdot IC^*_{g_2}\cdot z_{g_2}
$$
如果两个信号不相关，则调整 IC 等于原始 IC，精练预测等于两个信号各自的精练预测的和：
$$
\Phi = \sigma\cdot IC_{g_1}\cdot z_{g_1} + \sigma\cdot IC_{g_2}\cdot z_{g_2}
$$

另外一种方法是将原始信号正交化，得到新的不相关信号来构建精练预测，即对信号相关系数矩阵进行特征分解：
$$
\begin{pmatrix} 1 & \rho \\ \rho & 1 \end{pmatrix} = \begin{pmatrix} \frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2} \\ \frac{\sqrt{2}}{2} & -\frac{\sqrt{2}}{2} \end{pmatrix} \cdot \begin{pmatrix} 1+\rho & 0 \\ 0 & 1-\rho \end{pmatrix} \cdot \begin{pmatrix} \frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2} \\ \frac{\sqrt{2}}{2} & -\frac{\sqrt{2}}{2} \end{pmatrix}
$$
信号的标准分值变换为正交信号：
$$
\begin{aligned}
\begin{pmatrix} z^*_{g_1} \\ z^*_{g_2} \end{pmatrix} &= \begin{pmatrix} \frac{1}{\sqrt{1+\rho}} & 0 \\ 0 & \frac{1}{\sqrt{1-\rho}} \end{pmatrix} \cdot \begin{pmatrix} \frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2} \\ \frac{\sqrt{2}}{2} & -\frac{\sqrt{2}}{2} \end{pmatrix} \cdot \begin{pmatrix} z_{g_1} \\ z_{g_2} \end{pmatrix}\\
&= \begin{pmatrix} \frac{z_{g_1}+z_{g_2}}{\sqrt{2(1+\rho)}} \\ \frac{z_{g_1}-z_{g_2}}{\sqrt{2(1-\rho)}} \end{pmatrix}
\end{aligned}
$$
正交信号的 IC 为：
$$
\begin{aligned}
\begin{pmatrix} IC^{**}_{g_1} \\ IC^{**}_{g_2} \end{pmatrix} &= \begin{pmatrix} \frac{1}{\sqrt{1+\rho}} & 0 \\ 0 & \frac{1}{\sqrt{1-\rho}} \end{pmatrix} \cdot \begin{pmatrix} \frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2} \\ \frac{\sqrt{2}}{2} & -\frac{\sqrt{2}}{2} \end{pmatrix} \cdot \begin{pmatrix} IC_{g_1} \\ IC_{g_2} \end{pmatrix}\\
&= \begin{pmatrix} \frac{IC_{g_1}+IC_{g_2}}{\sqrt{2(1+\rho)}} \\ \frac{IC_{g_1}-IC_{g_2}}{\sqrt{2(1-\rho)}} \end{pmatrix}
\end{aligned}
$$
信号是正交的，从而精练预测等于两个信号各自的精练预测的和：
$$
\Phi = \sigma\cdot IC^{**}_{g_1}\cdot z^*_{g_1} + \sigma\cdot IC^{**}_{g_2}\cdot z^*_{g_2}
$$

精练预测作为两个信号的组合信号对于收益率的信息系数 IC 为：
$$
IC_\Phi = \sqrt{\frac{IC^2_{g_1} + IC^2_{g_2} - 2\rho\cdot IC_{g_1}\cdot IC_{g_2}}{1-\rho^2}}
$$
如果两个信号不相关，则组合信号和单个信号满足平方可加性：
$$
IC^2_\Phi = IC^2_{g_1} + IC^2_{g_2}
$$

### 时间序列回归模型

通常，在时间序列上有 T 个资产收益率和信号的样本用于预测未来一期的资产收益率，记收益率的样本向量为 $\mathbf{r} = \left(r_1,\ldots,r_T\right)'$，其中，$r_t$ 是 $(t-1,t]$ 时段上的资产收益率，信号的样本矩阵为：
$$
\mathbf{G} = \begin{pmatrix} \mathbf{g}_1 & \cdots & \mathbf{g}_K \end{pmatrix} = \begin{pmatrix} g_{1,1} & \cdots & g_{1,K} \\ \vdots & \ddots & \vdots \\ g_{T,1} & \cdots & g_{T,K} \end{pmatrix}
$$
其中，$g_{t,k}$ 是第 k 个信号在 t-1 时刻（为了使得信号和收益率对应，将信号滞后一期）的样本值。

可以建立资产收益率对信号的回归模型来预测未来收益率：
$$
r_t = \alpha + \mathbf{g}'_t\boldsymbol{\beta} + u_t,\ 1\leq t\leq T
$$

将 $T\times 1$ 维全 1 向量 $\boldsymbol{\iota}$ 和 $\mathbf{G}$ 合并记为 $\tilde{\mathbf{G}} = \left(\boldsymbol{\iota}, \mathbf{G}\right)$，则回归系数的 OLS 估计为：
$$
\begin{aligned}
\begin{pmatrix} \hat{\alpha} \\ \hat{\boldsymbol{\beta}} \end{pmatrix} &= \left(\tilde{\mathbf{G}}'\tilde{\mathbf{G}}\right)^{-1}\tilde{\mathbf{G}}'\mathbf{r} = \left(\begin{pmatrix} \boldsymbol{\iota}' \\ \mathbf{G}' \end{pmatrix}\begin{pmatrix} \boldsymbol{\iota} & \mathbf{G} \end{pmatrix}\right)^{-1}\begin{pmatrix} \boldsymbol{\iota}'\mathbf{r} \\ \mathbf{G}'\mathbf{r} \end{pmatrix}\\
&= \begin{pmatrix} T & T\cdot\bar{\mathbf{g}}' \\ T\cdot\bar{\mathbf{g}} & \mathbf{G}'\mathbf{G} \end{pmatrix}^{-1} \begin{pmatrix} T\cdot\bar{r} \\ \mathbf{G}'\mathbf{r} \end{pmatrix}
\end{aligned}
$$
其中，$\bar{\mathbf{g}} = \frac{1}{T}\mathbf{G}'\boldsymbol{\iota}$ 是信号的样本均值向量，$\bar{r} = \frac{1}{T}\mathbf{r}'\boldsymbol{\iota}$ 是收益率的样本均值。

由[分块矩阵的逆](../矩阵运算.ipynb/#InvOfBlockMatrix)可得：
$$
\begin{aligned}
\hat{\boldsymbol{\beta}} &= \left(\mathbf{G}'\mathbf{G} - T\bar{\mathbf{g}}\bar{\mathbf{g}}\right)^{-1}\left(\mathbf{G}'\mathbf{r}-T\bar{\mathbf{g}}\bar{r}\right)\\
&= \left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')'(\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\right)^{-1}\cdot\left(\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}'\right)'\left(\mathbf{r} - \boldsymbol{\iota}\bar{r}\right)
\end{aligned}
$$

由回归的性质有：
$$
\hat{\alpha} = \bar{r} - \bar{\mathbf{g}}'\cdot\hat{\boldsymbol{\beta}}
$$

对于给定的新信号值 $\mathbf{g}^*$，根据回归模型的收益率预测为：
$$
\begin{aligned}
r^* &= \hat{\alpha} + \hat{\boldsymbol{\beta}}'\mathbf{g}^* = \bar{r} + \hat{\boldsymbol{\beta}}'\left(\mathbf{g}^*-\bar{\mathbf{g}}\right)\\
&= \bar{r} + \left(\mathbf{r} - \boldsymbol{\iota}\bar{r}\right)'\left(\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}'\right) \cdot \left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')'(\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\right)^{-1}\left(\mathbf{g}^*-\bar{\mathbf{g}}\right)
\end{aligned}
$$

由[样本方差](../概率统计.ipynb/#SampleCov)的定义可得：
$$
r^* = \bar{r} + \widehat{\operatorname{Cov}}(r,\mathbf{g}) \cdot \widehat{\operatorname{Var}}^{-1}(\mathbf{g}) \cdot \left(\mathbf{g}^*-\bar{\mathbf{g}}\right)
$$

如果用样本均值作为预测基本公式中期望的估计，样本方差协方差矩阵作为预测基本公式中协方差矩阵的估计，那么时间序列回归模型对收益率的预测和预测基本公式是一致的。

## 多资产情形

一般的多资产多信号预测公式过于复杂，需要估计的参数非常多，在实践中难以直接使用，需要引入额外的假设或者结构来简化模型。

通常将信号划分为不同的信号源，记信号源的数量为 J，信号源 j 将产生信号 $\mathbf{g}_j=\left(g_{1,j},\ldots,g_{N,j}\right)'$，$g_{n,j}$ 表示信号源 j 对资产 n 产生的一个信号，所有信号源产生的信号为
$$
\mathbf{g} = \begin{pmatrix} \mathbf{g}_1 \\ \vdots \\ \mathbf{g}_J \end{pmatrix}
$$
信号总数为 $K = J \cdot N$。

通常的简化假设为：
* 不同信号源之间的相关性由相关系数矩阵 $\boldsymbol{\rho}_g$ 描述，即不同信号源对同一只资产产生的信号之间的相关系数为：
$$
\operatorname{Corr}(g_{n,j_1}, g_{n,j_2}) = \rho_{g,j_1,j_2},\ \forall n,j_1,j_2
$$
信号源 j 产生的信号之间的相关性由相关系数矩阵 $\boldsymbol{\rho}_j$ 描述，即信号源 j 对不同资产产生的信号之间的相关系数为：
$$
\operatorname{Corr}(g_{m,j}, g_{n,j}) = \rho_{j,m,n},\ \forall m,n,j
$$
通常进一步假设所有信号源的相关系数矩阵均相等：
$$
\boldsymbol{\rho}_j = \boldsymbol{\rho},\ j=1,\ldots,J
$$
在上述假设下，所有信号的相关系数矩阵为：
$$
\operatorname{Corr}(\mathbf{g}) = \boldsymbol{\rho}_g\otimes\boldsymbol{\rho}
$$
其中，$\otimes$ 是[克罗内克积](../矩阵运算.ipynb/#KroneckerProduct)。
* 信号源 j 产生的信号对其对应资产的预测能力均相同：
$$
\operatorname{Corr}(g_{n,j},r_n) = IC_j,\ \forall n, j
$$
信号源 j 对资产 m 产生的信号对资产 n 预测能力由相关系数矩阵 $\boldsymbol{\rho}$ 调整：
$$
\operatorname{Corr}(g_{m,j},r_n) = IC_j\cdot\rho_{m,n},\ \forall m\neq n, j
$$
记 $\mathbf{IC} = \left(IC_1,\ldots,IC_J\right)'$，则所有信号和资产的相关系数矩阵为：
$$
\operatorname{Corr}(\mathbf{g},\mathbf{r}) = \mathbf{IC}\otimes\boldsymbol{\rho}
$$

记 $\mathbf{G} = \left(\mathbf{g}_1,\ldots,\mathbf{g}_J\right)$，即 $\mathbf{g} = \operatorname{vec}(\mathbf{G})$，其中 $\operatorname{vec}$ 是向量化算子，它返回的是把矩阵的所有列堆起来所形成的列向量。再记 $\mathbf{Z}_G = \left(\mathbf{z}_{g_1},\ldots,\mathbf{z}_{g_J}\right)$，即 $\mathbf{z}_g = \operatorname{vec}(\mathbf{Z}_G)$。

在上述假设下，由[克罗内克积的性质](../矩阵运算.ipynb/#KroneckerProduct)可得预测基本公式为：
$$
\begin{aligned}
\Phi &= \operatorname{Std}(\mathbf{r})\cdot\operatorname{Corr}(\mathbf{r},\mathbf{g})\cdot\operatorname{Corr}^{-1}(\mathbf{g})\cdot\operatorname{Std}^{-1}(\mathbf{g})\cdot\left(\mathbf{g} - \mathbb{E}(\mathbf{g})\right)\\
&= \operatorname{Std}(\mathbf{r}) \cdot \left(\mathbf{IC}\otimes\boldsymbol{\rho}\right)' \cdot \left(\boldsymbol{\rho}_g\otimes\boldsymbol{\rho}\right)^{-1} \cdot \mathbf{z}_g\\
&= \operatorname{Std}(\mathbf{r}) \cdot \left[\left(\mathbf{IC}'\boldsymbol{\rho}^{-1}_g\right) \otimes \mathbf{I}\right] \cdot \mathbf{z}_g\\
&= \operatorname{Std}(\mathbf{r}) \cdot \mathbf{Z}_G \cdot \boldsymbol{\rho}^{-1}_g \cdot \mathbf{IC}
\end{aligned}
$$

最后的基本预测公式与 $\boldsymbol{\rho}$ 无关。可以将上式理解为先对信号进行标准化，然后以信号调整的 IC $\boldsymbol{\rho}^{-1}_g \cdot \mathbf{IC}$ 进行加权得到组合信号，最后用资产的波动率进行放缩得到预期收益率。

### 单信号源

假设只有一个信号源，信号总数 $K=N$，设信号源对所有资产产生的信号的相关系数矩阵为 $\operatorname{Corr}(\mathbf{g})=\boldsymbol{\rho}$。

上述假设变为：
1. 信号源对资产 n 产生的信号 $g_n$ 对资产 n 的信息系数关于所有资产都相等，记为 $IC$，即：
$$
\operatorname{Corr}(g_n, r_n) = IC
$$
2. 信号源对资产 m 产生的信号 $g_m$ 对资产 n 的预测能力仅与信号 $g_m$ 和 $g_n$ 的相关性有关，即：
$$
\operatorname{Corr}(g_m, r_n) = IC\cdot \rho_{m,n}
$$

预测基本公式为：
$$
\Phi = \operatorname{Std}(\mathbf{r})\cdot IC \cdot \mathbf{z}_g
$$


### 双信号源

假设有两个信号源 1、2，$\mathbf{g}=\begin{pmatrix} \mathbf{g}'_1 & \mathbf{g}'_2 \end{pmatrix}'$，信号总数 $K=2N$。

上述假设具体为：
1. 信号源 1(2) 对资产 n 产生的信号 $g_{n,1}$（$g_{n,2}$）对资产 n 的信息系数关于所有资产都相等，记为 $IC_1$（$IC_2$），即：
$$
\operatorname{Corr}(g_{n,1}, r_n) = IC_1,\ \operatorname{Corr}(g_{n,2}, r_n) = IC_2
$$
2. 信号源 1 产生的信号的相关系数矩阵和信号源 2 的相等，即：
$$
\operatorname{Corr}(\mathbf{g}_1)=\operatorname{Corr}(\mathbf{g}_2)=\boldsymbol{\rho}
$$
3. 信号源 1(2) 对资产 m 产生的信号 $g_{m,1}$（$g_{m,2}$）对资产 n 的预测能力仅与信号 $g_{m,1}$（$g_{m,2}$ ）和 $g_{n,1}$（$g_{n,2}$）的相关性有关，即：
$$
\operatorname{Corr}(g_{m,1}, r_n) = IC_1\cdot \rho_{m,n},\ \operatorname{Corr}(g_{m,2}, r_n) = IC_2\cdot \rho_{m,n}
$$
4. 两个信号源的相关性用 $\rho_{1,2}$ 表示，其产生的信号之间的相关系数为：
$$
\operatorname{Corr}(\mathbf{g}_1,\mathbf{g}_2) = \rho_{1,2}\cdot\boldsymbol{\rho}
$$

上面的假设赋予信号相关系数矩阵的结构为：
$$
\operatorname{Corr}(\mathbf{g}) = \begin{pmatrix} \boldsymbol{\rho} & \rho_{1,2}\cdot\boldsymbol{\rho} \\ \rho_{1,2}\cdot\boldsymbol{\rho} & \boldsymbol{\rho} \end{pmatrix}
$$

赋予信号对资产的信息系数矩阵的结构为：
$$
\mathbf{IC} = \operatorname{Corr}(\mathbf{g},\mathbf{r}) = \begin{pmatrix} IC_1\cdot\boldsymbol{\rho} \\ IC_2\cdot\boldsymbol{\rho} \end{pmatrix}
$$

从而，预测基本公式变为：
$$
\Phi = \operatorname{Std}(\mathbf{r}) \cdot \begin{pmatrix} \frac{IC_1-\rho_{1,2}\cdot IC_2}{1-\rho^2_{1,2}}\cdot\mathbf{I} & \frac{IC_2-\rho_{1,2}\cdot IC_1}{1-\rho^2_{1,2}}\cdot\mathbf{I} \end{pmatrix} \cdot \mathbf{z}_g
$$

### 横截面回归模型

假设有某个时刻的横截面样本，记收益率的样本向量为 $\mathbf{r} = \left(r_1,\ldots,r_N\right)'$，其中，$r_n$ 是资产 n 收益率，信号的样本矩阵为：
$$
\mathbf{G} = \begin{pmatrix} \mathbf{g}_1 & \cdots & \mathbf{g}_J \end{pmatrix} = \begin{pmatrix} g_{1,1} & \cdots & g_{1,J} \\ \vdots & \ddots & \vdots \\ g_{N,1} & \cdots & g_{N,J} \end{pmatrix}
$$
其中，$g_{n,j}$ 是信号源 j 产生的对资产 n 的期初预测信号。

可以建立资产收益率对信号的回归模型：
$$
r_n = \alpha + \mathbf{g}'_n\boldsymbol{\beta} + u_n,\ 1\leq n\leq N
$$

将 $N\times 1$ 维全 1 向量 $\boldsymbol{\iota}$ 和 $\mathbf{G}$ 合并记为 $\tilde{\mathbf{G}} = \left(\boldsymbol{\iota}, \mathbf{G}\right)$，则回归系数的 OLS 估计为：
$$
\begin{aligned}
\begin{pmatrix} \hat{\alpha} \\ \hat{\boldsymbol{\beta}} \end{pmatrix} &= \left(\tilde{\mathbf{G}}'\tilde{\mathbf{G}}\right)^{-1}\tilde{\mathbf{G}}'\mathbf{r} = \left(\begin{pmatrix} \boldsymbol{\iota}' \\ \mathbf{G}' \end{pmatrix}\begin{pmatrix} \boldsymbol{\iota} & \mathbf{G} \end{pmatrix}\right)^{-1}\begin{pmatrix} \boldsymbol{\iota}'\mathbf{r} \\ \mathbf{G}'\mathbf{r} \end{pmatrix}\\
&= \begin{pmatrix} N & N\cdot\bar{\mathbf{g}}' \\ N\cdot\bar{\mathbf{g}} & \mathbf{G}'\mathbf{G} \end{pmatrix}^{-1} \begin{pmatrix} N\cdot\bar{r} \\ \mathbf{G}'\mathbf{r} \end{pmatrix}
\end{aligned}
$$
其中，$\bar{\mathbf{g}} = \frac{1}{N}\mathbf{G}'\boldsymbol{\iota}$ 是信号的横截面样本均值向量，$\bar{r} = \frac{1}{N}\mathbf{r}'\boldsymbol{\iota}$ 是资产收益率的横截面样本均值。

由[分块矩阵的逆](../矩阵运算.ipynb/#InvOfBlockMatrix)可得：
$$
\begin{aligned}
\hat{\boldsymbol{\beta}} &= \left(\mathbf{G}'\mathbf{G} - T\bar{\mathbf{g}}\bar{\mathbf{g}}\right)^{-1}\left(\mathbf{G}'\mathbf{r}-T\bar{\mathbf{g}}\bar{r}\right)\\
&= \left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')'(\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\right)^{-1}\cdot\left(\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}'\right)'\left(\mathbf{r} - \boldsymbol{\iota}\bar{r}\right)
\end{aligned}
$$

由回归的性质有：
$$
\hat{\alpha} = \bar{r} - \bar{\mathbf{g}}'\cdot\hat{\boldsymbol{\beta}}
$$

对于给定的新的横截面信号值 $\mathbf{G}^*$，根据回归模型的收益率预测为：
$$
\begin{aligned}
\mathbf{r}^* &= \hat{\alpha}\boldsymbol{\iota} + \mathbf{G}^*\hat{\boldsymbol{\beta}} = \bar{r}\boldsymbol{\iota} + \left(\mathbf{G}^*-\boldsymbol{\iota}\bar{\mathbf{g}}'\right)\hat{\boldsymbol{\beta}}\\
&= \bar{r}\boldsymbol{\iota} + \left(\mathbf{G}^* - \boldsymbol{\iota}\bar{\mathbf{g}}'\right) \cdot \left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')'(\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\right)^{-1} \cdot \left(\mathbf{G}-\boldsymbol{\iota}\bar{\mathbf{g}}'\right)\left(\mathbf{r} - \boldsymbol{\iota}\bar{r}\right)\\
&= \bar{r}\boldsymbol{\iota} + \hat{\sigma}_r\mathbf{I} \cdot \left[\left(\mathbf{G}^* - \boldsymbol{\iota}\bar{\mathbf{g}}'\right)\hat{\boldsymbol{\sigma}}^{-1}_g\right] \cdot \left[\frac{1}{N-1}\left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\hat{\boldsymbol{\sigma}}^{-1}_g\right)'\left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\hat{\boldsymbol{\sigma}}^{-1}_g\right)\right]^{-1} \cdot \left[\frac{1}{N-1}\left((\mathbf{G}-\boldsymbol{\iota}\bar{\mathbf{g}}')\hat{\boldsymbol{\sigma}}^{-1}_g\right)'\left((\mathbf{r} - \boldsymbol{\iota}\bar{r})(\hat{\sigma}_r\mathbf{I})^{-1}\right)\right]
\end{aligned}
$$
其中，$\boldsymbol{\iota}$ 是 $N\times 1$ 维全 1 向量，$\hat{\sigma}_r$ 是资产收益率的横截面样本标准差，$\hat{\boldsymbol{\sigma}}_g$ 是信号的横截面样本标准差对角阵。

如果：
* 用资产收益率的横截面样本均值 $\bar{r}\boldsymbol{\iota}$ 作为 $\mathbb{E}(\mathbf{r})$ 的估计（所有资产的无条件预期收益率相同）；
* 用资产收益率的横截面样本标准差 $\sigma_r\mathbf{I}$ 作为 $\operatorname{Std}(\mathbf{r})$ 的估计（所有资产的无条件标准差相同）；
* 用信号源横截面信号的样本均值 $\operatorname{vec}(\boldsymbol{\iota}\bar{\mathbf{g}}')$ 作为信号期望 $\mathbb{E}(\mathbf{g})$ 的估计；
* 用信号源横截面信号的样本标准差 $\hat{\boldsymbol{\sigma}}_g\otimes\mathbf{I}$ 作为信号标准差 $\operatorname{Std}(\mathbf{g})$ 的估计；
* 用信号源横截面信号的样本相关系数矩阵 $\frac{1}{N-1}\left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\hat{\boldsymbol{\sigma}}^{-1}_g\right)'\left((\mathbf{G} - \boldsymbol{\iota}\bar{\mathbf{g}}')\hat{\boldsymbol{\sigma}}^{-1}_g\right)$ 作为信号源相关系数矩阵 $\boldsymbol{\rho}_g$ 的估计；
* 用信号源横截面信号和资产横截面收益率的样本相关系数矩阵 $\frac{1}{N-1}\left((\mathbf{G}-\boldsymbol{\iota}\bar{\mathbf{g}}')\hat{\boldsymbol{\sigma}}^{-1}_g\right)'\left((\mathbf{r} - \boldsymbol{\iota}\bar{r})(\hat{\sigma}_r\mathbf{I})^{-1}\right)$ 作为 $\mathbf{IC}$ 的估计；

那么横截面回归模型对收益率的预测和预测基本公式是一致的。

# 信息时间尺度（Information Horizon）

## 策略的时间尺度

记号：
* $\mathbf{h}(j)$ 策略滞后 j 期的主动组合；
* $\theta(j)$ 滞后 j 期的主动组合收益率，即 $\theta(j)=\mathbf{r}'\mathbf{h}(j)$；
* $\alpha(j)$ 滞后 j 期的主动组合的 alpha，即 $\alpha(j)=\boldsymbol{\alpha}'\mathbf{h}(j)$；
* $\omega(j)$ 滞后 j 期的主动组合的残差波动率，即 $\omega^2(j)=\mathbf{h}'(j)\mathbf{V}\mathbf{h}(j)$；
* $IR(j)$ 滞后 j 期的主动组合的信息率，即 $IR(j)=\frac{\alpha(j)}{\omega(j)}$；

<a id='InformationHorizonAssumption'></a>
基本假设：
1. 假设原策略组合的 beta 为 1，则其主动组合和残差组合相同；
2. 任意滞后期的主动组合具有同样的主动风险，即：
$$
\omega^2(j) = \mathbf{h}'(j)\mathbf{V}\mathbf{h}(j) = \omega^2
$$
从而，滞后组合信息率的衰减仅由 alpha 的衰减决定。
3. 任意两个滞后期主动组合的协方差或相关系数仅依赖于两期的时间间隔，即：
$$
\operatorname{Cov}(\theta(j),\theta(k)) = \mathbf{h}'(j)\mathbf{V}\mathbf{h}(k) = \omega^2\cdot \rho(|j-k|)
$$

假设滞后期 j 的主动组合的权重为 $y(j),j=0,1,2,\ldots$，按照此权重形成混合策略，记混合策略的主动组合为 $\mathbf{h}_m$：
$$
\mathbf{h}_m(0) = \sum\limits_{k\geq0}y(k)\mathbf{h}(k)
$$
混合策略滞后 j 期的主动组合为：
$$
\mathbf{h}_m(j) = \sum\limits_{k\geq0}y(k)\mathbf{h}(j+k)
$$

在上述基本假设下，混合策略有如下性质：
1. 混合策略的主动风险与滞后期无关，即其任意滞后期的主动组合具有同样的主动风险：
$$
\omega^2_m(j) = \mathbf{h}'_m(j)\mathbf{V}\mathbf{h}_m(j) = \omega_m^2
$$
2. 任意混合策略的相关性结构与原始策略的相关性结构相同，即任意两个滞后期主动组合的协方差或相关系数仅依赖于两期的时间间隔：
$$
\operatorname{Cov}(\theta_m(j),\theta_m(k)) = \mathbf{h}'_m(j)\mathbf{V}\mathbf{h}_m(j) = \omega_m^2\rho_m(|j-k|)
$$
3. 如果原策略的信息率以指数形式衰减，即：
$$
IR(j) = \gamma^j\cdot IR(0)
$$
其中，$\gamma$ 是衰减速率，那么其混合策略的信息率 $IR^*$ 也会以指数形式衰减且衰减速率相同，即：
$$
IR_m(j) = \gamma^j\cdot IR_m(0)
$$

**证明**

下证性质 1。对于混合策略，滞后 j 期的残差收益率方差为：
$$
\begin{aligned}
\operatorname{Var}(\theta_m(j)) &= \operatorname{Var}\left(\sum\limits_{k\geq0}y(k)\cdot\theta(j+k)\right)\\
&= \sum\limits_{k\geq0}\sum\limits_{l\geq0}y(k)\cdot\operatorname{Cov}\left(\theta(j+k),\theta(j+l)\right)\cdot y(l)\\
&= \sum\limits_{k\geq0}\sum\limits_{l\geq0}y(k)\cdot\operatorname{Cov}\left(\theta(k),\theta(l)\right)\cdot y(l)\ \ (\text{由假设 3})\\
&= \operatorname{Var}(\theta_m(0))\\
&= \omega^2_m
\end{aligned}
$$

下证性质 2。对于混合策略，滞后 j 期和滞后 k 期的残差收益率协方差为：
$$
\begin{aligned}
\operatorname{Cov}\left(\theta_m(j),\theta_m(k)\right) &= \operatorname{Cov}\left(\sum\limits_{m\geq0}y(m)\cdot\theta(j+m),\sum\limits_{n\geq0}y(n)\cdot\theta(k+n)\right)\\
&= \sum\limits_{m\geq0}\sum\limits_{n\geq0}y(m)\cdot\operatorname{Cov}\left(\theta(j+m),\theta(k+n)\right)\cdot y(n)\\
&= \sum\limits_{m\geq0}\sum\limits_{n\geq0}y(m)\cdot\operatorname{Cov}\left(\theta(j-k+m),\theta(n)\right)\cdot y(n) = \operatorname{Cov}(\theta_m(j-k), \theta_m(0))\ \ (\text{由假设 3})\\
&= \sum\limits_{m\geq0}\sum\limits_{n\geq0}y(m)\cdot\operatorname{Cov}\left(\theta(m),\theta(k-j+n)\right)\cdot y(n) = \operatorname{Cov}(\theta_m(0), \theta_m(k-j))\ \ (\text{由假设 3})\\
&= \omega^2_m\cdot\rho_m(|j-k|)
\end{aligned}
$$

下证性质 3。对于混合策略，滞后 j 期的信息率为：
$$
IR_m(j) = \frac{\alpha_m(j)}{\omega_m} = \frac{\sum\limits_{k\geq0}y(k)\cdot\alpha(j+k)}{\omega_m} = \frac{\sum\limits_{k\geq0}y(k)\cdot IR(j+k)\cdot\omega}{\omega_m}
$$
则：
$$
\frac{IR_m(j)}{IR_m(0)} = \frac{\sum\limits_{k\geq0}y(k)\cdot IR(j+k)}{\sum\limits_{k\geq0}y(k)\cdot IR(k)} = \frac{\sum\limits_{k\geq0}y(k)\cdot IR(0)\cdot\gamma^{j+k}}{\sum\limits_{k\geq0}y(k)\cdot IR(0)\cdot\gamma^{k}} = \gamma^{j}
$$

### 两期最优混合策略

只考虑策略当期组合和滞后一期组合的最优混合问题。记当前组合的权重 $w$，滞后一期组合的权重为 $1-w$，混合策略的主动组合为：
$$
\mathbf{h}_m(0) = w\cdot\mathbf{h}(0) + (1-w)\cdot\mathbf{h}(1)
$$
记当前组合和滞后组合的相关系数为 $\rho$。

在[基本假设](#InformationHorizonAssumption)下，再假设原策略的信息率以指数形式衰减 $IR(j) = \gamma^j\cdot IR(0)$。

混合策略的 alpha 为：
$$
\alpha_m = w\cdot\alpha(0) + (1-w)\cdot\alpha(1) = \omega\cdot IR(0)\cdot\left[w+(1-w)\gamma\right]
$$
混合策略的残差风险为：
$$
\omega^2_m = \omega^2\cdot\left[w^2 + (1-w)^2 + 2\rho w(1-w)\right]
$$
从而，混合策略的信息率为：
$$
IR^2_m(0) = \frac{\left[w+(1-w)\gamma\right]^2}{w^2 + (1-w)^2 + 2\rho w(1-w)}\cdot IR^2(0)
$$
最大化混合策略的信息率可得：
$$
w = \frac{1-\rho\gamma}{(1-\rho)(1-\gamma)}
$$
最优混合策略的信息率为：
$$
IR^*_m(0) = IR(0)\cdot\frac{(1-\rho\gamma)+\gamma(\gamma-\rho)}{\sqrt{(1-\rho\gamma)^2 + (\gamma-\rho)^2 + 2\rho(1-\rho\gamma)(\gamma-\rho)}}
$$

### 多期最优混合策略

考虑策略当前组合和所有滞后组合的最优混合问题。记滞后期 j 的主动组合的权重为 $y(j),j=0,1,2,\ldots$，则混合策略的主动组合为：
$$
\mathbf{h}_m(0) = \sum\limits_{k\geq0}y(k)\mathbf{h}(k)
$$

在[基本假设](#InformationHorizonAssumption)下，该问题即是求解最大化信息率组合的问题，这等价于固定 alpha 约束下最小化残差风险的优化问题，这里固定 alpha 为原策略的当期 alpha $\alpha(0)$：
$$
\begin{aligned}
\mathbf{h}_m(0) = \ & \mathop{\arg\min}_{\mathbf{y}}\operatorname{Var}\left(\sum\limits_{j=0}^\infty y(j)\cdot\theta(j)\right)\\
& \rm{s.t.}\ \sum\limits_{j=0}^\infty y(j)\alpha(j) = \alpha(0)
\end{aligned}
$$
以信息率表示为：
$$
\begin{aligned}
\mathbf{h}_m(0) = \ & \mathop{\arg\min}_{\mathbf{y}}\operatorname{Var}\left(\sum\limits_{j=0}^\infty y(j)\cdot\theta(j)\right)\\
& \rm{s.t.}\ \sum\limits_{j=0}^\infty y(j)\cdot IR(j) = IR(0)
\end{aligned}
$$
记最优混合策略滞后 j 期的主动组合为 $\mathbf{h}^*(0)$，残差收益率为 $\theta^*(j)$，残差风险为 $\omega^*$（由混合策略的性质 1 可知残差风险和滞后期无关），信息率为 $IR*(j)$，则最优混合策略有如下性质：
1. 最优混合策略滞后组合与当期组合的相关系数随着滞后期的增加而衰减，衰减速率等于信息率的衰减速率，即：
$$
\operatorname{Corr}(\theta^*(k),\theta^*(0)) = \frac{IR^*(k)}{IR^*(0)}
$$
2. 假设原策略的任意两滞后期主动组合的协方差和相关系数的形式为：
$$
\operatorname{Cov}(\theta(j),\theta(k)) = \omega^2\cdot\rho^{|j-k|},\ \operatorname{Corr}(\theta(j),\theta(k))\rho^{|j-k|}
$$
再假设原策略的信息率以指数形式衰减 $IR(j) = \gamma^j\cdot IR(0)$，那么最优混合策略的主动组合可表示为：
$$
\begin{aligned}
\mathbf{h}^*(0) &= \frac{1-\gamma^2}{1-\rho\gamma}\sum\limits_{k\geq0}\gamma^k\left[\mathbf{h}(k) - \rho\cdot\mathbf{h}(k+1)\right]\\
&= \frac{1-\gamma^2}{1-\rho\gamma}\left[\mathbf{h}(0) - \rho\cdot\mathbf{h}(1)\right] + \gamma\cdot\mathbf{h}^*(1)
\end{aligned}
$$

**证明：**
下证性质 1。优化问题的 Lagrange 函数为：
$$
L(\mathbf{y}, \lambda) = \operatorname{Var}\left(\sum\limits_{j=0}^{\infty}y(j)\cdot\theta(j)\right) + \lambda\left(\sum\limits_{j=0}^\infty y(j)\cdot IR(j) - IR(0)\right)
$$
最优解的一阶条件为：
$$
2\operatorname{Cov}\left(\theta(j), \sum\limits_{k=0}^{\infty}y(k)\cdot\theta(k)\right) + \lambda\cdot IR(j) = 0,\ j=0,1,2,\ldots
$$
将上式两边同乘以 $y(j)$ 并对所有的等式求和可得：
$$
\lambda = -\frac{2(\omega^*)^2}{IR(0)}
$$
从而有：
$$
\operatorname{Cov}(\theta(j),\theta^*(0)) = \frac{(\omega^*)^2\cdot IR(j)}{IR(0)}
$$
由
$$
IR(0) = \frac{\alpha(0)}{\omega} = \frac{\alpha^*(0)}{\omega} = \frac{\omega^*}{\omega}IR^*(0)
$$
得：
$$
\operatorname{Cov}(\theta(j),\theta^*(0)) = \frac{\omega\cdot\omega^*\cdot IR(j)}{IR^*(0)}
$$
再由最优混合策略滞后 k 期的残差收益率的定义：
$$
\theta^*(k) = \sum\limits_{j=0}^{\infty}y(j)\theta(j+k)
$$
得：
$$
\operatorname{Cov}(\theta^*(k),\theta^*(0)) = \frac{(\omega^*)^2\cdot IR^*(k)}{IR^*(0)}
$$
从而最优混合策略滞后组合与当前组合的相关系数为：
$$
\operatorname{Corr}(\theta^*(k),\theta^*(0)) = \frac{IR^*(k)}{IR^*(0)}
$$

下证性质 2。如果形式的记原策略残差收益率的相关系数矩阵为：
$$
\boldsymbol{\rho}_h = \begin{pmatrix} 1 & \rho & \rho^2 & \cdots & \rho^j & \cdots \\ \rho & 1 & \rho & \cdots & \rho^{j-1} & \cdots \\ \rho^2 & \rho & 1 & \cdots & \rho^{j-2} & \cdots \\ \vdots & \vdots & \vdots & \ddots & \vdots & \cdots \\ \rho^{j} & \rho^{j-1} & \rho^{j-2} & \cdots & 1 & \cdots \\ \vdots & \vdots & \vdots & \vdots & \vdots & \ddots \end{pmatrix}
$$
这是一个无限的 Toeplitz 矩阵，参考类似的[有限 Toeplitz 矩阵的逆](../矩阵运算/#SpecialToeplitzMatrix)可验证其逆为：
$$
\boldsymbol{\rho}_h^{-1} = \frac{1}{1-\rho^2}\begin{pmatrix} 1 & -\rho & 0 & \cdots \\ -\rho & 1+\rho^2 & -\rho & \cdots \\ 0 & -\rho & 1+\rho^2 & \cdots \\ \vdots & \vdots & \vdots & \ddots \end{pmatrix}
$$
记权重向量和信息率向量为：
$$
\mathbf{y} = \begin{pmatrix} y(0) \\ y(1) \\ y(2) \\ \vdots \end{pmatrix},\ \mathbf{IR} = IR(0)\cdot\boldsymbol{\gamma} = IR(0)\cdot\begin{pmatrix} 1 \\ \gamma \\ \gamma^2 \\ \vdots \end{pmatrix}
$$
则原问题等价为：
$$
\begin{aligned}
\mathbf{h}_m(0) = \ & \mathop{\arg\min}_{\mathbf{y}}\mathbf{y}'\boldsymbol{\rho}_h\mathbf{y}\\
& \rm{s.t.}\ \mathbf{y}'\boldsymbol{\gamma} = 1
\end{aligned}
$$
形式上可由[等式约束条件下的二次规划](..\最优化.ipynb\#QuandprogWithEqConstraint)可得：
$$
\mathbf{y} = \frac{\boldsymbol{\rho}_h^{-1}\boldsymbol{\gamma}}{\boldsymbol{\gamma}'\boldsymbol{\rho}_h^{-1}\boldsymbol{\gamma}} = \frac{1-\gamma^2}{1-\rho\gamma}\begin{pmatrix} 1 \\ \gamma-\rho \\ \gamma(\gamma-\rho) \\ \gamma^2(\gamma-\rho) \\ \vdots \end{pmatrix} = \frac{1-\gamma^2}{1-\rho\gamma}\left(\boldsymbol{\gamma} - \rho\cdot\mathbf{S}\cdot\boldsymbol{\gamma}\right)
$$
其中：
$$
\mathbf{S} = \begin{pmatrix} 0 & 0 & 0 & \cdots \\ 1 & 0 & 0 & \cdots \\ 0 & 1 & 0 & \cdots \\ \vdots & \vdots & \vdots & \ddots \end{pmatrix}
$$
$\mathbf{S}$ 左乘一个矩阵的作用是将该矩阵的所有行向下平移一行，首行填充 0，右乘矩阵的作用是将该矩阵的所有列向左平移一列。
记：
$$
\mathbf{H} = \begin{pmatrix}\mathbf{h}(0) & \mathbf{h}(1) & \cdots & \mathbf{h}(j) & \cdots \end{pmatrix}
$$
则最优混合策略的主动组合可以表示为：
$$
\begin{aligned}
\mathbf{h}^*(0) &= \mathbf{H}\cdot\mathbf{y}\\
&= \frac{1-\gamma^2}{1-\rho\gamma}\left[\mathbf{H}\boldsymbol{\gamma} - \rho\cdot(\mathbf{H}\cdot\mathbf{S})\cdot\boldsymbol{\gamma}\right]\\
&= \frac{1-\gamma^2}{1-\rho\gamma}\sum\limits_{k\geq0}\gamma^{k}\left[\mathbf{h}(k) - \rho\cdot\mathbf{h}(k+1)\right]\\
&= \frac{1-\gamma^2}{1-\rho\gamma}\left[\mathbf{h}(0) - \rho\cdot\mathbf{h}(1)\right] + \frac{1-\gamma^2}{1-\rho\gamma}\gamma\sum\limits_{k\geq1}\gamma^{k-1}\left[\mathbf{h}(k) - \rho\cdot\mathbf{h}(k+1)\right]\\
&= \frac{1-\gamma^2}{1-\rho\gamma}\left[\mathbf{h}(0) - \rho\cdot\mathbf{h}(1)\right] + \gamma\cdot\frac{1-\gamma^2}{1-\rho\gamma}\sum\limits_{l\geq0}\gamma^{l}\left[\mathbf{h}(l+1) - \rho\cdot\mathbf{h}(l+2)\right]\ (l=k-1)\\
&= \frac{1-\gamma^2}{1-\rho\gamma}\left[\mathbf{h}(0) - \rho\cdot\mathbf{h}(1)\right] + \gamma\cdot\mathbf{h}^*(1)
\end{aligned}
$$

## 信号的时间尺度

只考虑单个资产和单个信号的情形。

记号和基本假设：
* 信号以时间间隔 $\Delta t$ 产生；
* $s(t)$ 时刻 t 信号的标准 z 分值；
* $r(0,t)$ 表示区间 $(0,t]$ 上的资产收益率；
* 资产收益率单位时间的波动率为 $\omega$，即 $(0,t]$ 上的波动率为 $\omega\cdot\sqrt{t}$。

### 两期最优混合信号

记当期信号对随后一个时段收益率的信息系数为 $IC_1 = \operatorname{Corr}(s(0), r(0,\Delta t))$，当期信号对随后第二个时段收益率的信息系数为 $IC_2 = \operatorname{Corr}(s(0), r(\Delta t,2\Delta t))$，则信号对随后两个时期的总收益率的信息系数为：
$$
\begin{aligned}
IC &= \frac{\operatorname{Cov}(s(0), r(0,\Delta t)+r(\Delta t,2\Delta t))}{\operatorname{Std}(r(0,\Delta t)+r(\Delta t,2\Delta t))}\\
&= \frac{\omega\cdot\sqrt{\Delta t}\cdot\left(IC_1 + IC_2\right)}{\omega\cdot\sqrt{2\Delta t}}\\
&= \frac{IC_1+IC_2}{\sqrt{2}}
\end{aligned}
$$

下面考虑当期信号 $s(0)$ 和上一期信号 $s(-\Delta t)$ 的最优混合问题。记当期信号对随后一个时段收益率的信息系数为 $IC_1 = \operatorname{Corr}(s(0), r(0,\Delta t))$，上期信号对同一时段收益率的信息系数为 $IC_2 = \operatorname{Corr}(s(-\Delta t), r(0,\Delta t))$，两期信号的相关系数为 $\rho$。只需将滞后信号作为一个新的信号和当期信号按照预测基本公式进行叠加，结果类似于[单资产双信号情形下的预测基本公式](#SingleAssetTwoSignal)，混合信号即为精练预测：
$$
\begin{aligned}
s^*(0) &= \omega\cdot\sqrt{\Delta t}\cdot\begin{pmatrix} IC_1 & IC_2 \end{pmatrix}\cdot\begin{pmatrix} 1 & \rho \\ \rho & 1 \end{pmatrix}^{-1}\cdot\begin{pmatrix} s(0) \\ s(-\Delta t) \end{pmatrix}\\
&= \omega\cdot\sqrt{\Delta t}\cdot\left(\frac{IC_1-\rho IC_2}{1-\rho^2}s(0) + \frac{IC_2-\rho IC_1}{1-\rho^2}s(-\Delta t)\right)
\end{aligned}
$$
混合信号的信息系数为：
$$
IC^* = \sqrt{\frac{IC_1^2 + IC_2^2 - 2\rho\cdot IC_1\cdot IC_2}{1-\rho^2}}
$$

滞后信号 $s(-\Delta t)$ 对提高预测质量有两方面的作用：
* 分散化，作为收益率的第二个预测信号，当 $\rho<\frac{IC_2}{IC_1}$ 时，两期信号的相关性较低，滞后信号有相对较强的补充作用，分散化作为主导作用；
* 对冲，减少当期信号中噪声的一种方法，当 $\rho>\frac{IC_2}{IC_1}$ 时，两期信号的相关性较高，滞后信号的补充作用较弱，但可以用于对冲当期信号中的噪声，多冲将作为主导作用；

当 $\rho=\frac{IC_2}{IC_1}$ 时，滞后信号对提高预测质量没有贡献，不需要混合信号。

### 多期最优混合信号

考虑信号对时段 $(0, t]$ 收益率的预测能力。假设当期信号 $s(0)$ 对时间区间 $(0, \Delta t]$ 上资产收益率的信息系数是 IC，信号对未来每个 $\Delta t$ 时段上收益率的信息系数按照因子 $\delta$ 指数衰减，即：
$$
\operatorname{Corr}\left(s(0), r(j\Delta t, (j+1)\Delta t)\right) = IC\cdot\delta^{j},\ j=0,1,\ldots,\frac{t}{\Delta t}-1
$$
则信号对整个时段 $(0, t]$ 收益率的信息系数为：
$$
\begin{aligned}
\operatorname{Corr}\left(s(0), r(0, t)\right) &= \frac{\operatorname{Cov}\left(s(0), r(0, t)\right)}{\operatorname{Std}\left(r(0,t)\right)}\\
&= \frac{\operatorname{Cov}\left(s(0), \sum\limits_{j=0}^{\frac{t}{\Delta t}-1}r(j\Delta t, (j+1)\Delta t)\right)}{\operatorname{Std}\left(\sum\limits_{j=0}^{\frac{t}{\Delta t}-1}r(j\Delta t, (j+1)\Delta t)\right)}\\
&= \frac{IC\cdot\omega\cdot\sqrt{\Delta t}\cdot\sum\limits_{j=0}^{\frac{t}{\Delta t}-1}\delta^{j}}{\omega\sqrt{t}}\\
&= IC\cdot\sqrt{\frac{t}{\Delta t}}\cdot\frac{1-\delta^{\frac{\Delta t}{t}}}{1-\delta}
\end{aligned}
$$

下面考虑当期信号和所有滞后信号的最优混合问题。进一步假设任意两滞后信号的协方差和相关系数的形式为：
$$
\operatorname{Cov}(s(-j\Delta t),s(-k\Delta t)) = \operatorname{Corr}(s(-j\Delta t),s(-k\Delta t)) = \rho^{|j-k|}
$$
同最优混合策略时类似，所有信号的相关系数矩阵一个无限的 Toeplitz 矩阵：
$$
\boldsymbol{\rho}_s = \begin{pmatrix} 1 & \rho & \rho^2 & \cdots & \rho^j & \cdots \\ \rho & 1 & \rho & \cdots & \rho^{j-1} & \cdots \\ \rho^2 & \rho & 1 & \cdots & \rho^{j-2} & \cdots \\ \vdots & \vdots & \vdots & \ddots & \vdots & \cdots \\ \rho^{j} & \rho^{j-1} & \rho^{j-2} & \cdots & 1 & \cdots \\ \vdots & \vdots & \vdots & \vdots & \vdots & \ddots \end{pmatrix}
$$
其逆为：
$$
\boldsymbol{\rho}_s^{-1} = \frac{1}{1-\rho^2}\begin{pmatrix} 1 & -\rho & 0 & \cdots \\ -\rho & 1+\rho^2 & -\rho & \cdots \\ 0 & -\rho & 1+\rho^2 & \cdots \\ \vdots & \vdots & \vdots & \ddots \end{pmatrix}
$$
记信号向量和信息系数向量为：
$$
\mathbf{s} = \begin{pmatrix} s(0) \\ s(-\Delta t) \\ s(-2\Delta t) \\ \vdots \end{pmatrix},\ \mathbf{IC} = IC\cdot\boldsymbol{\delta} = IC\cdot\begin{pmatrix} 1 \\ \delta \\ \delta^2 \\ \vdots \end{pmatrix}
$$
根据预测基本公式，最优混合信号为：
\begin{equation*}
\begin{aligned}
s^*(0) &= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \boldsymbol{\delta}'\cdot\boldsymbol{\rho}_s^{-1}\cdot\mathbf{s}\\
&= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\begin{pmatrix} 1 & \delta-\rho & \delta(\delta-\rho) & \delta^2(\delta-\rho) & \cdots \end{pmatrix}\cdot\mathbf{s}\\
&= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\left(\boldsymbol{\delta} - \rho\cdot\mathbf{S}\cdot\boldsymbol{\delta}\right)\cdot\mathbf{s}\\
&= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\sum\limits_{k\geq0}\left[s(-k\Delta t) - \rho\cdot s\left(-(k+1)\Delta t\right)\right]\\
&= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\left[s(0) - \rho\cdot s(-\Delta t)\right] + \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\delta\sum\limits_{k\geq1}\delta^{k-1}\left[s(-k\Delta t) - \rho\cdot s(-(k+1)\Delta t)\right]\\
&= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\left[s(0) - \rho\cdot s(-\Delta t)\right] + \delta\cdot\omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\sum\limits_{l\geq0}\delta^{l}\left[s(-(l+1)\Delta t) - \rho\cdot s(-(l+2)\Delta t)\right]\ (l=k-1)\\
&= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\left[s(0) - \rho\cdot s(-\Delta t)\right] + \delta\cdot s^*(-\Delta t)
\end{aligned}
\end{equation*}
其中：
$$
\mathbf{S} = \begin{pmatrix} 0 & 0 & 0 & \cdots \\ 1 & 0 & 0 & \cdots \\ 0 & 1 & 0 & \cdots \\ \vdots & \vdots & \vdots & \ddots \end{pmatrix}
$$
总之，最优混合信号表达为：
$$
\begin{aligned}
s^*(0) &= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\sum\limits_{k\geq0}\left[s(-k\Delta t) - \rho\cdot s\left(-(k+1)\Delta t\right)\right]\\
&= \omega\cdot\sqrt{\Delta t}\cdot IC\cdot \frac{1-\rho\delta}{1-\rho^2}\cdot\left[s(0) - \rho\cdot s(-\Delta t)\right] + \delta\cdot s^*(-\Delta t)
\end{aligned}
$$
这同最优混合策略在形式上是一致的。

最优混合信号的标准差为：
$$
\operatorname{Std}(s^*(0)) = \omega\cdot\sqrt{\Delta t}\cdot IC\cdot\frac{1-\rho\delta}{\sqrt{(1-\rho^2)(1-\delta^2)}}
$$
最优混合信号的信息系数为：
$$
IC^* = IC\cdot\frac{1-\rho\delta}{\sqrt{(1-\rho^2)(1-\delta^2)}} \geq IC
$$
当且仅当 $\rho=\delta$ 时有 $IC^*=IC$。

# 多因子模型

考虑多个特征或者因子：$\mathbf{X} = (\mathbf{x}_1, \ldots, \mathbf{x}_K)$，其中，$\mathbf{x}_k$ 是一个特征向量，$\mathbf{X}$ 是 K 个特征组成的特征矩阵或因子暴露矩阵。

对于资产收益率建立多因子模型：
\begin{equation}\label{MultiFactorModel}
\mathbf{r} = \mathbf{X}\mathbf{b} + \mathbf{u}
\end{equation}
其中，$\mathbf{X}$ 是资产对于因子的暴露矩阵（$N\times K$ 维），$X_{n,k}$ 是资产 n 对因子 k 的暴露（Exposure）或者因子载荷（Factor Loading）；$\mathbf{b}$ 是因子收益率向量（$K\times 1$ 维），$b_k$ 是因子 k 的收益率，$\mathbf{u}$ 是资产的特异收益率（Specific Return）向量（$N\times 1$ 维），$u_n$ 资产 n 的特异收益率。

记资产的预期超额收益率为 $\mathbf{f}=\mathbb{E}\mathbf{r}$，则预期收益率模型主要是要对 $\mathbf{f}$ 建模。

**基本假设**
1. 假设特异收益率 $\mathbf{u}$ 与因子收益率 $\mathbf{b}$ 互不相关：$\operatorname{Cov}(u_n, b_k)=0$ 对任意 n 和 k 成立，即：
$$
\operatorname{Cov}(\mathbf{u}, \mathbf{b})=\mathbb{E}\left[(\mathbf{u}-\mathbb{E}\mathbf{u})(\mathbf{b}-\mathbb{E}\mathbf{b})'\right]=\mathbf{0}
$$
2. 因子暴露 $\mathbf{X}$ 于期初已知。
3. 最小方差组合的预期超额收益率大于 0，即 $f_C = \mathbf{f}'\mathbf{h}_C >0$，从而组合 Q 存在且 $f_Q>0$。

在上述假设 1 下，资产收益率的协方差矩阵 $\mathbf{V}$ 表示为：
$$
\mathbf{V} = \mathbf{X}\mathbf{F}\mathbf{X}'+\boldsymbol{\Delta}
$$
其中，$\mathbf{F}=\operatorname{Var}(\mathbf{b})$ 为因子收益的协方差矩阵，$\boldsymbol{\Delta}=\operatorname{Var}(\mathbf{u})$ 是资产特异收益率的协方差矩阵。

# 套利定价（APT）模型

APT 认为每个资产的预期超额收益率由资产的因子暴露决定。

因子模型解释了预期超额收益率 $\mathbf{f}$，如果预期超额收益率 $\mathbf{f}$ 可以表示为因子暴露 $\mathbf{X}$ 的线性组合，即存在 **因子预测** $\mathbf{m}$，使得：
$$
\mathbf{f} = \mathbf{X}\cdot\mathbf{m}
$$
APT 只给出了因子预测的存在性，但没有给出构造因子预测的方法。

记多因子模型为 $(\mathbf{X},\mathbf{F},\boldsymbol{\Delta})$，称组合 p 相对于因子模型 **充分分散**，如果组合 p 在所有与其具有相同因子暴露的组合中风险最低，即:
$$
\sigma_p\leq\sigma_r,\forall \mathbf{X}'\mathbf{h}_r=\mathbf{x}_q
$$

**套利定价理论**：因子模型 $(\mathbf{X},\mathbf{F},\boldsymbol{\Delta})$ 能够解释预期超额收益率 $\mathbf{f}$ 当且仅当组合 Q 相对于该模型充分分散。

**证明**：

(1) 首先假设组合 Q 相对于 $(\mathbf{X},\mathbf{F},\boldsymbol{\Delta})$ 充分分散，即 $\mathbf{h}_Q$ 是如下优化问题的一个解：
$$
\begin{aligned}
& \mathop{\min}_{\mathbf{h}}\mathbf{h}'\mathbf{V}\mathbf{h}\\
& \rm{s.t.}\ \mathbf{X}'\mathbf{h} = \mathbf{x}_Q
\end{aligned}
$$
Lagrange 函数：
$$
L(\mathbf{h},\boldsymbol{\lambda}) = \mathbf{h}\mathbf{V}\mathbf{h} + \boldsymbol{\lambda}'\left(\mathbf{X}'\mathbf{h}-\mathbf{x}_Q\right)
$$
则 $\mathbf{h}_Q$ 满足一阶条件
$$
\left\{\begin{aligned}
& \frac{\partial L}{\partial\mathbf{h}} = 2\mathbf{V}\mathbf{h}_Q + \mathbf{X}\boldsymbol{\lambda} = \mathbf{0}\\
& \frac{\partial L}{\partial\boldsymbol{\lambda}} = \mathbf{X}'\mathbf{h}_Q - \mathbf{x}_Q = \mathbf{0}
\end{aligned}\right.
$$
则：
$$
\mathbf{f} = f_Q\boldsymbol{\beta}^Q = f_Q\frac{\mathbf{V}\mathbf{h}_Q}{\sigma_Q^2} = \frac{f_Q}{\sigma_Q^2}\mathbf{X}\boldsymbol{\lambda}
$$
令 $\mathbf{m}=\frac{f_Q}{\sigma_Q^2}\boldsymbol{\lambda}$ 即得。

(2) 假设因子模型 $(\mathbf{X},\mathbf{F},\boldsymbol{\Delta})$ 能够解释预期超额收益率 $\mathbf{f}$，即 $\exists\mathbf{m},s.t.\ \mathbf{f}=\mathbf{X}\cdot\mathbf{m}$，对于任意组合 p，如果 $\mathbf{x}_p=\mathbf{x}_Q$，则：
$$
f_p = \mathbf{m}'\mathbf{X}'\mathbf{h}_p = \mathbf{m}'\mathbf{x}_p = \mathbf{m}'\mathbf{x}_Q = f_Q
$$
因为组合 Q 是最大夏普率组合，则：
$$
\frac{f_p}{\sigma_p}\leq\frac{f_Q}{\sigma_Q}
$$
由假设 3，$f_C>0$ 可得 $f_Q>0$，
从而 $\sigma_Q\leq\sigma_p$。

# 因子组合

定义 k 因子的 **因子组合（Factor Portfolio）** 或者 **因子模拟组合（Factor Mimicking Portfolio）** 为对 k 因子具有单位暴露，而对其他因子零暴露的最小风险投资组合：
$$
\begin{aligned}
\mathbf{h}_k = \ & \mathop{\arg\min}_{\mathbf{h}}\mathbf{h}'\mathbf{V}\mathbf{h}\\
& \rm{s.t.}\ \mathbf{X}'\mathbf{h}=\boldsymbol{\iota}_k
\end{aligned}
$$
由[等式约束条件下的二次规划](.\最优化.ipynb\#QuandprogWithEqConstraint)可解得：
$$
\mathbf{h}_k = \mathbf{V}^{-1}\mathbf{X}(\mathbf{X}'\mathbf{V}^{-1}\mathbf{X})^{-1}\boldsymbol{\iota}_k
$$
所有 K 个因子的因子组合表示为：
$$
\mathbf{W} = (\mathbf{h}_1,\ldots,\mathbf{h}_K) = \mathbf{V}^{-1}\mathbf{X}(\mathbf{X}'\mathbf{V}^{-1}\mathbf{X})^{-1}
$$
其中，$\mathbf{W}$ 的第 k 列即为 k 因子的因子组合。

**多因子模型下的因子组合**

考虑在多因子模型假设下的因子组合，其优化的目标函数可调整为：
$$
\mathbf{h}'\mathbf{V}\mathbf{h} = (\mathbf{X}'\mathbf{h})'\cdot\mathbf{F}\cdot(\mathbf{X}'\mathbf{h}) + \mathbf{h}'\boldsymbol{\Delta}\mathbf{h} = \boldsymbol{\iota}'_k\mathbf{F}\boldsymbol{\iota}_k + \mathbf{h}'\boldsymbol{\Delta}\mathbf{h} = F_{k,k} + \mathbf{h}'\boldsymbol{\Delta}\mathbf{h}
$$
原问题等价于最小化 $\mathbf{h}'\boldsymbol{\Delta}\mathbf{h}$。

解为：
$$
\mathbf{W} = \boldsymbol{\Delta}^{-1}\mathbf{X}(\mathbf{X}'\boldsymbol{\Delta}^{-1}\mathbf{X})^{-1}\boldsymbol{\iota}_k
$$

如果将 (\ref{MultiFactorModel}) 式看做横截面回归方程，则 $\mathbf{W}'$ 刚好是广义最小二乘(GLS)估计的投影矩阵：
$$
\hat{\mathbf{b}} = (\mathbf{X}'\boldsymbol{\Delta}^{-1}\mathbf{X})^{-1}\mathbf{X}'\boldsymbol{\Delta}^{-1}\mathbf{r} = \mathbf{W}'\mathbf{r}
$$
从而，因子收益率可以看做其对应的因子组合的收益率。
