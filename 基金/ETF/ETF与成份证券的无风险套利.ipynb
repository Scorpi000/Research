# 摘要

本部分主要研究 ETF 一二级市场间套利策略. ETF 的申购赎回机制和 T+0 回转交易使得在一二级市场间套利成为可能. 下面主要给出了套利策略的基本模型以及套利成本的分析. 

**关键字**: ETF 套利, 溢价套利, 折价套利, 折溢价率, 冲击成本, 延时成本

# 理论模型

## 折溢价套利

<img src="./images/ETF折溢价套利.png" title="ETF 折溢价套利"/>

**折价套利**: ETF 市场价格 P 小于 ETF 份额参考净值 IOPV 时, 通过在场内买入 ETF 赎回成一篮子股票再立即卖出所有股票获利. 假设该过程的平均单位套利成本为 $C_1$, 其数值等于套利过程中付出的总成本除以 ETF 的总份额(如果以最小申赎单位来操作, 则总份额为最小申赎单位), 如果没有套利机会, 就应该有 $IOPV\le P+C_1$.

**溢价套利**: ETF 市场价格 P 高于 ETF 份额参考净值 IOPV 时, 通过在场外买入一篮子股票申购成 ETF 再立即卖出 ETF 获得收益. 假设该过程的平均单位套利成本为 $C_2$, 其数值等于套利过程中付出的总成本除以 ETF 的总份额(如果以最小申赎单位来操作, 则总份额为最小申赎单位), 如果没有套利机会, 就应该有 $P\le IOPV+C_2$.

**折/溢价率** ＝ (ETF 二级市场价格 P - ETF 份额参考净值 IOPV) / ETF 份额参考净值 IOPV × 100%

<img src="./images/ETF折溢价套利区间.png" title="ETF 折溢价套利区间"/>

综上所得, ETF 市场价格的无套利区间为:
$$
IOPV_t − C_1 \le P_t \le IOPV_t + C_2
$$
或者以折溢价率的形式表达为:
$$
-\frac{C_1}{IOPV_t} \le \frac{P_t-IOPV_t}{IOPV_t} \le \frac{C_2}{IOPV_t}
$$

## 成本分析

要想套利成功，必须考虑市场价格与 ETF 净值之间的差异是否足够覆盖套利交易的成本，这部分成本包括固定成本和可变成本。

![ETF 折价套利相关成本](./images/ETF折价套利相关成本.png)


### 固定成本

固定成本包括股票交易佣金、交易税费、ETF 交易佣金和 ETF 申购赎回费用, 详情可见: ETF 相关费用

### 可变成本

可变成本主要包括冲击成本和延时成本:
1. **冲击成本(或流动性成本)**: 套利交易时需在现货市场买进卖出相当规模之股票和 ETF, 可能影响到的市场价格变化;
2. **延时成本(或等待成本)**: 在交易过程中, 买卖一篮子股票组合或基金份额时, 只有等到流动性最差的交易环节完成了交易, 这个套利过程才算完成, 在等待成交的过程中需要付出的成本就是延时成本.

一般来说, 在交易过程中, 为了降低冲击成本, 扩大套利交易的规模, 投资者会选择延长套利交易的时间, 而延长交易时间的同时, 交易的不确定性也在增加, 套利者的委托指令可能不能全部成交, 从而修改指令价格, 导致增加成本. 如何优化交易时间长度, 使得套利利润最大化对于套利者来说是最重要的, 而正确估计套利交易的可变成本是得到最优套利策略的基础.

单只证券的可变成本，取决于投资者需要交易的证券数量及证券的流动性, 一个组合的可变成本同样也取决于该组合的流动性, 但组合的流动性是由组合所有成份证券中流动性最差的证券决定的. 在构建组合时, 整个组合交易金额及成份证券的比例决定了每一只成份证券的交易数量. 只有组合中挂单数量与个券需要交易的数量的比例最小, 即相对流动性最差的证券也完成了交易， 整个组合才能得以构建完成, 其它先行完成交易的成份证券与最后完成交易的成份证券一样都有相同的等待成本. 对于冲击成本而言, 构建各个证券的头寸而支付的冲击成本是不会完全相同的, 组合冲击成本等于所有成份证券冲击成本之和. 因此组合的可变成本与个券的可变成本显然存在着明显的区别。

**冲击成本**

冲击成本可以看成是需要成交的证券数量与证券当前的流动性的二元函数.

为了简化分析, 暂且固定证券规模变量, 即限定一次套利的规模为最小申赎单位下讨论买卖股票和 ETF 的冲击成本.

在计算一篮子证券的冲击成本时, 为了降低等待成本, 需要迅速完成交易, 则需要考虑到买卖价差的影响.
为了迅速成交最小申赎单位下各股票所需要交易的手数, 则需要以对手盘档位上的 **最优一次全部成交价格** 成交.

最优一次全部成交价格的解释以溢价套利为例, 最小申赎单位 ETF 对应的股票组合中贵州茅台为 300 股即 3 手, 当前的卖一价位 203.96, 卖一量为 2 手, 卖二价为 203.97, 卖二量为 12 手，此时为了完成交易则最优一次全部成交价格为卖二价 203.97 元. 

臧大年、林文政(1996)对买卖一篮子股票组合的市场冲击成本估计方法如下:
$$
\sum\limits_{i=1}^{N}\frac{T_i-P_i}{P_i}\times\varphi_i
$$
其中, $T_i$ 为证券 i 的最优一次全部成交价格, $P_i$ 为证券 i 的当前的市场价格, $\varphi_i$ 为证券 i 在整个组合中的权重.

**延时成本**

一种估计延时成本的方法是基于风险准备金的思想.

估算延时成本也就是要测定套利者持有股票(或 ETF)的期间风险并对该风险进行补偿，一种最直接的思路是对于不同持有期的指数头寸提取一定的风险准备金，如假定持有指数头寸一年需要提取 10% 的风险准备(假定数字下同), 而持有一天指数头寸需要提取 20 BP 的风险准备, 持有一小时需要提取 10 BP 的风险准备, 持有指数头寸 10 分钟需要提取 2 BP 的风险准备等. 至于具体的数字取决于在一定概率保证条件下指数在该时间内可能的下跌幅度. 这里可以借鉴 VAR 指标的定义思路, VAR 指标指的是在一个较大的概率保证下(如 95%, 两倍标准差)一项资产可能的最大损失值, 而这里的风险准备是指在一定概率保证下一项资产可能发生的损失值,  并以此风险准备来对风险暴露进行足量的补偿, 从而合理估算等待成本. 在计算 VAR 值时采用较大的概率保证水平是因为 VAR 为了反映可能损失的最大值, 而在本文主要是对风险在一个较长的时间内进行足额的补偿. 因此本文中只是借用 VAR 的概率算法来计算在一定时间持有指数头寸应该提取的合适的风险准备比例而已, 如(56%，0.15 倍标准差).

延时成本的计算公式为:
$$
\bar{R}-\sigma\times\lambda_\sigma
$$
其中, $\bar{R}$ 为头寸的平均收益率, $\sigma$ 为收益率的标准差, $\lambda_\sigma$ 为为取定的标准差倍数.

另外一种计算方法是先估计延迟的时间, 然后考察该时间内证券价格的变化. 假设在交易前证券的价格为 $P_t$, 交易完成后的价格为 $P_{t+1}$, 则此时可以按如下方法计算成本，即:
$$        
\frac{P_{t+1}-P_t}{P_t}
$$
这里使用第二种方法来估算延时成本.

```python
# coding=utf-8
import sys
sys.path.append("C:\\Users\\hushuntai\\svn\\python")
sys.path.append("D:\\HST\\QuantStudio")
import datetime as dt

import numpy as np
import pandas as pd
import statsmodels.api as sm
from matplotlib.pylab import mpl
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm_notebook

import QuantStudio.api as QS
WDB = QS.FactorDB.WindDB2()
```

下面主要以上证 50 ETF 为例进行相关的实证分析, 使用的数据包括:
1. ETF 申购赎回清单. 来自于 Wind 量化研究数据库中的表: CHINAETFPCHREDMLIST, CHINAETFPCHREDMMEMBERS.

3. ETF 和申赎清单中成份股票的日内数据. 来自于 epsilon 数据库, 采样频率 3 秒.

设置参数并准备数据.

```python
# 参数设置和数据准备
TargetID = '510050.SH'# 测试用例: 50ETF, 510050.SH, 深 100ETF, 159901.SZ, 嘉实富时中国A50ETF, 512550.SH
StartDT = dt.datetime(2017, 10, 19)
EndDT = dt.datetime(2017, 11, 17, 23, 59, 59, 999999)
tms_intv = 3# data matrix interval
if TargetID[-2:]=="SH":
    DayTimeNum = int(28800/2/tms_intv+2)
else:
    DayTimeNum = int(28440/2/tms_intv+2)

WDB.connect()
DTs = WDB.getTable("中国A股日行情").getDateTime(start_dt=StartDT, end_dt=EndDT)
PCF = WDB.getTable("中国ETF申购赎回清单").readData(ids=[TargetID], dts=DTs).iloc[:, :, 0]
PCFCon = WDB.getTable("中国ETF申购赎回成份").readData(dts=DTs, args={"基金ID":"510050.SH"})
PreClose = WDB.getTable("中国A股日行情").readData(factor_names=["昨收盘价", "交易状态"], ids=list(PCFCon.minor_axis), dts=DTs)
TradeStatus, PreClose = PreClose["交易状态"], PreClose["昨收盘价"]
PreClose.insert(loc=0, column=TargetID, value=WDB.getTable("中国封闭式基金日行情").readData(factor_names=["昨收盘价(元)"], ids=[TargetID], dts=DTs).iloc[0, :, 0])

from QuantStudio.FactorDataBase.CloverDB import CloverDB
CDB = CloverDB()
CDB.connect()
FT = CDB.getTable("Tick 数据", args={"时间间隔":tms_intv})

DateNum = PCF.shape[0]

# 计算 ETF 份额参考净值
# price: ETF 成分证券的价格, DataFrame(index=[时点], columns=[ID])
# pcf: 申购赎回清单, Series(index=[现金差额, 最小申购赎回单位资产净值, 预估现金部分, 现金替代比例上限, 最小申购赎回单位, 申购赎回允许情况]
# pcf_cons: 申购赎回清单的成分证券, DataFrame(index=[ID], columns=[股票数量, 现金替代标志, 现金替代溢价比例, 固定替代金额])
#           现金替代标志：0：禁止现金替代；1：允许现金替代；2：必须现金替代；3：退补现金替代
# 返回: IOPV, Series(index=[时点])
def calcIOPV(price, pcf, pcf_cons):
    MinPCUnitAsset = np.zeros(price.shape[0]) + pcf.loc["预估现金部分"]# 预估现金差额
    for i in range(pcf_cons.shape[0]):
        if pcf_cons["现金替代标志"].iloc[i]==2:# 必须现金替代
            MinPCUnitAsset += pcf_cons["固定替代金额"].iloc[i]
        else:# 禁止现金替代, 允许现金替代, 退补现金替代
            iPrice = price.iloc[:, i].values
            MinPCUnitAsset += iPrice * pcf_cons["股票数量"].iloc[i]
    return pd.Series(MinPCUnitAsset / pcf.loc["最小申购赎回单位"], index=price.index)
```

## 套利空间

根据每个交易日的日内数据, 首先计算 IOPV, 而后计算折溢价率, 记录每个时点的折溢价率和该交易日的平均折溢价率, 最后计算所有折溢价率数据以及日平均数据的描述统计量并绘制分布直方图. 通过统计数据来看, 50 ETF 的平均折溢价率为 -6.5 BP, ETF 长期处于折价状态.

```python
# 折溢价率
IOPV = np.zeros((DayTimeNum, DateNum))
ZYRate = np.zeros((DayTimeNum, DateNum))# 折溢价率
for i, iDT in enumerate(tqdm_notebook(PCF.index, desc="日期 loop")):
    iPCFConMask = pd.notnull(PCFCon["现金替代标志"].iloc[i])
    iPCFConIDs = list(iPCFConMask[iPCFConMask].index)
    iLast = FT.readData(factor_names=["lst"], ids=[TargetID]+iPCFConIDs, 
                        start_dt=dt.datetime.combine(iDT.date(), dt.time(0)), end_dt=iDT).iloc[0]
    iLast.iloc[0] = iLast.iloc[0].where(pd.notnull(iLast.iloc[0]), PreClose.iloc[i].loc[iLast.columns])
    iLast.fillna(method="pad", inplace=True)
    iIOPV = calcIOPV(iLast.iloc[:, 1:], PCF.iloc[i], PCFCon.iloc[:, i].loc[iPCFConIDs])
    IOPV[:, i] = iIOPV.values
    ZYRate[:, i] = ((iLast.iloc[:, 0] - iIOPV) / iIOPV).values
ZYRate_DayAvg = np.nanmean(ZYRate, axis=0)

Fig, Axes = plt.subplots(1, 2, figsize=(25, 8))
Axes[0].hist(ZYRate.flatten()[pd.notnull(ZYRate.flatten())], bins="auto")
Axes[0].set_title("折溢价率-日内数据")
Axes[1].hist(ZYRate_DayAvg, bins="auto")
Axes[1].set_title("折溢价率-日平均数据");
```

## 固定成本分析

溢价套利固定成本 = 买入股票的交易税费 + ETF 申购费用 + 卖出 ETF 的交易税费 = (0.00487% + 0.002% + 0.002% + 0.003%) + (0.025%) + (0.00487% + 0.003%) = 0.0474% = 5 BP

折价套利固定成本 = 买入 ETF 的交易税费 + ETF 赎回费用 + 卖出股票的交易税费 = (0.00487% + 0.003%) + (0.025%) + (0.1% + 0.00487% + 0.002% + 0.002% + 0.003%) = 0.1474% = 15 BP

## 冲击成本分析

**组合证券的冲击成本**

根据每个交易日的日内数据, 在每个时点上假设进行 ETF 最小申赎单位证券组合的买卖, 对于买入证券组合, 计算以卖一价(以后有 level 2 行情数据应以最优一次全部成交价格来计算)成交证券组合相对于最新价(对于当前最新价缺失的数据, 以前收盘价数据填充)的变动率作为该时点的冲击成本, 对于卖出证券组合, 计算以买一价(以后有 level 2 行情数据应以最优一次全部成交价格来计算)成交相对于最新价的变动率作为该时点的冲击成本. 记录每个时点的冲击成本和该交易日的平均冲击成本.

**ETF 冲击成本**

根据每个交易日的日内数据, 在每个时点上假设进行 ETF 最小申赎单位的买卖, 对于买入 ETF, 计算以卖一价(以后有 level 2 行情数据应以最优一次全部成交价格来计算)成交相对于最新价(对于当前最新价缺失的数据, 以第一笔交易的成交价填充, 以后有收盘价数据当以前收盘价填充)的变动率作为该时点的冲击成本, 对于卖出 ETF, 计算以买一价(以后有 level 2 行情数据应以最优一次全部成交价格来计算)成交相对于最新价的变动率作为该时点的冲击成本.

通过统计数据来看, 按照 95% 的置信水平, 可以设置组合证券的冲击成本为: 5 BP, ETF 的冲击成本为: 4 BP.

```python
# 冲击成本估计
BuyETFImpactCost = np.zeros((DayTimeNum, DateNum))# 买入 ETF 的冲击成本
SellETFImpactCost = np.zeros((DayTimeNum, DateNum))# 卖出 ETF 的冲击成本
Val_Bid = np.zeros((DayTimeNum, DateNum))# 以买一价计算的证券组合价值
Val_Last_Bid = np.zeros((DayTimeNum, DateNum))# 以最新价计算的证券组合价值
Val_Ask = np.zeros((DayTimeNum, DateNum))# 以卖一价计算的证券组合价值
Val_Last_Ask = np.zeros((DayTimeNum, DateNum))# 以最新价计算的证券组合价值
for i, iDT in enumerate(tqdm_notebook(PCF.index, desc='日期 loop')):
    iPCFCon = PCFCon.iloc[:, i, :]
    iPCFCon = iPCFCon[pd.notnull(iPCFCon["现金替代标志"])]
    iIntradayData = FT.readData(factor_names=["lst", "bid", "ask"], ids=list(iPCFCon.index)+[TargetID], 
                                start_dt=dt.datetime.combine(iDT.date(), dt.time(0)), end_dt=iDT)
    iIntradayData["lst"].iloc[0] = iIntradayData["lst"].iloc[0].where(pd.notnull(iIntradayData["lst"].iloc[0]), PreClose.iloc[i].loc[iIntradayData.minor_axis])
    iIntradayData["lst"].fillna(method="pad", inplace=True)
    # 估计组合证券的冲击成本
    for j, jID in enumerate(iPCFCon.index):
        if iPCFCon.loc[jID, "现金替代标志"]!=2:
            ijLast = iIntradayData["lst"].iloc[:, j]
            ijAsk = iIntradayData["ask"].iloc[:, j]
            ijMask = pd.notnull(ijAsk)
            Val_Ask[:, i] += (ijAsk * iPCFCon["股票数量"].iloc[j]).where(ijMask, 0)
            Val_Last_Ask[:, i] += (ijLast * iPCFCon["股票数量"].iloc[j]).where(ijMask, 0)
            ijBid = iIntradayData["bid"].iloc[:, j]
            ijMask = pd.notnull(ijBid)
            Val_Bid[:, i] += (ijBid * iPCFCon["股票数量"].iloc[j]).where(ijMask, 0)
            Val_Last_Bid[:, i] += (ijLast * iPCFCon["股票数量"].iloc[j]).where(ijMask, 0)
    iLast = iIntradayData["lst"].iloc[:, -1]
    # 估计 ETF 的买入冲击成本
    iAsk = iIntradayData["ask"].iloc[:, -1]
    BuyETFImpactCost[:, i] = (iAsk - iLast) / iLast
    # 估计 ETF 的卖出冲击成本
    iBid = iIntradayData["bid"].iloc[:, -1]
    SellETFImpactCost[:, i] = (iLast - iBid) / iLast
BuyConImpactCost = (Val_Ask - Val_Last_Ask) / Val_Last_Ask# 买入证券组合冲击成本
SellConImpactCost = (Val_Last_Bid - Val_Bid) / Val_Last_Bid# 卖出证券组合冲击成本
BuyConImpactCost_DayAvg = np.nanmean(BuyConImpactCost, axis=0)# 买入证券组合冲击成本, 日平均
SellConImpactCost_DayAvg = np.nanmean(SellConImpactCost, axis=0)# 卖出证券组合冲击成本, 日平均
BuyETFImpactCost_DayAvg = np.nanmean(BuyETFImpactCost, axis=0)# 买入 ETF 冲击成本, 日平均
SellETFImpactCost_DayAvg = np.nanmean(SellETFImpactCost, axis=0)# 卖出 ETF 冲击成本, 日平均

Fig, Axes = plt.subplots(4, 2, figsize=(15, 16))
Axes[0, 0].hist(BuyConImpactCost.flatten()[pd.notnull(BuyConImpactCost.flatten())], bins="auto")
Axes[0, 0].set_title("买入证券组合冲击成本-日内数据")
Axes[0, 1].hist(SellConImpactCost.flatten()[pd.notnull(SellConImpactCost.flatten())], bins="auto")
Axes[0, 1].set_title("卖出证券组合冲击成本-日内数据")
Axes[1, 0].hist(BuyConImpactCost_DayAvg[pd.notnull(BuyConImpactCost_DayAvg)], bins="auto")
Axes[1, 0].set_title("买入证券组合冲击成本-日平均数据")
Axes[1, 1].hist(SellConImpactCost_DayAvg[pd.notnull(SellConImpactCost_DayAvg)], bins="auto")
Axes[1, 1].set_title("卖出证券组合冲击成本-日平均数据")
Axes[2, 0].hist(BuyETFImpactCost.flatten()[pd.notnull(BuyETFImpactCost.flatten())], bins="auto")
Axes[2, 0].set_title("买入 ETF 冲击成本-日内数据")
Axes[2, 1].hist(SellETFImpactCost.flatten()[pd.notnull(SellETFImpactCost.flatten())], bins="auto")
Axes[2, 1].set_title("卖出 ETF 冲击成本-日内数据")
Axes[3, 0].hist(BuyETFImpactCost_DayAvg[pd.notnull(BuyETFImpactCost_DayAvg)], bins="auto")
Axes[3, 0].set_title("买入 ETF 冲击成本-日平均数据")
Axes[3, 1].hist(SellETFImpactCost_DayAvg[pd.notnull(SellETFImpactCost_DayAvg)], bins="auto")
Axes[3, 1].set_title("卖出 ETF 冲击成本-日平均数据");
```

## 延时成本分析

**组合证券的延时成本**

这里以累积交易量超过待交易的证券数量作为该证券完成交易的标志, 计算这段时间的价格变动作为该证券的延时成本, 组合的等待时间定义为组合中等待时间最长的证券的等待时间.

根据每个交易日的日内数据, 在每个时点上假设进行 ETF 最小申赎单位对应的证券组合的买卖, 对于每个证券, 以未来累积成交量超过最小申赎单位中该证券数量的时间点距离当前时间点的时间长度(以秒为单位)作为本次该证券交易的等待时间, 以该段时间中证券价格变化率作为延时成本, 所有证券的延时成本按照组合权重加权平均作为组合的延时成本, 组合的等待时间为所有证券等待时间的最大值. 记录每个时点的等待时间、延时成本和该交易日的平均等待时间、平均延时成本.

通过统计数据来看, 按照 95% 的置信水平, 组合证券延时 30 秒, 可以设置组合证券的延时成本为: 1 BP

```python
# 组合证券的延时成本估计
ConDecayTime = np.zeros((DayTimeNum, DateNum))# 交易证券组合的延迟时间
ConDecayCost = np.zeros((DayTimeNum, DateNum))# 在延迟时间中的收益率
ijMask = None
for i, iDT in enumerate(tqdm_notebook(PCF.index, desc='日期 loop')):
    iPCFCon = PCFCon.iloc[:, i, :]
    iPCFCon = iPCFCon[pd.notnull(iPCFCon["现金替代标志"])]
    ijMask = np.ones((DayTimeNum+1, iPCFCon.shape[0]))
    iIntradayData = FT.readData(factor_names=["lst", "vol"], ids=list(iPCFCon.index), 
                                start_dt=dt.datetime.combine(iDT.date(), dt.time(0)), end_dt=iDT)
    iIntradayData["lst"].iloc[0] = iIntradayData["lst"].iloc[0].where(pd.notnull(iIntradayData["lst"].iloc[0]), PreClose.iloc[i].loc[iIntradayData.minor_axis])
    iIntradayData["lst"].fillna(method="pad", inplace=True)
    iLast, iVol = iIntradayData["lst"].values, iIntradayData["vol"].values
    iConTargetNum = iPCFCon["股票数量"].values / 100
    iConTargetNum[(iPCFCon["现金替代标志"].values==2) | (TradeStatus.iloc[i][iPCFCon.index].values!="交易")] = 0# 必须现金替代, 停牌
    for j in range(DayTimeNum):
        ijLast, ijVol = iLast[j:, :], iVol[j:, :]
        ijMask[j, :] = 1
        ijMask[j+1:, :] = (np.cumsum(ijVol, axis=0) < np.repeat(iConTargetNum.reshape((1, iConTargetNum.shape[0])), axis=0, repeats=DayTimeNum-j))
        ijDecayTime = np.sum(ijMask[j+1:, :], axis=0)
        ConDecayTime[j, i] = np.max(ijDecayTime) * tms_intv
        ijAvgPrice = np.nansum(ijMask[j:-1, :] * ijVol * ijLast, axis=0) / np.nansum(ijMask[j:-1, :] * ijVol, axis=0)
        ijAvgPrice[np.isnan(ijAvgPrice)] = ijLast[0, np.isnan(ijAvgPrice)]
        ijCost = ijAvgPrice / ijLast[0, :] - 1
        ijPortfolio = ijLast[0, :] * iConTargetNum / np.nansum(ijLast[0, :] * iConTargetNum)
        ConDecayCost[j, i] = np.nansum(ijPortfolio * ijCost)
ConDecayTime_DayAvg = np.nanmean(ConDecayTime, axis=0)
ConDecayCost_DayAvg = np.nanmean(ConDecayCost, axis=0)

Fig, Axes = plt.subplots(2, 2, figsize=(15, 8))
Axes[0, 0].hist(ConDecayTime.flatten()[pd.notnull(ConDecayTime.flatten())], bins="auto")
Axes[0, 0].set_title("交易证券组合延迟时间-日内数据")
Axes[0, 1].hist(ConDecayCost.flatten()[pd.notnull(ConDecayCost.flatten())], bins="auto")
Axes[0, 1].set_title("交易证券组合延迟成本-日内数据")
Axes[1, 0].hist(ConDecayTime_DayAvg[pd.notnull(ConDecayTime_DayAvg)], bins="auto")
Axes[1, 0].set_title("交易证券组合延迟时间-日平均数据")
Axes[1, 1].hist(ConDecayCost_DayAvg[pd.notnull(ConDecayCost_DayAvg)], bins="auto")
Axes[1, 1].set_title("交易证券组合延迟成本-日平均数据");
```

**ETF 延时成本**

这里以累积交易量超过待交易的 ETF 数量作为完成交易的标志, 计算这段时间的价格变动作为延时成本.

根据每个交易日的日内数据, 在每个时点上假设进行 ETF 最小申赎单位的买卖,  以未来累积成交量超过最小申赎单位份额的时间点距离当前时间点的时间长度(以秒为单位)作为本次交易的等待时间, 以该段时间中 ETF 价格变化率作为延时成本. 记录每个时点的等待时间、延时成本和该交易日的平均等待时间、平均延时成本.

通过统计数据来看, 按照 95% 的置信水平, ETF 交易的延时 21 秒, 可以设置 ETF 的延时成本为: 4 BP.

```python
# ETF的延时成本估计
ETFDelayTime = np.zeros((DayTimeNum, DateNum))# 交易 ETF 的延迟时间
ETFDelayCost = np.zeros((DayTimeNum, DateNum))# 在延迟时间中的收益率
for i, iDT in enumerate(tqdm_notebook(PCF.index, desc='日期 loop')):
    ijMask = np.ones(DayTimeNum+1)
    iIntradayData = FT.readData(factor_names=["lst", "vol"], ids=[TargetID], 
                                start_dt=dt.datetime.combine(iDT.date(), dt.time(0)), end_dt=iDT).iloc[:, :, 0]
    if pd.isnull(iIntradayData["lst"].iloc[0]):
        iIntradayData["lst"].iloc[0] = PreClose[TargetID].iloc[i]
    iIntradayData["lst"].fillna(method="pad", inplace=True)
    iLast, iVol = iIntradayData["lst"].values, iIntradayData["vol"].values
    iTargetNum = PCF["最小申购赎回单位"].iloc[i] / 100
    for j in range(DayTimeNum):
        ijLast, ijVol = iLast[j:], iVol[j:]
        ijMask[j] = 1
        ijMask[j+1:] = (np.cumsum(ijVol, axis=0) < iTargetNum)
        ETFDelayTime[j, i] = np.sum(ijMask[j+1:]) * tms_intv
        ijAvgPrice = np.nansum(ijMask[j:-1] * ijVol * ijLast) / np.nansum(ijMask[j:-1] * ijVol)
        ETFDelayCost[j, i] = ijAvgPrice / ijLast[0] - 1
ETFDelayTime_DayAvg = np.nanmean(ETFDelayTime, axis=0)
ETFDelayCost_DayAvg = np.nanmean(ETFDelayCost, axis=0)

Fig, Axes = plt.subplots(2, 2, figsize=(15, 8))
Axes[0, 0].hist(ETFDelayTime.flatten()[pd.notnull(ETFDelayTime.flatten())], bins="auto")
Axes[0, 0].set_title("交易 ETF 延迟时间(秒)-日内数据")
Axes[0, 1].hist(ETFDelayCost.flatten()[pd.notnull(ETFDelayCost.flatten())], bins="auto")
Axes[0, 1].set_title("交易 ETF 延迟成本-日内数据")
Axes[1, 0].hist(ETFDelayTime_DayAvg[pd.notnull(ETFDelayTime_DayAvg)], bins="auto")
Axes[1, 0].set_title("交易 ETF 延迟时间(秒)-日平均数据")
Axes[1, 1].hist(ETFDelayCost_DayAvg[pd.notnull(ETFDelayCost_DayAvg)], bins="auto")
Axes[1, 1].set_title("交易 ETF 延迟成本-日平均数据");
```

## 现金差额成本

另外, 在 ETF 套利时的现金差额也是一个不确定因素, 下面统计进行溢价套利和折价套利的现金差额引起的成本.

根据每个交易日的日内数据, 首先计算 IOPV. 在每个时点上假设进行折价套利和溢价套利. 对于溢价套利, 计算该交易日实际发生的每单位基金份额对应的现金差额与 IOPV (溢价套利需首先申购 ETF, 单位基金份额投入的资金对应于 IOPV)的比率作为现金差额引起的成本; 对于折价套利, 计算该交易日实际发生的每单位基金份额对应的现金差额与 ETF 交易价格(折价套利需首先买入 ETF, 单位基金份额投入的资金对应于交易价格)的比率作为现金差额引起的成本. 记录每个时点的现金差额成本和该交易日的平均现金差额成本, 最后计算所有现金差额成本以及日平均数据的描述统计量并绘制分布直方图如下.

通过统计数据来看, 按照 95% 的置信水平, 可以设置 ETF 溢价套利的现金差额成本为: 2 BP, 折价套利的现金差额成本为: 5 BP.

```python
# 现金差额成本估计
YJCost = np.zeros((DayTimeNum, DateNum-1))# 进行溢价套利的现金差额引起的成本
ZJCost = np.zeros((DayTimeNum, DateNum-1))# 进行折价套利的现金差额引起的成本
for i, iDT in  enumerate(tqdm_notebook(PCF.index[:-1], desc="日期 loop")):
    iPCFCon = PCFCon.iloc[:, i, :]
    iPCFCon = iPCFCon[pd.notnull(iPCFCon["现金替代标志"])]
    iIntradayData = FT.readData(factor_names=["lst"], ids=[TargetID]+iPCFConIDs, 
                                start_dt=dt.datetime.combine(iDT.date(), dt.time(0)), end_dt=iDT).iloc[0]
    iIntradayData.iloc[0] = iIntradayData.iloc[0].where(pd.notnull(iIntradayData.iloc[0]), PreClose.iloc[i].loc[iIntradayData.columns])
    iIntradayData.fillna(method="pad", inplace=True)
    iIOPV = calcIOPV(iIntradayData.iloc[:, 1:], PCF.iloc[i], PCFCon.iloc[:, i].loc[iPCFConIDs])
    # 溢价套利的现金差额成本
    YJCost[:, i] = (PCF["现金差额"].iloc[i+1] - PCF["预估现金部分"].iloc[i]) / PCF["最小申购赎回单位"].iloc[i] / iIOPV
    # 折价套利的现金差额成本
    ZJCost[:, i] = (PCF["预估现金部分"].iloc[i] - PCF["现金差额"].iloc[i+1]) / PCF["最小申购赎回单位"].iloc[i] / iIntradayData.values[:, 0]
YJCost_DayAvg = np.nanmean(YJCost, axis=0)# 进行溢价套利的现金差额引起的成本的日平均
ZJCost_DayAvg = np.nanmean(ZJCost, axis=0)# 进行折价套利的现金差额引起的成本的日平均

Fig, Axes = plt.subplots(2, 2, figsize=(15, 8))
Axes[0, 0].hist(YJCost.flatten()[pd.notnull(YJCost.flatten())], bins="auto")
Axes[0, 0].set_title("溢价套利的现金差额引起的成本-日内数据")
Axes[0, 1].hist(ZJCost.flatten()[pd.notnull(ZJCost.flatten())], bins="auto")
Axes[0, 1].set_title("折价套利的现金差额引起的成本-日内数据")
Axes[1, 0].hist(YJCost_DayAvg[pd.notnull(YJCost_DayAvg)], bins="auto")
Axes[1, 0].set_title("溢价套利的现金差额引起的成本-日平均数据")
Axes[1, 1].hist(ZJCost_DayAvg[pd.notnull(ZJCost_DayAvg)], bins="auto")
Axes[1, 1].set_title("折价套利的现金差额引起的成本-日平均数据");
```

## 代码说明

**计算 IOPV**

IOPV: ETF 份额参考净值, 根据申购赎回清单中的组合证券的实时市值并加计含预估现金差额计算所得.

IOPV = 最小申购、赎回单位 基金资产净值 / 最小申购、赎回单位基金份数

最小申购、赎回单位基金资产净值 = 申购、赎回清单中各组合证券执行数量 × 最新价（未停牌部分） + ETF 清算单各组合证券执行数量 × 昨收盘价（停牌部分） + 必须现金替代金额 + 预估现金差额

Matlab 函数： calcIOPV.m


**统计 ETF 折溢价率**

折/溢价率 ＝ (ETF 二级市场价格 P - ETF 份额参考净值 IOPV) / ETF 份额参考净值 IOPV × 100%

根据每个交易日的日内数据，计算 IOPV, 而后计算折溢价率，记录每个时点的折溢价率和该交易日的平均折溢价率，最后计算所有折溢价率数据以及日平均数据的描述统计量并绘制分布直方图。

Matlab 脚本： ETFZYRatio.m


**统计 ETF 现金差额引起的成本**

T-1 日现金差额 ＝ T-1 日最小申购、赎回单位对应的基金资产净值 - T-1 日最小申购、赎回单位净值

T-1 日最小申购、赎回单位净值 = ∑ (T-1 日申购赎回清单中必须现金替代的固定替代金额) + ∑ (T-1 日申购赎回清单中允许, 退补或者禁止现金替代的成份证券数量 × T-1 日收盘价)

根据每个交易日的日内数据，首先计算 IOPV。每个时点假设进行折价套利和溢价套利， 对于溢价套利，计算该交易日实际发生的每单位基金份额对应的现金差额与IOPV（溢价套利需首先申购 ETF，单位基金份额投入的资金对应于IOPV）的比率作为现金差额引起的成本， 对于折价套利， 计算该交易日实际发生的每单位基金份额对应的现金差额与 ETF交易价格（折价套利需首先买入 ETF, 单位基金份额投入的资金对应于交易价格）的比率作为现金差额引起的成本。记录每个时点的现金差额成本和该交易日的平均现金差额成本，最后计算所有现金差额成本以及日平均数据的描述统计量并绘制分布直方图。

Matlab 脚本： ETFCashBalanceCost.m


**统计买卖 ETF 引起的冲击成本**

根据每个交易日的日内数据，每个时点假设进行ETF 的买卖， 对于买入 ETF，计算以卖一价成交相对于最新价（对于当前最新价缺失的数据，以第一笔交易的成交价填充，以后有收盘价数据当以前收盘价填充）的变动率作为该时点的冲击成本， 对于卖出 ETF， 计算以买一价成交相对于最新价的变动率作为该时点的冲击成本。记录每个时点的冲击成本和该交易日的平均冲击成本，最后计算所有冲击成本以及日平均数据的描述统计量并绘制分布直方图。

Matlab 脚本： ETFImpactCost.m


**统计买卖最小申赎单位证券组合引起的冲击成本**

根据每个交易日的日内数据，每个时点假设进行ETF 最小申赎单位证券组合的买卖， 对于买入证券组合，计算以卖一价成交证券组合相对于最新价（对于当前最新价缺失的数据，以前收盘价数据填充）的变动率作为该时点的冲击成本， 对于卖出证券组合， 计算以买一价成交相对于最新价的变动率作为该时点的冲击成本。记录每个时点的冲击成本和该交易日的平均冲击成本，最后计算所有冲击成本以及日平均数据的描述统计量并绘制分布直方图。

Matlab 脚本： ETFConImpactCost.m


**统计买卖 ETF 引起的延时成本**

这里以累积交易量超过待交易的 ETF 数量作为完成交易的标志，计算这段时间的价格变动作为延时成本。

根据每个交易日的日内数据，每个时点假设进行最小申赎单位 ETF 的买卖， 以未来累积成交量超过最小申赎单位份额的时间点距离当前时间点的秒数作为本次交易的等待时间，以该段时间中ETF价格变化率作为延时成本。记录每个时点的等待时间、延时成本和该交易日的平均等待时间、平均延时成本，最后计算所有等待时间、延时成本以及日平均数据的描述统计量并绘制分布直方图。

Matlab 脚本： ETFDelayCost.m


**统计买卖最小申赎单位证券组合引起的延时成本**

这里以累积交易量超过待交易的证券数量作为该证券完成交易的标志，计算这段时间的价格变动作为该证券的延时成本，组合的等待时间定义为组合中成交时间最长的证券的等待时间。

根据每个交易日的日内数据，每个时点假设进行最小申赎单位证券组合的买卖， 对于每个证券， 以未来累积成交量超过最小申赎单位中该证券数量的时间点距离当前时间点的秒数作为本次该证券交易的等待时间，以该段时间中证券价格变化率作为延时成本， 所有证券的延时成本按照组合权重加权平均作为组合的延时成本，组合的等待时间为所有证券等待时间的最大值。记录每个时点的等待时间、延时成本和该交易日的平均等待时间、平均延时成本，最后计算所有等待时间、延时成本以及日平均数据的描述统计量并绘制分布直方图。

Matlab 脚本： ETFConDelayCost.m
