## 基本定义

交易型开放式指数基金（简称“交易所交易基金”, Exchange Traded Funds, ETF）, 是指经依法募集的, 投资特定指数所对应组合证券或基金合同约定的其他投资标的的开放式基金, 其基金份额用组合证券、现金或基金合同约定的其他对价进行申购、赎回, 并在证券交易所上市交易.
 
ETF 基金的 特点:
1. 指数型基金, 属于被动投资, 目标是跟踪指数
2. 开放式基金, 可以申购和赎回, 只不过申赎所对应的是标的指数所涵盖的一揽子股票.
3. 上市交易, 和其他上市交易的分级基金、封闭式基金、LOF 一样, 交易价格会随着供给变化与净值产生折溢价

## 发行方式

1. 场内现金认购(网上): 一般是发行期的最后几天或一天, 投资者直接通过交易所的网上系统进行认购即可.
2. 场外现金认购(网下): 涵盖整个发行期, 必须通过签订了代销协议的代销券商进行认购.
3. 网上组合证券认购: 通过交易所的网上系统进行认购.
4. 网下组合证券认购: 投资者付出的对价可以不是现金, 而是持仓的股票, 该股票必须是该 ETF 公告的备选成分股. 而且无论认购期哪一天付出的股票对价, 都是按发行期最后一天的收盘价去计算他的价值然后扣除手续费算出份额给你.

## ETF 份额折算

**建仓期**

基金合同生效后, 基金管理人应逐步调整实际组合直至达到跟踪指数要求, 此过程为 ETF 建仓期, ETF 建仓期不超过 3 个月.

**折算时间**

基金建仓期结束后, 基金管理人通常会以某一选定的日期作为基金份额折算日, 以标的指数的 0.1% 或者 1% 作为份额净值, 对原来的基金份额及其份额净值进行折算, 但折算后基金份额持有人持有的基金份额占基金份额总额的比例不发生变化.

**折算方法**

假设基金管理人确定了基金份额折算日为 T 日, T 日收市后, 基金管理人计算当日的基金资产净值 X 和基金份额总额 Y, T 日标的指数的收盘值为 I, 若以标的指数值的比例 c (1% 或者 0.1%)作为基金份额净值进行折算, 则 T 日的目标基金份额净值为 $c\cdot I$, 折算比例为:
$$  
折算比例=\frac{X/Y}{c\cdot I}
$$
最后结果以四舍五入的方法保留至小数点后 8 位, 折算后基金持有人的持有份额 = 原持有份额 × 折算比例

## 申购赎回

**参与券商**

ETF 的一级市场仅允许参与券商(一级交易商)进行, 参与券商(Participating Dealer, 简称 PD 或者 Authorized Participant, 简称 AP)是指与 ETF 发行人签订了参与契约(Participating Agreement), 可自营或代理非一级交易商及其他投资者的 ETF 申购, 赎回等相关业务的中介结构. 

ETF 的申购和赎回必须通过参与证券商来进行. 当然作为股票经纪商, 参与券商也完全可以代理 ETF 二级市场的买卖交易.

基金管理人在开始申购/赎回业务前会公告参与券商(上交所称为"代理证券公司", 深交所称为"代办证券公司")名单.


### 申赎规则
 
| 规则 | 说明 |
| ---- | ---- |
| 申购赎回开始日 | 基金份额折算日之后可开始办理申购, 基金合同生效日后不超过 3 个月的时间起开始办理赎回 |
| 申购赎回时间 | 交易所的开市时间:  周一至周五的上午9:30-11:30和下午1:00-3:00（节假日除外） |
| 申购赎回地点 | 参与券商或者参与券商提供的其他方式 |
| 申购赎回方式 | 在交易日的交易时间通过参与参与券商委托下单 |
| 申购赎回条件 | 需开立证券账户 |
| 申购赎回单位 | 申购赎回清单中规定的最小申购赎回单位或其整数倍 |
| 申购赎回费用 | 无印花税, 佣金不超过申购、赎回份额的 0.5％ |
| 申购赎回限制 | 申购赎回申请提交后不能撤销 |


**申赎原则**

1. 场内申购/赎回 ETF 采用份额申购, 份额赎回的方式, 即申购/赎回均以份额申请. 场内申购/赎回对价包括组合证券, 现金替代, 现金差额以及其他对价.
2. 部分 ETF 基金管理人提供场外申购/赎回模式, 场外申购/赎回采用金额申购, 份额赎回的方式, 即申购以金额申请, 赎回以份额申请. 场外申购/赎回对价均为现金.


**回转交易**

1. 当日申购的 ETF 份额, 实时到账, 同日可以卖出, 但不得赎回; 其中, 对标的指数为跨市场指数的交易所交易基金, 当日申购且同日未卖出的基金份额, 清算交收完成后方可卖出和赎回; 
2. 当日赎回的 ETF 份额, 证券实时到账, 同日可以卖出, 但不得用于申购基金份额; 
3. 当日买入的 ETF 份额, 当日可以赎回, 但不得卖出; 
4. 当日买入的证券, 同日可以用于申购 ETF 份额, ETF 份额即时可交易; 

总之, 单独在一级市场申赎和单独在二级市场买卖都是 T+1, 但是一二级市场结合起来, 申购卖出、买入赎回可以是 T+0.


**申购赎回的清算交收与登记**

投资者在 T 日申购/赎回成功后, 登记结算机构在 T 日收市后为投资者办理基金份额与组合证券的清算交收以及现金替代的清算, 在 T+1 日办理现金替代等的交收以及现金差额的清算, 在 T+2 日办理现金差额的交收, 并将结果发送给参与券商, 基金管理人和基金托管人.

中国证券登记结算公司针对 ETF 实行 DVP 结算制度.

当投资者做申购赎回业务时, 在交易所前端对相关证券份额进行实时过户的情况下, 中央登记结算系统并未实时过户(将 ETF 以及证券篮子作为待处分证券), 而是在 T+1 日检查资金无透支的情况下, 才真正过户. 若出现会员透支, T+1 日交易被限制. 透支额须 T+2 日补齐, 若 T+2 日未能补齐, 中登公司将在 T+3 日变现待处分证券以补足欠款. 如果是券商自营账户的话, 还可以进一步通过卖出自营账户中的其他证券的办法来强制补足.

## 申购赎回清单

ETF 申购和赎回的对价均是一揽子股票, 这样交易成本最低, 跟踪误差最小. 这一揽子股票具体是什么是由基金公司确定的, 具体的内容在每个交易日开盘之前公布的“申购赎回清单”中.

**一级市场基金代码**: 指申购赎回时使用的基金代码.

**最小申购、赎回单位**: 一次申购赎回的最小基金份额.

**最小申购、赎回单位净值**: 基于 T-1 日的收盘价计算的最小申购、赎回单位对应的基金资产净值.

**基金份额净值**: 基于 T-1 日的收盘价计算的单位份额对应的基金资产净值

**申购上限**: 本日申购基金时不能超过的最大基金份额.

**赎回上限**: 本日赎回基金时不能超过的最大基金份额.

**申购赎回的允许情况**: 本日申购赎回是否允许.


### 现金替代

理想情况下, ETF 是"以物易物", 但在成分股停牌或涨跌停等特殊情况下, 若严格执行“物物交换”, 将会使投资人买不到成分股去申购 ETF, 影响效率. 为避免这种情况, ETF 基金通常会允许现金替代, 即允许使用现金作为全部或部分该成份股票的替代, 对于大部分成分股, 即使处于正常交易时期, 基金公司也是允许现金替代的, 现金替代有四种类型:
1. **禁止现金替代**: 表示申购或赎回本基金份额时必须使用股票, 不接受现金替代; 例如有的股票长期停牌, 基金经理很难拿着你的现金从二级市场购入股票, 如果现金替代量大而且该股票权重大的话, 可能导致跟踪误差偏大. 其实禁止现金替代体现的是 ETF 的本意.
2. **允许现金替代**: 表示申购本基金份额时可以先使用股票, 在股票不足的情况下可以使用现金替代持有不足的股票; 但在赎回基金份额时, 该成份股不允许使用现金作为替代.
3. **必须现金替代**: 表示在申购或赎回本基金份额时只能使用现金而不接受股票. 这种设置一般适用于标的指数调整, 即将被剔除的成份证券, 或者是政策规定基金管理人不能买卖的成份证券, 又或者基金公司出于保护持有人利益原则等原因认为有必要实行必须现金替代的成份股. 对于必须现金替代的股票, 基金管理人在申购赎回清单中公告替代的一定数量的现金, 即**固定替代金额**(见成份股信息内容的最后一列), 固定替代金额 = T 日预计开盘价 × 申赎清单中的证券数量
4. **退补现金替代**: 表示在申购、赎回本基金份额时只能使用现金, 待基金公司代为买卖之后, 根据实际成本进行多退少补. 


**现金替代溢价比例(现金替代保证金率)**

对于使用现金替代(类型为允许现金替代和退补现金替代)的股票, 由于基金经理需在 T+1 和 T+2 日去购买对应的股票, 会产生交易费用, 而且购买的成本也有可能与 T 日不同, 为此基金公司会向申购者收取一定的现金溢价, 并在 T+2 日盘后进行交收, 多退少补. 这个溢价称为: **现金替代溢价比例** 或者 **现金替代保证金率** .
1. 对于类型为允许现金替代的股票, 申购 ETF 时, 替代的现金为: (1+现金替代溢价比例) × 证券参考价格 × 替代证券数量; 赎回 ETF 时得到申赎清单中给出的相应数量的股票, 不退回现金. 其中, 申购时的证券参考价格确定原则为: 
    * 证券正常交易时，采用最新成交价。
    * 证券正常交易中出现涨停/跌停时, 采用涨停/跌停价格。
    * 证券停牌且当日有成交时, 采用最新成交价。
    * 证券停牌且当日无成交时, 采用经除权调整后的前一交易日收盘价
2. 对于类型为退补现金替代的股票, 申购 ETF 时其替代的股票数量固定为申赎清单中给出的股票数量, 所以替代的现金为: (1+现金替代溢价比例) × T 日预计开盘价 × 申赎清单中的股票数量 = (1+现金替代溢价比例) × 申赎清单中的固定替代金额; 赎回 ETF 时将得到基金公司卖出申赎清单中对应数量的股票而所得的现金.
3. 对于类型为必须现金替代的股票, 申购 ETF 时将以申赎清单中的固定替代金额直接替代股票; 赎回时得到的也是申赎清单中给出的固定替代金额.


**现金替代比例上限**

为有效控制基金的跟踪偏离度和跟踪误差, 基金管理人可以规定投资者使用允许现金替代的比例合计不得超过申购基金份额资产净值的一定比例, 25%, 50%都有, 例如 上图中的 A50 ETF, 其现金替代比例的上限就是 50%. 

现金替代比例 = ∑ (第 i 只替代证券的数量 × 证券参考价格) ÷ (最小申购、赎回单位 × 最新基金份额参考净值 IOPV) × 100%


**现金替代处理程序**

由于基金公司会使用替代的现金来购买相应证券, 就有可能产生现金替代退补款, 指投资者支付的现金替代与基金购入被替代成份证券的成本及相关费用的差额. 若现金替代大于基金购入被替代成份证券的成本及相关费用, 则基金需向投资者退还差额, 若现金替代小于基金购入被替代成份证券的成本及相关费用, 则投资者须向基金补缴差额. 

T 日, 基金管理人在申赎清单中公布现金替代溢价比例, 并据此收取替代金额; 在 T 日后被替代的成份证券有正常交易的 2 个交易日(简称 T+2 日)内, 基金管理人将已收到的替代金额买入被替代的部分证券; T+2 日终, 若已购入全部被替代的证券, 则以替代金额与所购入的部分被替代证券实际购入成本（包括买入价格与交易费用）加上按照 T+2 日收盘价计算的未购入的部分被替代证券价值的差额，确定基金应退还申购投资者或申购投资者应补交的款项。

特例情况: 若在 T 日后, 上交所正常交易日已达到 20 日而该证券正常交易日低于 2 日, 则以替代金额与所购入的部分被替代证券实际购入成本（包括买入价格与交易费用）加上按照最近一次收盘价计算的未购入的部分被替代证券价值的差额, 确定基金应退还申购投资者或申购投资者应补交的款项。

T+2 日后第一个交易日(若在特例情况下, 则为 T 日后第 21 个交易日), 基金管理人将应退款或补款的明细以及汇总数据发送给相关申购赎回代理券商和基金托管人, 相关款项的清算交收于此后 3 个工作日内完成. 结算公司对于申购产生的现金替代退补款实行非担保交收. 基金管理人负责计算投资者应收（付）的现金替代退补款, 并将其传送到结算公司. 结算公司根据每日收到的基金管理人数据进行现金替代退补款的清算, 并于次日进行交收. 


### 现金差额

理想情况下, ETF 是"以物易物", 申购赎回清单在最理想的状态下应该是和基金资产完全一样的一系列股票数量的组合和现金（基金资产中也会有少量现金）, 但是在现实中为了基金能够顺利运作, 往往会产生一些差异, 其中一个原因为: 清单里的股票数量都是 100 的整数倍, 也就是我们平常熟知的“一手”的概念. 而基金资产中的股票有可能由于送配股或交易等原因, 其股数并不是 100 的整数倍, 和清单中取整的股数是有差异的, 这样一来基金资产和清单中的对价就会有差别. 这个差额就称为: 现金差额. 在 T 日已经可以得到 T-1 日的收盘价, 因此可以准确计算出 T-1 日的现金差额, 计算公式如下:

T-1 日最小申购、赎回单位净值 = ∑ (T-1 日申购赎回清单中必须现金替代的固定替代金额) + ∑ (T-1 日申购赎回清单中允许, 退补或者禁止现金替代的成份证券数量 × T-1 日收盘价)

T-1 日现金差额 ＝ T-1 日最小申购、赎回单位对应的基金资产净值 - T-1 日最小申购、赎回单位净值. 


**预估现金差额**

指由基金管理人计算并在 T 日开盘前在申购赎回清单中公布的 T 日现金差额预估值(对应图中: 最小申购、赎回单位的预估现金部分). 

在 T 日, 每个最小申赎单位的基金资产净值和申赎清单中最小申赎单位的对价的现金差额在 T 日盘后清算之前是不确定的, 但投资者要进行申赎的话, 为了对价公平, 要把这一差额补上(该差额为正)或基金公司退回(该差额为负), 所以就需要在 T 日开盘前, 就有一个预计的差额来进行指导, 这就是所谓的预估现金差额. 其计算公式为:

T 日最小申购、赎回单位预估净值 = ∑ (T 日申购赎回清单中必须现金替代的固定替代金额) + ∑ (T 日申购赎回清单中允许, 退补或者禁止现金替代的成份证券数量 × T 日预计开盘价)

T 日预估现金差额 ＝ T-1 日最小申购、赎回单位对应的基金资产净值 - T 日最小申购、赎回单位预估净值

其中, T 日预计开盘价主要根据上交所提供的标的指数成份证券的预计开盘价确定. 另外, 若 T 日为基金分红除息日, 则计算公式中的"T-1 日最小申购、赎回单位对应的基金资产净值"需扣减相应的收益分配数额.

这样一来, 如果预估现金差额是正数, 那么投资者在 T 日按照清单申购的时候, 不仅要付出清单上的对价, 还要加上这部分预估现金差额. 如果投资者赎回, 得到的对价除了申赎清单上的东西, 还会得到一部分预估现金差额. 如果预估现金差额是负数, 那么申购者则会得到预估现金差额的退款, 而赎回者则要补缴这部分预估现金. 

预估现金差额由代办证券公司预先冻结, 但这部分现金只是预估的, T 日实际上的现金差额很有可能和预估的数不一样, 甚至连正负都相反, 这个数字会在 T+1 日申赎清单中一同公布. 真实差额和预估差额有差异时进行多退少补, T+1 日根据实际现金差额进行清算, T+2 日进行交收并且差额多退少补.


### 申购赎回对价

投资者进行申购 ETF 时, 提交的对价组成部分如下:
1. 股票, 包括: 类型为禁止现金替代的股票(申赎清单中规定的固定数量), 类型为允许现金替代的股票(投资者控制数量)
2. 替代现金, 包括: 类型为必须现金替代的股票替代金(申赎清单中规定的固定替代金额), 类型为允许现金替代的股票替代数量(投资者控制数量) × 证券参考价格 × (1+现金替代溢价比率), 类型为退补现金替代的股票替代金(申赎清单中规定的固定替代金额) × (1+现金替代溢价比率)
3. 预估现金差额, 为正表示投资者支付, 为负表示投资者获得

<img src="./images/ETF基金申购图示.png" title="ETF 基金申购图示">

投资者进行赎回 ETF 时, 得到的对价组成部分如下:
1. 股票, 包括: 类型为禁止现金替代的股票(申赎清单中规定的固定数量), 类型为允许现金替代的股票(申赎清单中规定的固定数量)
2. 替代现金, 包括: 类型为必须现金替代的股票替代金(申赎清单中规定的固定替代金额), 类型为退补现金替代的股票(申赎清单中规定的固定数量)卖出所得的现金
3. 预估现金差额, 为正表示投资者获得, 为负表示投资者支付


### ETF 份额参考净值(IOPV)

**IOPV**: ETF 份额参考净值, 交易所在交易时间内根据申购赎回清单中的组合证券的实时市值并加计含预估现金差额计算所得, 每 15 秒发布一次.

IOPV = 最小申购、赎回单位 基金资产净值 / 最小申购、赎回单位基金份数

最小申购、赎回单位基金资产净值 = 申购、赎回清单中各组合证券执行数量 × 最新价（未停牌部分） + ETF 清算单各组合证券执行数量 × 昨收盘价（停牌部分） + 必须现金替代金额 + 预估现金差额

### 上交所跨市场 ETF

上交所跨市场 ETF 主要采取“本交易所实物+非本交易所现金替代”的方案，即非本交易所股票由基金管理人代买，这种方式的优点是可以实行 T+0，但缺点是投资者不能实时锁定另外一个交易所股票组合的构建成本，对实时风险控制也提出了较高的要求。深市股票补券原则: 时间优先、实时申报
1. 当日申购的 ETF 份额, 实时到账, 同日可以卖出, 但不得赎回; 当日申购且同日未卖出的基金份额, T+2 日清算交收完成后方可卖出和赎回; 
2. 当日赎回的 ETF 份额, 证券实时到账, 同日可以卖出, 但不得用于申购基金份额; 
3. 当日买入的 ETF 份额, 同日可以赎回, 但不得卖出; 
4. 当日买入的证券, 同日可以用于申购 ETF 份额, ETF 份额即时可交易; 

<img src="./images/上交所跨市场ETF申购流程.png" title="上交所跨市场 ETF 申购流程">

1. T 日投资者持有上海成份股，同时预留深圳成份股权重相当（券商和客户可根据公告用申购赎回清单计算）的现金并提出申购申请。
2. 券商根据篮子清单检查投资者上海成份股以及资金是否足额，足额方报上海证券交易所。
3. 上交所根据基金管理公司的篮子清单确认申购是否有效，如符合清单，将申购成功确认返回券商和基金公司。
4. 上交所实时完成投资者与基金之间的上海成份股和 ETF 份额的可用数量的实时增减，券商为投资者增加 ETF 可用数量，投资者申购的 ETF 份额当日可卖出。
5. 上交所将实时申购数据发送给 ETF 基金管理人。
6. T 日交易时段，基金公司根据 T 日申购“时间优先、实时申报”原则，买入相对应的深圳成份股。
7. T 日收市后，沪深交易所分别将当日申购数据和成份股买入数据发送给中国结算沪深分公司。
8. T 日日终，中国结算沪深分公司进行 ETF 申购及组合证券买入的清算交收。
9. 基金管理人向托管银行发出划款交收指令。
10. 基金托管银行将现金替代款从中国结算上海分公司划到中国结算深圳分公司，完成基金管理人在深圳买入股票的资金交收.
11. T+2 日日终，若已购入全部被替代的证券，则以替代金额与被替代证券的实际购入成本（包括买入价格与交易费用）的差额，确定基金应退还申购投资者或申购投资者应补交的款项；若未能购入全部被替代的证券，则以替代金额与所购入的部分被替代证券实际购入成本（包括买入价格与交易费用）加上按照 T+2 日收盘价计算的未购入的部分被替代证券价值的差额，确定基金应退还申购投资者或申购投资者应补交的款项。 

<img src="./images/上交所跨市场ETF赎回流程.png" title="上交所跨市场 ETF 赎回流程">

1. T 日投资者持有跨市场 ETF 份额，并提出赎回申请。
2. 券商检查投资者 ETF 份额是否足额，足额方报上海证券交易所。
3. 上交所根据基金管理公司持有的组合证券和投资者的 ETF 份额数量，确认赎回是否有效，如符合要求，将赎回成功确认返回券商和基金公司。
4. 上交所实时完成投资者与基金之间的上海成份股和 ETF 份额的可用数量的实时增减，券商为投资者增加上海成份股的可用数量，投资者赎回的上海成份股当日可卖出。
5. 上交所将实时赎回数据发送给 ETF 基金管理人。
6. T 日交易时段，基金公司根据 T 日赎回“时间优先、实时申报”原则，卖出相对应的深圳成份股。
7. T 日收市后，沪深交易所分别将当日赎回数据和成份股交易数据发送给中国结算沪深分公司。
8. T 日日终，中国结算沪深分公司进行 ETF 赎回及组合证券卖出的清算交收；T+1 日完成相关的资金交收。
9. T+2 日，基金管理人通过结算公司的代收代付平台，与代办券商完成 ETF 赎回涉及深圳现金替代款的交收。


### 深交所跨市场 ETF

深交所跨市场 ETF 采用国际通用模式, 实行场外实物申购赎回、非担保交收, 由于采用这种场外申赎模式, 跨市场 ETF 的买卖申赎原则与单市场 ETF 不同:
1. T 日申购的 ETF 份额, T+2 日可卖出或赎回.
2. T 日买入的 ETF 份额, T+1 日可赎回或卖出.
3. T 日买入的组合证券, T 日可申购 ETF 份额.
4. T 日赎回 ETF 份额, 得到的组合证券, T+2 日可卖出或用于申购.

深交所跨市场 ETF 的运行模式非 T+0.

## 跨境 ETF

跨境 ETF 是指由境内基金公司发起设立的, 通过 QDII 形式, 以复制的方法追踪境外证券指数, 投资特定境外证券指数所对应的组合证券.

跨境 ETF 采用份额申购, 份额赎回的方式, 申购/赎回对价为"全现金替代", 申购/赎回结算为"逐笔全额结算". 

1. 当日申购的 ETF 份额, T+2 日清算交收完成后方可卖出和赎回; 
2. 当日买入的 ETF 份额, 同日可以赎回, 但不得卖出; 

另外, 由于跨境涉及外汇管理, 如果投资者当日申购总额, 赎回总额超出基金管理人设定限额的, 超出额度的申购, 赎回申报为无效申报.

## 债券 ETF

债券 ETF 是以债券类指数为跟踪标的交易型证券投资基金. 

1. 当日申购的 ETF 份额, 实时到账, 同日可以卖出或者赎回; 
2. 当日赎回的 ETF 份额, 债券实时到账, 同日可以卖出或者进行质押式回购; 
3. 当日买入的 ETF 份额, 同日可以赎回, 但不得卖出; 
4. 当日买入的债券, 同日可以用于申购 ETF 份额, ETF 份额即时可交易;

## 商品 ETF

商品 ETF 一般可分为实物支持 ETF 与非实物支持 ETF 两类, 实物支持 ETF 直接持有实物资产, 主要用于贵金属领域. 非实物支持 ETF 主要投资于大宗商品期货等金融衍生品, 主要覆盖工业金属, 能源, 农产品等资产类型.

黄金 ETF 是指由国内基金管理人发起设立的, 追踪国内黄金现货价格(基础资产为上海黄金交易所的黄金实盘现货合约), 并在交易所上市交易的 ETF. 黄金 ETF 采用现金申赎或者黄金现货实盘合约申赎两种模式:
1. 现金申赎: 对价为全现金替代, 基金管理人在金交所内代理买卖黄金合约, 现金申赎结算为逐笔全额结算, T 日现金申购, T+1 日清算交收后可卖、可赎回, T 日买入的份额, 同日可以赎回, 当日赎回的份额实时记减.
2. 实物申赎: 对价为金交所黄金现货合约, T 日申购, T 日可卖、可赎回; T 日买入份额, T+1 日可黄金赎回, T 日赎回的黄金, T 日可卖. 需同时使用证券账户和黄金账户, 且确保两个账户属于同一投资者. 办理黄金现货合约申购、赎回前, 投资者必须于基金管理人处进行备案.

黄金 ETF 存在 10% 的涨跌幅限制, 而基金投资的黄金现货涨跌幅上限为 30%. 因此, 基金净值涨跌幅可能超过 10%, 导致黄金 ETF 在二级市场上出现折溢价.

2013 年 7 月, 华安基金管理公司和国泰基金管理公司开发的我国第一只黄金 ETF 产品在上海证券交易所上市交易. 该基金主要投资于上海黄金交易所挂牌交易的纯度高于(含) 99.95% 的黄金现货合约. 同时, 通过黄金租赁业务, 黄金互换业务等覆盖部分基金运营管理费用, 以期让投资者能够完全享受到金价的投资收益. 该产品每份基金份额含 0.01 克黄金, 最小交易单位为 1 手(100 份, 对应 1 克黄金).

## 投资策略

**日内趋势策略**

由于 ETF 存在独特的一, 二级市场联动机制, 所以投资者可以通过类似折溢价套利的原理实现日内 ETF 的回转交易. 日内趋势交易策略的成功与否主要取决于投资者对于日内行情的正确研判.

假设当前投资者对未来看涨, 可以通过以下两种方式进行交易:
1. 投资者在低点买入一揽子股票, 申购成 ETF 基金份额, 持有至盘中反弹后抛出 ETF 基金份额
2. 投资者在低点买入一定数量的 ETF 份额, 赎回一揽子股票组合, 持有至盘中反弹后将股票组合在市场中卖出.

投资者最终选择以何种方式进行回转交易, 主要取决于两个因素: 一是成分股以及 ETF 二级市场的流动性; 二是买入时点的折价还是溢价以及其相应幅度.

**事件策略**

成份股相关的公司事件有可能给投资者带来交易机会, 这些事件主要包括: 增发, 配股, 股改等. 公司事件交易主要针对停牌股, 即在成分股因公告, 股改, 增发等事件而导致的停牌期间, 投资者预估其价格复牌后会有大幅涨跌的可能. 在 ETF 二级市场价格还没有充分反映其复牌后价格预期之前, 投资者可以进行溢价(申购方向)或者折价(赎回方向)套利操作, 获取收益.

![ETF 事件交易](./images/ETF事件交易.png)

公司事件交易成功与否主要取决于 ETF 管理人对于停牌股票的标志处理方式, 在必须现金替代模式下该交易策略不具备可操作性.

**套利策略**

ETF 套利可分为无风险套利及风险套利。

在无风险套利的情况下， ETF 相关产品的理论价格不等于市场价格时，套利机会便会产生。 ETF 衍生金融工具包括 ETF 期货、 ETF 所追踪指数的期货、 ETF 期权、 ETF 所追踪指数的期权、 ETF 认购（售）权证、 ETF 连结票据等。这些商品之间的价格具有关联性，当产品价格之间产生偏差，便可藉相关产品的一买一卖来消弭错价。由下图错综复杂之价格连结所产生的可能无风险套利机会约 36 种，但其中部分套利在实际操作时可能无法进行。
