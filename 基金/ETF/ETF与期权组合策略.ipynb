# ETF 与期权组合策略

## 摘要

## 理论模型

### 无风险套利

期权平价理论(Put-Call Parity)的基本原理是相同标的、相同到期日、相同执行价格的认购认沽期权满足下述等式关系:
$$
C_t + Ke^{-r(T-t)} = S_t + P_t
$$
其中, $S_t,C_t,P_t$ 分别为现货, 认购期权和认股期权的市场价格; $K$ 为认购认沽期权的执行价格; $r$ 为无风险利率(年化, 连续复利计算); $T-t$ 为年化的到期时间.

为直观的理解期权平价套利的原理, 我们可以假想构造两个投资组合 A 和 B，其中组合 A 由认购期权和现金 :math:`Ke^{-r(T-t)}` 组成， 组合 B 由现货和认沽期权组成，根据期权的支付特点，到期时两个组合的现金流是相同的，所以根据无套利原理, 任何时刻这两个组合的市场价格应该是相同的，若存在差异则可能存在套利机会。

直观上讲, 认购期权、认沽期权和标的现货所组成的这种平价关系, 意味着认购期权和认沽期权所内含的隐含波动率应该一致, 若隐含波动率相差过大, 则意味着存在套利机会.

**多头套利(正向套利)**: 当 $C_t + Ke^{-r(T-t)} > S_t + P_t$ 时, 卖出一张认购期权, 同时买入一张认沽期权和对应一合约乘数的现货, 持有至到期交割.

**空头套利(反向套利)**: 当 $C_t + Ke^{-r(T-t)} < S_t + P_t$ 时, 买入一张认购期权, 同时卖出一张看跌期权, 并融券卖出一合约乘数的现货, 持有至到期交割.

考虑到存在交易费用、保证金占用以及融资融券成本, 假设相关参数如下:
* 设定相关费用为 $f_t$ (包括现货交易费用, 期权交易费用, 期权交割费用等),
* 融资融券利率为 $r_l$,
* 融券的保证金比例为 $m$,
* 卖出认购期权保证金占用 $M_t^C$,
* 卖出认沽期权保证金占用 $M_t^P$,
* 无风险收益率(年化, 连续复利计算)为 $r$
* 投资者的最低投资收益率为 $r_e$

则多头套利(正向套利)的开仓条件为: $\frac{C_t e^{r(T-t)}+K-P_t-S_t-f_t-f_T}{S_t+P_t+M^C_t+f_t}>r_e$

即期初的资金投入是初始成本:
* 买入现货支出 $S_t$,
* 买入认沽期权支出 $P_t$,
* 卖出认购期权的保证金支出 $M^C_t$,
* 相关费用支出 $f_t$,
* 总计 $S_t+P_t+M^C_t+f_t$.

期末的产生的资金收入为:
* 现货卖出收入 $S_T$
* 认沽期权行权收入 $P_T=\left(K-S_T\right)^+$
* 认购期权行权支出 $C_T=\left(S_T-K\right)^+$
* 认购期权保证金收回 $M^C_t$
* t 时认购期权卖出收入 $C_t e^{r(T-t)}$
* 相关费用支出 $f_T$
* 总计 $S_T+\left(K-S_T\right)^+ -\left(S_T-K\right)^+M^C_t+C_t e^{r(T-t)}-f_T = C_t e^{r(T-t)}+K+M^C_t-f_T$

多头套利(正向套利)期间收益为: $C_t e^{r(T-t)}+K-S_t-P_t-f_t-f_T$.

空头套利(反向套利)的开仓条件为: $\frac{P_t e^{r(T-t)}+S_t e^{r(T-t)}-C_t-K-S_t r_l (T-t)-f_t-f_T}{C_t+K e^{-r(T-t)}+M^P_t+mS_t+f_t}>r_e$.

即期初的资金投入是初始成本: 
* 买入认购期权支出 $C_t$,
* 现金支出 $Ke^{-r(T-t)}$,
* 融券保证金支出 $mS_t$,
* 卖出认沽期权的保证金支出 $M^P_t$,
* 相关费用支出 $f_t$,
* 总计 $C_t+Ke^{-r(T-t)}+mS_t+M^P_t+f_t$.

期末的产生的资金收入为: 
* 认购期权行权收入 $C_T=\left(S_T-K\right)^+$
* 现金收入 $K$
* 融券卖出收入 $S_t e^{r(T-t)}-S_T$
* 融券保证金收回 $mS_t$
* 融券利息支出 $S_t r_l (T-t)$
* 认沽期权行权支出 $P_T=\left(K-S_T\right)^+$
* 认沽期权保证金收回 $M^P_t$
* t 时认沽期权卖出收入 $P_t e^{r(T-t)}$
* 相关费用支出 $f_T$
* 总计 $\left(S_T-K\right)^+ +K+S_t e^{r(T-t)}-S_T+mS_t-S_t r_l (T-t)-\left(K-S_T\right)^++M^P_t+P_t e^{r(T-t)}-f_T = P_t e^{r(T-t)}+S_t e^{r(T-t)}+mS_t-S_t r_l (T-t)+M^P_t-f_T$

期间收益为: $P_t e^{r(T-t)}+S_t e^{r(T-t)}-C_t-K-S_t r_l (T-t)-f_t-f_T$

持仓期间应该逐日监测持有期权的保证金仓位水平控制持仓风险, 同时实时监测头寸的盈利状况.

平仓阶段分两种情形: 如果未到交割日但持有期间组合的盈利已经大于预期的收益率则选择实时平仓获利, 否则持有期权至临近交割日等待期权价值回归理论价值后平仓或交割.

在实施平价套利的过程中, 值得注意的两个问题是:

1. 平价套利策略收益的最晚确定日期: 根据现行上交所的规定, 设定期权的交割日为 T 日,  则期权交割后认购期权的义务方或认沽期权的权利方将 T+1 日获得行权资金, 所以多头套利的收益最晚将在 T+1 日确认; 另一方面, 认购期权的权利方或认沽期权的义务方将在 T+1 日交收获得现券, T+2 日 ETF 可以卖出最终确认收益;
2. 平价套利策略受 50 ETF 除息的影响: 由于标的现货在出现分红后一张期权所对应的 50 ETF 数量会改变, 则原有开仓时所准备的 50 ETF 会有所不足, 为避免该情况出现应该选择红利再投资的基金分红方式.
