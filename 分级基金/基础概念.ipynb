# 基础概念

本部分主要探讨分级基金的相关研究. 


## 基本概念

分级基金是指通过事先约定基金的风险收益分配，将母基金份额分为预期风险收益不同的子份额，并可将其中部分或全部类别份额上市交易的结构化证券投资基金。

上证分级基金的母份额为 LOF 基金（可由基金管理人选择是否上市交易），可通过上海证券交易所场内证券经营机构申购、赎回，可实时分拆为子份额；子份额在上交所场内上市交易，可实时合并为母份额，但子份额不可单独申购、赎回。

收益约定: A 份额获得按照基金合同约定的年收益率, 基金收益首先分配给 A 份额, 剩余的收益归 B 份额.

基金管理人将母基金整体合并运作, 进行股票或者债券投资, 按收益分配机制分成 A、B 两类子份额.

## 申购赎回

**申赎原则** 
1、“未知价”原则: 即母基金份额申购、赎回价格以申请当日收市后计算的母基金份额净值为基准进行计算;
2、“金额申购、份额赎回”原则: 即母基金份额申购以金额申请，赎回以份额申请。 

分级基金母基金份额可以通过场外、场内按净值办理申购和赎回, A 份额和 B 份额不能单独申购赎回，只能在场内买卖。

**折算机制**

**定期折算**: 阶段性兑现收益, 每个计息周期结束后将 A 份额的约定收益（净值大于 1 的部分）折算为基础份额，投资者可以通过赎回份额实现收益，B 份额不参与定期折算。

![分级基金定期折算](./images/分级基金定期折算.png)

折算基准日当天 A 份额和 B 份额是正常交易的，但是母基金当天不能申购和赎回.

**不定期折算**: 调节 B 类杠杆范围

向上不定期折算: 当母基金份额净值大于某个值（一般 1.5 或者 2）时，将母基金份额、A 份额和 B 份额净值大于 1 的部分折算为母基金，三者的净值调整为 1 元，实现收益并提高 B 份额的杠杆。

![分级基金向上不定期折算](./images/分级基金向上不定期折算.png)

向下不定期折算: 当 B 份额净值小于某个值（一般为 0.25）时，通过份额再平衡，将母份额、A 份额和 B 份额净值归 1.

![分级基金向下不定期折算](./images/分级基金向下不定期折算.png)

**初始杠杆**

初始杠杆 = (A 份额数量 + B 份额数量) / B 份额数量

**净值杠杆**

母基金份额净值每变动 1 个单位，B 份额净值的变动率

净值杠杆 = 初始杠杆 × 母基金净值 / B 份额净值

**价格杠杆**

母基金份额净值没变动 1 个单位, B 份额价格的变动率

价格杠杆 = 初始杠杆 × 母基金份额净值 / B 份额价格
