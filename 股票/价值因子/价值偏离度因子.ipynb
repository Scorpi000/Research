# 价值偏离度因子

## 基本逻辑

价值投资的鼻祖当属格雷厄姆，在格雷厄姆的价值投资理念中，安全边际被认为是至关重要的。巴菲特是格雷厄姆价值投资理念的集大成者，巴菲特在其投资生涯中也一直将安全边际作为重要的投资准则，他晚年的时候感叹安全边际原则仍然非常正确，非常有效，永远是投资成功的基石.

我把投资成功、永不亏损的秘密精练成四个字的座右铭--安全边际
——本杰明· 格雷厄姆

安全边际本质上就是股票价格低于其内在价值的部分，衡量了上市公司被低估的程度.

据此, 可以定义估值偏离度为:
$$
D = M - V
$$
其中, $M$ 为上市公司总市值, $V$ 为上市公司内在价值, $D$ 为估值偏离度. 从上式可以看到，得到公司的估值偏离度的关键在于计算公司的内在价值，而公司的内在价值无法直接观测，因此我们只能用各种估值手段和方法对其内在价值进行估计。由于内在价值估算的方式的选择是见仁见智的，所得到的估值偏离度因子也各不相同。

理论上对内在价值的估计越精确，所构建的估值偏离度指标也越有效。

下面采取一种对公司内在价值可能更为精确的估计方式——剩余收益估值模型(Residual Income Valuation Model, RIV)

剩余收益估值模型又被称为 EBO 模型，最早是由 Edwards 和 Bell 于 1961 年提出来的，1995 年美国学者 Ohlson 对这个方法进行了系统的阐述。剩余收益模型使用公司权益的账面价值和预期剩余收益的现值来表示股票的内在价值，如下式所示:
$$
V_0 = BV_0+\sum\limits_{t=1}^{\infty}\frac{RI_t}{(1+r)^t}
$$
其中, $V_0$ 为公司当前内在价值, $BV_0$ 为公司当前的账面价值, $RI_t$ 为公司未来的剩
余收益, 剩余收益为公司的净利润与股东所要求的报酬之差, 即: 
$$
RI_{t+1} = NI_{t+1} - rBV_t
$$
假设净利润和账面价值均以一个固定比率增长, 分别记为$g_1, g_2$, 则有:
$$
V_0=BV_0+\sum\limits_{t=1}^\infty\frac{NI_t-rBV_{t-1}}{(1+r)^t}=\left(1-\frac{r}{r-g_2}\right)BV_0+\frac{1+g_1}{r-g_1}NI_0
$$
从而总市值可以表达为:
$$
M_0=\left(1-\frac{r}{r-g_2}\right)BV_0+\frac{1+g_1}{r-g_1}NI_0+D_0
$$
即公司的总市值可以分解为账面价值, 净利润, 以及估值偏离度的组合.

由于个股总市值数量级相差较大，数据呈现明显的右偏性，为了降低数据的右偏性对回归参数估计的准确性所的影响，避免残差项的异方差问题，我们采取对数回归的形式:
$$
\ln M_{it}=\alpha_t+\beta_{1t}\ln BV_{it}+\beta_{2t}\ln NI_{it}+\varepsilon_{it}
$$
考虑到公司净利润可能为负, 以及公司财务杠杆对总市值的影响，对上式进行一些修改, 加入了净利润哑变量以及财务杠杆因子:
$$
\ln M_{it}=\alpha_t+\beta_{1t}\ln BV_{it}+\beta_{2t}\ln|NI_{it}|I_{NI_{it}>0}+\beta_{3t}\ln|NI_{it}|I_{NI_{it}<0}+\beta_{4t}LEV_{it}+\varepsilon_{it}
$$
其中, $I_{.}$ 是示性函数, $LEV_{it}$ 是财务杠杆 = 负债合计 / 股东权益合计(不含少数股东权益)

理论上, 上市公司的估值偏离度因子的因子值越大, 表明该公司被高估的程度越大, 未来股票收益相对较低; 而因子值越小, 表明公司被
低估的程度越大, 未来股票收益相对更高.
