# 保险

保险的职能：
* 基本职能
    - 分散风险
    - 经济补偿或给付
* 派生职能
    - 资金融通
    - 风险管理

中国商业保险的种类：
* 人身保险：以人的寿命和身体为保险标的的保险，按照保险责任分为：
    - 人寿保险
    - 年金保险
    - 健康保险
    - 意外伤害保险
* 财产保险：以财产及其有关利益为保险标的的保险
    - 财产损失保险
    - 责任保险
    - 信用保证保险

保险的基本原则：
* 最大诚信原则
* 保险利益原则：保险利益又称可保利益，是指投保人或者被保险人对保险标的所具有的法律上承认的经济利益
* 近因原则
* 损失补偿原则：其派生原则有：
    - 代位原则
    - 重复保险的分摊原则

根据《人身保险公司保险条款和保险费率管理办法》，人身保险的命名规则：“保险公司名称”+“吉庆、说明性文字”+“险种类别”+“（设计类型）”，其中保险公司名称可用全称或简称，吉庆、说明性文字字数不超过 10 个，附加保险的定名应当在“保险公司名称”后标注“附加”字样。分红型、投资连结型和万能型人身保险应当注明设计类型，普通型无需注明。

## 人寿保险

人寿保险是以保险人的寿命作为保险标的，以被保险人在保险期间内生存或身故为给付保险金条件的保险。广义的人寿保险包括年金保险。

人寿保险的特征：
* 保额的确定没有人的生命的实际价值作为客观依据，实务中人寿保险的保额由投保人和保险人双方约定后确定，主要考虑两个方面：
    - 被保险人对人寿保险需要的程度，比如可通过生命价值法或遗属需要法测算
    - 投保人缴纳保费的能力
* 人寿保险的主要风险因素是死亡率
* 人寿保险中，保险利益只是订立保险合同的前提条件，并不是维持保险合同效力、保险人给付保险金的条件。
* 人寿保险是定额给付性保险，不适用补偿原则，不存在比例分摊和代位补偿问题

**自然保费**：以人的每一年龄及一年期间的死亡率为基础指定的各年龄的保费，其数额等于预期的当年成本，每期缴费随死亡率的上升而逐步提高。

**均衡保费**：保险人将被保险人不同年龄的自然保费（考虑时间价值），均匀分配到每个缴费期限的保费，每期缴费相同。

**现金价值**：现金价值是指投保人退保时可以得到的现金数额，也成解约金或退保金。现金价值是均衡保费高于自然保费的差额累计生息形成的。对采用均衡保费方式缴费的投保人，退保时应当返还这部分利益。

**风险保额**：是指保险公司支付的死亡保险金与现金价值的差额，表示保险人所给付的死亡保险金中，除了报单持有人自己的贡献外，保险人所承担的部分。

人寿保险的种类
* 普通型人寿保险：按照寿险产品的保障内容分为：
    - 定期寿险
    - 终身寿险
    - 两全保险
* 新型人寿保险：按照寿险产品的设计形态分为：
    - 分红保险
    - 万能保险
    - 投资连结保险


### 定期寿险

在合同中规定一定时期为保险有效期（可以是年限或者约定的年龄），若被保险人在期限内死亡，保险人即给付受益人约定的保险金；若被保险人在期限届满时仍然生存，契约即行终止，保险人无给付义务，亦不退还已收的保费。按照保额是否可变，分为：
* 保额恒定定期寿险
* 保额递增定期寿险：保额按照约定金额或比例递增，比如按生活费用指数递增的 COLA（Cost-OF-Living-Adjustment）保单（可乐保单）
* 保额递减定期寿险：保额按照约定金额或比例递减，比如抵押贷款保证定期寿险保单（Mortgage Protection Term），保险金额与抵押贷款未偿还余额保持一致，或者家庭收入保险（Family Income Policy）。

定期寿险的特殊条款：
* 可续保条款
* 可转换条款


### 终身寿险

终身寿险是指以死亡为给付保险金条件，且保险期限为终身的人寿保险。按照缴费方式可分为：
* 普通终身寿险：即保费终身分期缴付。
* 限期缴付终身寿险：即保费在规定期限内分期缴付，期满后不再缴付保费，但仍享有保险保障。
* 趸缴终身寿险：即在投保时一次全部缴清保费。

终身寿险会一直累积现金价值，并在保险人生存至生命表年龄上限时等于保额，不同缴费期限决定了现金价值的增长速度，普通终身寿险的现金价值增长最慢，趸缴型增长最快，限期型期限越短增长越快。

### 两全保险

两全保险是指被保险人在保险合同有效期内死亡或合同期满时仍生存，保险人按照合同均承担给付保险金责任的保险。两全保险既保障死亡又保障生存，不仅使受益人得到保障，也使被保险人本身享受其利益。



## 年金保险

年金是一系列定期有规则的款项支付，分为确定型年金和不确定型年金。确定型年金不含保险因素，而年金保险属于不确定型年金，是指以被保险人生存为给付条件，并按约定的时间间隔给付生存保险金的人身保险。市场上的年金保险通常为：
* 养老年金保险
* 教育年金保险

## 健康保险

健康保险是指被保险人患病、分娩过程中，以及由此造成死亡或者残废的情况下，由保险人给付医疗费、护理费或保险金。国外有时把健康保险和人身意外伤害保险归为一类。

按照损失原因分为：
* 意外伤害保险
* 疾病保险

按照损失种类分为：
* 收入保险：意外伤害保险对完全丧失工作能力和部分丧失工作能力都给付收入保险金，疾病保险一般只对完全丧失工作能力给付收入保险金。
* 死亡和残废保险：意外伤害保险在被保险人意外死亡情况下向指定受益人给付一次性保险金，也对意外丧失肢体、视力等给付一次性保险金，疾病保险不给付死亡保险金。
* 费用保险：比如住院费用保险单、普通医疗费用保险单、大额医疗保险单等。

按照承保标准分为：
* 简易健康保险
* 老年健康保险
* 特种风险健康保险
* 弱体健康保险

收入保险和意外伤害保险通常使用定值保险单，即在保险事故发生时，被保险人或者受益人得到预定金额的保险金。费用保险按照给付方式分为：
* 定值方式
* 报销方式
* 提供服务方式

### 人身意外伤害保险

人身意外伤害保险指在保险期内因发生意外事故致使被保险人死亡或伤残，保险人按照合同规定给付保险金。意外事故的构成具备三要素：
* 意外发生的：即被保险人未预料到的和非故意的事故
* 外来原因造成的：即被保险人身体外部原因造成的事故，疾病所致伤害，是人体内部生理故障，不属于意外事故。
* 突然发生的：即事故的原因与伤害的结果之间具有很直接的关系，在瞬息间造成伤害，来不及预防。

人身意外伤害保险可以单独办理，也可以作为一种附加险。具有下列特点：
* 保险期限较短：一般不超过 1 年，这与财险相似
* 净保险费根据保险金额损失率计算
* 年末未到期责任准备金是按当年保险费收入的一定比例提存的

人身意外伤害保险的保险责任项目包括死亡给付、伤残给付、医疗费给付和丧失工作能力收入给付等，一般来说，死亡保险金和伤残保险金的给付合为一个保险金额。

### 医疗保险

医疗保险分为：
* 住院费用保险：一般给付：
    - 病房和膳食费用
    - 医药费和杂费
* 外科费用保险
* 普通医疗费用保险
* 大额医疗保险
* 补充医疗保险
* 长期护理保险

# 保险规划


客户未来一生的支出的现值等于过去累积资产净值与客户未来一生的净收入（营生资产）之和，则客户可以安度一生。如果事故发生，比如死亡、失能、失业，那么客户的个人收入将降低或中断，营生资产随之降低甚至为 0，这就应该通过保险来弥补缺口。

购买保险的基本原则：
1. 以风险缺口为基础：总风险缺口对应总保险需求，分项缺口对应分项需求。
2. 以风险保障为先导：先风险保障再储蓄投资。
3. 注重风险防范顺序：按风险事件影响程度可依次考虑死亡、残疾、重疾、医疗、退休等，按保险标的依次考虑家庭主要收入者、次要收入者。
4. 优化保险产品组合：用最少的保费做最大的风险保障（如充分利用附加险、消费型险种、定期与终身寿险转换）。
5. 保费负担量力而行
6. 选择优质保险公司



# 人寿保险规划

## 寿险保额的确定

**倍数法则**

倍数法则是一种简单估算死亡保险金额的方法，比如根据十一法则，家庭需要的人寿保险的死亡风险保额，大约应该等于家庭税后年收入的 10 倍（这里没有考虑任何支出）。适合于一般的估算和速算。缺点：
* 没有考虑不同家庭的支出、投资和负债情况
* 没有考虑被保险人的家庭角色和家庭责任

**生命价值法**

应有保额应该等于未来收入的折现值减去未来支出的折现值。这种算法类似产险的保额，是以投保物本身的价值为上限。应有保额通常受下列因素影响：
* 年龄
* 个人支出占个人收入的比例
* 个人收入的成长率
* 投资收益率

通常收入假定期末发生，支出则假定期初就准备好。生命价值法的思路是从被保险人出险以后家庭可能遭受的经济损失触发，将未来出险所造成的所有经济损失的折现值作为保额，用来规避被保险人出险导致家庭收入损失的风险。这种保额的计算适合目前生息资产少或者未来收入非常可观的情况，例如刚步入社会的新人以及具有高财富创造能力的高净值人群。

案例：被保险人 30 岁，年收入 10 万元，个人年支出 3 万元，预计 60 岁退休，投资收益率为 3%，不考虑收入和支出的增长率，按照生命价值法计算的寿险保额。

30 年收入的现值为 PV(n=30, I=3%, PMT=10, FV=0, 期末年金) = -196 万元，30 年支出的现值为 PV(n=30, I=3%, PMT=3, FV=0, 期初年金) = -60.57 万元，则生命价值，即保险金额为 196 - 60.57 = 135.43 万元。

**遗属需要法**

在被保险人死亡时，遗属终其一生的生活需要扣除被保险人生前的累积净值，就是应有保额，即：

应有保额 = 遗属生活费用缺口 + 紧急预备金 + 子女高等教育金现值 + 房贷及其他负债 + 丧葬最终支出现值 - 生息资产变现值

生息资产变现值是指当保险事故发生后可以处理的资产在当时的市场价值。假设在最差的市场情况下，可变现资产可能的损失率为 $\beta$，那么可变现资产的净值需要乘以 $1-\beta$。

遗属需要法的思路是从被保险人出险以后家庭收入减少，原有的家庭财务目标出现缺口出发，将这些不同时期的财务目标可能产生的缺口数额折现到当期作为保额，用来规避因保险人出险导致家庭原有财务目标无法实现的风险。这种保额计算思路更多适用于普通家庭的常见财务目标规划。遗属需要法可以方便的调整原来生涯规划中不足或超出部分，故经常采用该法计算应有保额。

案例：假设被保险人现在死亡，被保险人还需要生活 50 年，每年的生活开支是 5 万元，目前可变现资产价值 20 万元，预计投资收益率 3%，不考虑其他变量，遗属的未来生活开支的现值为 PV(n=50, I=3%, PMT=-5, FV=0, 期初年金) = 132.51 万元，则应有保额 = 132.51 - 20 = 112.51 万元。

年金保险是不确定性年金，是指以被保险人生存为给付保险金条件，并按约定的时间间隔给付生存保险金的人身保险。



# 附录

## 生命表

生命表又称“死亡表”，是反映一个国家或一个区域人口生存死亡规律的调查统计表。生命表是对相当数量的人口自出生(或一定年龄)开始，直至这些人口全部去世为止的生存与死亡记录。通常以 10 万（或 100 万）人作为 0 岁的生存人数，然后根据各年中死亡人数，各年末生存人数计算各年龄人口的死亡率、生存率，列成表格，直至此 10 万人全部死亡为止。生命表上所记载的死亡率、生存率是决定人寿保险费的重要依据，还可根据它计算人口的平均预期寿命。此表系根据分年龄死亡率编制，并主要反映各年龄死亡水平，故又称死亡率表。生命表的编制为经营人寿保险业务奠定了科学的数理基础，是计算人身保险的保险费、责任准备金、退保金的主要依据。

生命表中的年龄间隔，可以是 1 岁一组，也可以是 5 岁一组或 10 岁一组。前者称为**完全生命表**，后者称为**简略生命表**。简略生命表中往往也把 0 岁组单独列出。生命表的原意是反映同一批人陆续死亡减少过程当中各年龄人数之间的比例关系，以及这批人活到各年龄时的平均余寿。整个生命表的推理过程都是针对一批人而言的，但实际编制生命表时，通常都是根据某一年或连续几年不同批人的死亡率，因而反映的是该时期的死亡水平，称为**时期生命表**。反之，根据某同批人经历各年龄时的死亡率而编制的生命表，则称为**同批人生命表**或**一代人生命表**。根据某地区、某时期实际死亡率水平编制的生命表，只能反映该地区、该时期情况。综合许多不同死亡水平的生命表，编制成与各种不同 0 值相应的成套的生命表，称为**模型生命表**，可供缺乏资料的国家或地区使用。

因为我国保险业发展时间短暂，改革开放之初基本参照了日本的经验生命表。直到1995年，当时的中国人民保险公司参考其 1990 年至 1993 年的保单数据，制订了中国人寿保险业第一张经验生命表。这张生命表从 1996 年开始使用。20 世纪 90 年代中国人民保险（集团）公司组织了大量的专家，成功地编制出《中国人寿保险业经验生命表》，并于 1997 年 4 月 1 日起正式运用于人寿保险业务的经营核算中。中国保监会 2003 年 8 月启动我国第二张《生命表》的编制工作，第一次在全行业范围内进行经验死亡率调查，参调公司为国内经营时间较长、数据量较大的六家公司，所用保单数据一亿多条，占行业数据量的 98% 以上。

生命表中的死亡概率 $q_x$ 由实际调查所得的分年龄死亡率 $m_x$ 换算得到。通常用的最简单的公式为 1 岁组:
$$
q_x = \frac{2m_x}{2+m_x}
$$
其中，$q_x$ 为 x 至 x+1 岁的死亡概率，$m_x$ 为 x 至 x+1 年龄组的死亡率。对于 n 岁组为：
$$
q_{x,n} = \frac{2\cdot n\cdot m_{x,n}}{2 + n\cdot m_{x,n}}
$$
其中，$q_{x,n}$ 为 x 至 x+n 岁的死亡概率，$m_{x,n}$ 为 x 至 x+n 年龄组的死亡率。


生命表的内容如下：

|年龄 $x$|死亡率 $q_x$|生存人数 $l_x$|死亡人数 $d_x$|生存人年数 $L_x$|x 岁以后生存人年数总和 $T_x$|平均余命 $\dot{e}_x$
|----|----|----|----|----|----|----|
|0|0.000867|1000000.00|867|999566.50|76420141.87|76.42|
|1|0.000615|999133|614.47|998825.77|75420575.37|75.49|
|2|0.000445|998518.53|444.34|998296.36|74421749.60|74.53|
|…|…|…|…|…|…|…|
|105|1|85.43|85.43|42.72|42.72|0.5|

其中:
* $x$ 表示年龄，$x=0$ 表示刚出生的那一刻，假设的最大年龄为 $T$，中国为 105 岁。
* $q_x$ 表示死亡概率，活到每一个确切年龄 x 的人在未来的一岁间隔（也即未来的一年时间）内，均有死亡的可能。不同年龄的这种死亡可能性大小不同，死亡概率即指活到 x 岁的人在 x 至 x+1 岁间死亡的可能性，$q_x=P(在 x 至 x+1 岁死亡|x 岁生存)$。假设在最大年龄的死亡率为 1，即 $q_T=1$。各年龄死亡概率的数值根据实际调查的各年龄死亡率计算而得，它是整个生命表的基础，其他指标均由此衍生而出。
* $p_x$ 表示生存概率，即指活到 x 岁的人在 x 至 x+1 岁间生存的可能性。$p_x=P(x+1 岁生存|x 岁生存)=1-q_x$。
* $l_x$ 表示活到确切年龄 x 岁的人数，$l_0$ 即确切年龄 0 岁人数，也即出生人数。生命表中的计算均从 $l_0$ 开始，故 $l_0$ 称为生命表的基数。显然，表中其他各栏人数均随 $l_0$ 的取值而定。一般生命表中为使人数保持整数，中国多取 $l_0=1000000$；也可取$l_0=1$，则表中各栏人数均成为与 $l_0$ 的比例系数。$l_x$ 和 $q_x$ 的换算关系为：
$$
\begin{aligned}
& l_x = l_0\cdot\prod\limits_{y=0}^{x-1}P(y+1 岁生存 | y 岁生存) = l_0\cdot\prod\limits_{y=0}^{x-1}(1-q_y)\\
& q_x = \frac{l_x - l_{x+1}}{l_x}
\end{aligned}
$$
* $d_x$ 表示活到 x 岁的人在 x 至 x+1 岁间死亡的人数，即 $d_x = l_x - l_{x+1}$。
* $L_x$ 生存人年数指已活到确切年龄 x 岁的人在 x 至 x+1 岁间所活的人年数。由于在这一岁间隔内死亡的人是陆续死去的，既活了一段时间，又未满一年，所以 $l_x>L_x>l_{x+1}$。在 1 岁一组的年龄组中，$L_x$ 也可看做是 x 至 x+1 岁这一年龄间隔内的平均人数。若假定死亡是均匀分布的，则 $L_x = \frac{l_x+l_{x+1}}{2}$。
* $T_x$ 将已活到确切年龄 x 岁者以后预期可活的人年数加以累计，即得 x 岁以后预期生存人年总数 $T_x$，即所有年龄 x 岁的人 $l_x$ 的余命分布为：

|死亡年龄段|平均余命|人数|
|----|----|----|
|x 至 x+1|0+1/2|$d_x$|
|x+1 至 x+2|1+1/2|$d_{x+1}$|
|…|…|…|
|x+n 至 x+n+1|n+1/2|$d_{x+n}$|
|…|…|…|
|T-1 至 T|T-1-x+1/2|$d_{T-1}$|
|T 至 T+1|T-x+1/2|$d_{T}$|

$$
\begin{aligned}
T_x &= \sum\limits_{n=0}^{T-x}\left(n+\frac{1}{2}\right)(l_{x+n} - l_{x+n+1})\\
&= \frac{1}{2}\sum\limits_{n=0}^{T-x}(l_{x+n} - l_{x+n+1}) + \sum\limits_{n=0}^{T-x}nl_{x+n} - \sum\limits_{n=0}^{T-x}nl_{x+n+1}\\
&= \frac{1}{2}l_x + \sum\limits_{n=1}^{T-x}nl_{x+n} - \left(\sum\limits_{k=1}^{T-x}(k-1)l_{x+k} + (T-x)l_{T+1}\right), (k=n+1)\\
&= \frac{1}{2}l_x + \sum\limits_{n=1}^{T-x}nl_{x+n} - \sum\limits_{n=1}^{T-x}(n-1)l_{x+n}\\
&= \frac{1}{2}l_x + \sum\limits_{n=1}^{T-x}l_{x+n}\\
&= \sum\limits_{n=0}^{T-x}\frac{l_{x+n}+l_{x+n+1}}{2}\\
&= \sum\limits_{n=0}^{T-x}L_{x+n}
\end{aligned}
$$
* $\dot{e}_x$ 已活到确切年龄 x 岁的人的平均余命，$\dot{e}_x = \frac{T_x}{l_x} = \frac{1}{2} + \frac{1}{l_x}\sum\limits_{n=1}^{T-x}l_{x+n}$。

## 人寿保险数学原理


![保险现金流.png](./images/保险现金流.png)


假设被保险人当前年龄为 x，当前时刻为 0，保险有效期为 T。第 t 期的净保费为 $P_t$，保费的缴纳发生在期初。第 t 期如果发生了风险事件，保险人支付的保险金为 $P_t$，保险金的支付发生在期末。根据生命表，第 t 期的死亡概率为 $q_{x+t-1}$，第 t 期初的生存人数为 $l_{x+t-1}$，第 t 期的死亡人数为 $d_{x+t-1}$。第 t 期的投资收益率为 $r_t$。

首先分析保险金支付的概率分布：

|时期|保险金|支付概率|
|----|----|----|
|1|$B_1$|$\frac{d_x}{l_x}$|
|2|$B_2$|$\frac{d_{x+1}}{l_x}$|
|…|…|…|
|t|$B_t$|$\frac{d_{x+t-1}}{l_x}$|
|…|…|…|
|T|$B_T$|$\frac{d_{x+T-1}}{l_x}$|

接着分析净保费缴纳的概率分布

|时期|净保费|缴纳概率|
|----|----|----|
|1|$P_1$|$\frac{l_x}{l_x }$|
|2|$P_2$|$\frac{l_{x+1}}{l_x}$|
|…|…|…|
|t|$P_t$|$\frac{l_{x+t-1}}{l_x}$|
|…|…|…|
|T|$P_T$|$\frac{l_{x+T-1}}{l_x}$|

根据净现值公式，让净现值为 0，可得：
$$
\sum\limits_{t=1}^T\frac{B_t\cdot\frac{d_{x+t-1}}{l_x}}{\prod\limits_{\tau=1}^t(1+r_\tau)} = \sum\limits_{t=1}^T\frac{P_t\cdot\frac{l_{x+t-1}}{l_x}}{\prod\limits_{\tau=1}^{t-1}(1+r_\tau)}
$$
即：
$$
\sum\limits_{t=1}^T\frac{B_t\cdot d_{x+t-1}}{\prod\limits_{\tau=1}^t(1+r_\tau)} = \sum\limits_{t=1}^T\frac{P_t\cdot l_{x+t-1}}{\prod\limits_{\tau=1}^{t-1}(1+r_\tau)}
$$
通常投资收益率为固定利率 $r$，则：
$$
\sum\limits_{t=1}^T\frac{B_t\cdot d_{x+t-1}}{(1+r)^t} = \sum\limits_{t=1}^T\frac{P_t\cdot l_{x+t-1}}{(1+r)^{t-1}}
$$
如果保险金固定为 B，则：
$$
B\cdot\sum\limits_{t=1}^T\frac{d_{x+t-1}}{(1+r)^t} = \sum\limits_{t=1}^T\frac{P_t\cdot l_{x+t-1}}{(1+r)^{t-1}}
$$


### 保险准备金

下面考察理论保险准备金的计算。

**过去法（Retrospective Method）**：理论准备金 = 过去收到的净保险费的复利终值 - 过去已给付保险金的复利终值，假设当前保险公司年龄为 x 岁的投保人数为生命表上的 $l_x$，对于保险公司整体而言，第 t 期末这部分投保人的准备金为：
$$
V_t = \sum\limits_{\tau=1}^t\left(\prod\limits_{s=1}^{t-\tau+1}(1+r_s)\right)\cdot P_\tau l_{x+\tau-1} - \sum\limits_{\tau=1}^t\left(\prod\limits_{s=1}^{t-\tau}(1+r_s)\right)\cdot B_\tau d_{x+\tau-1}
$$
当 $t=T$ 时，由保险费和保险金满足的公式可知：$V_T = 0$。投资收益率固定时为：
$$
V_t = \sum\limits_{\tau=1}^t(1+r)^{t-\tau+1}\cdot P_\tau l_{x+\tau-1} - \sum\limits_{\tau=1}^t(1+r)^{t-\tau}\cdot B_\tau d_{x+\tau-1}
$$

**未来法（Prospective Method）**：理论准备金 = 将来给付保险金的现值 - 将来净保费收入的现值，按照上面的假设，第 t 期末保险公司的准备金为：
$$
V'_t = \sum\limits_{\tau=t+1}^T\frac{B_\tau d_{x+\tau-1}}{\prod\limits_{s=t+1}^{\tau}(1+r_s)} - \sum\limits_{\tau=t+1}^T\frac{P_\tau l_{x+\tau-1}}{\prod\limits_{s=t+1}^{\tau-1}(1+r_s)}
$$
投资收益率固定时为：
$$
V'_t = \sum\limits_{\tau=t+1}^T\frac{B_\tau d_{x+\tau-1}}{(1+r)^{\tau-t}} - \sum\limits_{\tau=t+1}^T\frac{P_\tau l_{x+\tau-1}}{(1+r)^{\tau-t-1}}
$$
通过做差可以证明两种方法计算的准备金是相等的，$V_t=V'_t$。

对于单个投保人，如果在 t 期末其生存，则其账户的准备金为：
$$
v_t = \frac{V_t}{l_{x+t}}
$$

考虑具体的产品，对于**定期寿险**，如果一次缴清保费，即 $P_t=0,t\geq2$，则 0 时刻的净保费为：
$$
P = B\cdot\frac{\sum\limits_{t=1}^{T}(1+r)^{-t}d_{x+t-1}}{l_x}
$$
如果是均衡保费，缴费期为 $n$，即 $P_t=P,t\leq n,P_t=0,t\geq n+1$，则均衡净保费为：
$$
P = B\cdot\frac{\sum\limits_{t=1}^{T}(1+r)^{-t}d_{x+t-1}}{\sum\limits_{t=1}^{n}(1+r)^{-t+1}l_{x+t-1}}
$$

对于**终身寿险**，x+T 就是生命表上最大年龄，

```python
# 终身寿险
r = 0.04# 收益率
B = 10000# 保险金额
x = 40# 当前年龄
T = CLTable.index[-1] - x# 保险有效期
t = np.arange(1, T+1).astype(float)
n = 1# 缴费期数, 1：趸缴, T: 终身缴费
d = CLTable.iloc[x:-1, 2].values
d[-1] += CLTable.iloc[-1, 2]
l = CLTable.iloc[x:-1, 1].values

P = B * np.sum((1+r)**(-t) * d) / l[0]
V = (1+r)**t * P * l[0] - B * (1+r)**t * np.cumsum((1+r)**(-t)*d)
v = V / CLTable.iloc[x+1:, 1].values
pd.Series(v[:-1]).plot();
```
