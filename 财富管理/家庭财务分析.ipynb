# 家庭会计

**会计要素**：对会计对象的基本分类：
* 资产：主体过去的交易或事项形成的、由主体拥有或者控制的、预期会给主体带来效用或经济利益的资源
* 负债：主体过去的交易或事项形成的主体在将来的付款义务
* 所有者权益（净值）：资产扣除负债后的余额
* 收入：主体在一定期间内获得的导致所有者权益（净值）增加的经济利益流入
* 费用（支出）：主体在一定期间内发生的导致所有者权益（净值）减少的经济利益流出
* 利润（储蓄）：主体在一定期间内的经济利益成果

资产、负债和所有者权益（净值）三项会计要素反映主体的财务状况，收入、费用（支出）和利润（储蓄）三项会计要素反映主体在一定期间的经营成果。

**会计科目**：对会计要素的具体内容进行分类的项目。

**会计等式**：
* 资产 = 负债 + 所有者权益（净值）
* 利润（储蓄） = 收入 - 费用（支出）

**财务报表**：对主体财务状况、经营成果和现金流量的结构性表述，至少包括：资产负债表、利润表、现金流量表以及附注等。

家庭会计和企业会计的区别：
* 会计核算原则不同：家庭会计使用收付实现制，企业会计使用权责发生制：
    - 收付实现制：现金制，即以收入带来的现金“收到”时间和由费用导致的现金“付出”时间为准来确认收入和费用的一种会计核算基础，有现金流入或流出才记账。
    - 权责发生制：应计制，即以导致收入实现和费用发生的“行为”的发生时间为准来确认收入和费用的一种会计核算原则
* 财务信息适用范围不同
* 范式要求不同
* 核算准则不同
* 家庭财务管理倾向于处理现金而不是未来会有的收入，即家庭财务会计几乎不进行收入或费用的资本化。

收入和支出是流量概念，通常以收入循环一次的时间作为确定流量的期间，比如国内工薪阶层一般是领月薪，收入、支出、储蓄等流量通常按月计算，汇总月流量便得到一季或者一年的流量。资产、负债是存量概念，通常以月底、季度底或是年底作为资产负债的结算基准日。

对于资产价值变动的处理：
* 对于市价评估不易、流动性较差的房地产、汽车、古董或未上市股票、债券等资产，可以根据成本入账，平常不进行资产重估调整，在处置此类资产时其处分收益直接列入净值变动额。但如果市价在短期内却有很大变化，应自行估价调整。
* 对于市价变动频繁且有客观依据来评判的上市公司股票、基金等，应于每期编制资产负债表时，将未实现的资本利得或损失反映在当期净值的变动上。

将资产重估增减值和市价变动后未实现的资本利得或损失作为净值调整项目后，期初期末净值变动额和当期储蓄之间的恒等式为：
**当期净值变动额 = 当期储蓄额 + 当期资产重估调整值 + 当期未实现资本利得或损失**

资产负债表的恒等关系：
* 净值 = 资产 - 负债
* 当期储蓄 = 以成本计价的期末净值 - 以成本计价的期初净值
* 当期储蓄 + 资产账面价值变动 = 以市值计价的期末净值 - 以市值计价的期初净值

当期可以分别以成本以及市值计价来编制资产负债表，一个用于收支储蓄表相对照，一张用于显示家庭当前的实际财富状况。如果由资产负债表计算的净值变动与收支储蓄表计算的储蓄不一致时，应以两期净值的差异为准来调整收支储蓄表的储蓄数字，差异额计入其他收入或其他支出中。

    - 储蓄型保费：养老险、还本险、退休年金以及投资型保险等属于储蓄型保险，其保费列为储蓄，所累积的现金价值列为投资性资产；终身寿险兼有保障性和储蓄性，每年保费超过自然保费的部分为储蓄，储蓄的目的是以最能节税的方式累积遗产

    - 个人失业保险费为保障型保费
    - 对于医疗保险，单位所缴保费中拨入个人账户的及个人所缴保费为限制支配收入，并形成储蓄；个人医疗保险账户余额为资产
    - 个人所缴的基本养老保险费拨入养老保险社会统筹基金部分相当于保障型保费；拨入个人账户部分为限制支配收入与储蓄，个人养老保险账户余额为资产
    - 个人与单位住房公积金缴存全部为限制支配收入与储蓄，住房公积金账户余额为资产
    - 三险一金中属于或相当于保障型保费部分同样可列为收入的减项

# 家庭财务分析

## 资产负债表分析

资产负债的结构分析：
* **流动性资产（CA） = 流动性负债（CL） + 流动性净值（CE）**：流动性负债是透支信用借钱来消费，等于在结算时点时所积欠的余额；流动性资产的特点是可保本，安全性和流动性最强，获利性最低，其持有动机：
    - 交易性动机
    - 预防性动机：家庭需要准备**紧急预备金** = (月生活费预算 + 月房租或房贷) × 失业保障月数 + 意外超额支出准备
    - 投机性动机
* **投资性资产（IA） = 投资性负债（IL） + 投资性净值（IE）**：除流动性资产外，所有可以产生利息收入或资本利得的资产都可算作投资性资产；投资性负债是扩张信用借钱来投入投资性资产的借款余额。
* **自用性资产（UA） = 自用性负债（UL） + 自用性净值（UE）**：自用性资产是个人或家庭使用的且拥有产权的资产，其可以提供使用价值，不以赚取买卖价差为主要目的；自用性负债是用来购买自用性资产的贷款余额。


## 收支储蓄表分析

收支储蓄的结构分析
* **工作收入 - 生活支出 = 工作储蓄**：工作收入是其他财富的来源，但其具有时限性；生活支出应该进行预算控制。
* **理财收入 - 理财支出 = 理财储蓄**：退休后只有理财收入，在工作期间应逐年提高理财收入比重；由于退休之后无工作，退休前应把贷款还清，保费缴清，退休之后不再有理财支出，只有生活支出。

储蓄运行分析：**总储蓄 = 自由储蓄 + 固定用途储蓄**
* 固定用途储蓄：是指已经安排的本金还款或投资，包括当月拨入个人住房公积金账户和个人基本养老账户的金额、房贷定期摊还的本金额、应缴储蓄型保费以及其他目标的固定缴存（比如教育储蓄存款等）
* 自由储蓄：即一般人所认定的储蓄额，是可以自由决定如何使用的储蓄，可以用于实现一些计划外的短期理财目标或者提前还清贷款

收入支出象限与储蓄方向：
* 高收入低支出：理财重点是有效运用储蓄，积极投资来提高投资收益率
* 高收入高支出：理财重点在节流，通过有计划的消费来调高储蓄率
* 低收入低支出：理财重点在开源，通过加班、兼职等开创收入
* 低收入高支出：理财重点在于量入为出，尽量避免消费借贷等

## 财务比率分析

### 家庭偿债能力

**资产负债率 = 总负债 / 总资产**

通常资产负债率应保持在 60% 以下，若超出此范围，则说明负债水平过高；若是负债主要为长期摊还的房贷则可接受，否则应立即进行减债计划。

**流动比率 = 流动性资产 / 流动性负债**

通常流动比率应该保持在 200% 以上。

**融资比率 = 投资性负债 / 投资性资产**

投资性负债的存在说明家庭投资时利用了财务杠杆，这使得投资的收益和亏损都放大了，通常融资比率保持在 50% 以下。

**财务负担率 = 年本息支出 / 年可支配收入**

财务负担率应该控制在 40% 以下。

**平均负债利率 = 年利息支出 / 负债总额**

平均负债利率反映了家庭实际承担的贷款利率水平，一般应在基准利率的 1.2 倍以下。

**自用贷款乘数 = 自用性负债 / 自用性资产**

如果自用性资产主要是自用房产，则自用贷款乘数 = 房贷乘数 = 房贷额 / 自用房产市值


**资产负债率分析**

资产负债率 = 总负债 / 总资产

= (流动性负债 + 自用性负债 + 投资性负债) / 总资产

= 流动比率 × 流动性资产比例 + 自用贷款乘数 × 自用性资产比例 + 融资比率 × 投资性资产比例



### 家庭应急能力

**紧急预备金倍数 = 流动性资产 / 月总支出**

一般流动性资产应该能应付 3~6 个月的支出

### 家庭保障能力

**保费负担率 = 保费 / 税后工作收入**

一般以工作收入的 10% 为合理商业保险保费预算标准。

### 家庭储蓄能力

**储蓄率 = (税后总收入 - 总支出) / 税后总收入**

储蓄率一般保持在 25% 以上。

**工作储蓄率 = (税后工作收入 - 消费支出) / 税后工作收入**

税后工作收入包括社保缴费收入，该指标一般应保持在 20% 以上。

**自由储蓄率 = (储蓄 - 固定用途储蓄) / 税后总收入

该指标一般保持在 10% 以上。

### 家庭宽裕度

**收支平衡点收入 = 固定负担 / 工作收入净结余比例**

其中：
* 固定负担：包括每月固定生活费用支出、房贷本息支出等近期内每月固定要流出的支出
* 工作收入净结余：等于工作收入 - 所得税扣缴额 - 三险一金缴费额 - 为了工作所必须支付的费用(比如通勤的交通费、工作用餐费等)
* 工作收入净结余比例：等于工作收入净结余 / 工作收入

**安全边际率 = (当前收入 - 收支平衡点收入) / 当前收入**

用于衡量收入减少或固定费用增加时有多少缓冲空间。


### 家庭财富增值能力

**生息资产比率 = 生息资产 / 总资产**

生息资产包括流动性资产和投资性资产，该指标衡量家庭资产中有多少可以拿来应付流动性、成长性与保值性需求，通常保持在 50% 以上。

**平均投资收益率 = 理财收入 / 生息资产**

衡量家庭投资绩效，一般应比通货膨胀率高 2 个百分点以上

### 家庭成长性

**资产增长率 = 资产增加额 / 期初总资产**

上式假设无负债情形。

**净值增长率 = 净值增加值 / 期初净值**

上式假设有负债，净值增长率代表家庭累积净值的速度。

### 家庭财务自由度

**财务自由度 = 年理财收入 / 年总支出**

理想的目标值是退休之际财务自由度大于等于 1，即包括退休金在内的资产，其利息可以维持生活。
