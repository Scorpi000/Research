# 跨期交易

## 摘要

## 理论模型

国债期货的理论价值为:
$$
F_t = \frac{P_t - Carry_t - Option_t}{CF}
$$
其中, $P_t$ 是 CTD 券的净价, $Carry_t$ 是持有收益, $Option_t$ 是交割期权的价格, $CF$ 为转换因子.

假设近月合约的理论价格为 $F_{1t}$, 远月合约的理论价格为 $F_{2t}$, 则跨期价差的理论值为:
$$
F_{1t}-F_{2t} = \frac{P_{1t}}{CF_1} - \frac{P_{2t}}{CF_2} - \left(\frac{Carry_{1t}}{CF_1}-\frac{Carry_{2t}}{CF_2}\right) - \left(\frac{Option_{1t}}{CF_1}-\frac{Option_{2t}}{CF_2}\right)
$$

假设 CTD 券是同一只, 通常转换因子都接近于 1, 如果忽略转换因子的差别, 以及不考虑交割期权价值的情况下, 跨期价差的理论价值为:
$$
F_{1t}-F_{2t} \approx Carry_{2t} - Carry_{1t}
$$
远月合约交割日比近月合约交割日多了 3 个月, 在上述假定下, 两者的持有收益之差就是以 CTD 国债票面利率为利息收入, 以 3 个月融资利率为资金成本而持有 CTD 券 3 个月的持有收益.


**跨期价差交易隐含利率(IRP)**

做多跨期价差可以理解为在近月合约到期时用发票价格 $F_{1t}^d=F_{1t}*CF_1+AI_{1T_1}$ 买入 CTD 国债, 持有该券至远月合约交割, 则获得远月合约的发票价格 $F_{2t}^d=F_{2t}*CF_2+AI_{2T_2}$, 把该笔交易的收益率定义为跨期价差交易隐含利率:
* 假设在 $[T_1, T_2]$ 期间债券不发生付息, 则:
$$
IRP_t = \frac{F_{2t}^d-F_{1t}^d}{F_{1t}^d}\cdot\frac{365}{T_2-T_1}
$$
* 在 $[T_1, T_2]$ 期间债券发生付息, 则 IRP 是下述方程的解:
$$
IRP_t = \frac{F_{2t}^d+\sum\limits_{T_1<t_i^c<T_2}\frac{C}{f}\cdot(1+IRP_t\frac{T_2-t^c_i}{365})-F_{1t}^d}{F_{1t}^d}\cdot\frac{365}{T_2-T_1}
$$
IRP 计算了做多跨期价差交易在远月合约交割日与近月合约交割日之间 3 个月的年化收益率, 在不考虑交易费用, 保证金等因素的前提下, 如果 IRP 显著大于平均融资成本, 则可以考虑进行做多价差交易, 在完成两次期货合约交割后可以基本锁定收益, 相当于无风险套利.


### 换月移仓交易策略

**临近交割月规律**

大部分期货持仓不会进入交割流程, 同时由于滚动交割阶段的空头举手规则, 多头持仓进入交割月面临两个不确定性: (1) 交割时间不确定, (2) 交割品种不确定. 交割的不确定性让很多持仓提前转移, 多头可能比空头更早的主动提前移仓. 总之, 在临近交割月时, 理论上合约多头总是倾向于平仓或者移仓, 两种行为都会促使跨期价差有缩小的趋势, 从而使得跨期价差在临近交割月时有先缩小后扩大的规律.


**换月移仓规律**

换月移仓(跨季移仓)是主力合约逐渐从近月合约向远月合约转移的过程. 多头主导的移仓会导致跨期价差先缩小后扩大, 空头主导的移仓会导致跨期价差先扩大后缩小. 如果能把握大规模移仓时点和多空双方的主导权, 则可以据此进行跨期价差交易.

移仓时点的判断可以观察近月合约和远月合约持仓量的变化, 如果近月合约持仓量连续大量减少, 远月合约持仓量连续大量增加, 则可以认为换月移仓开始.

多空移仓主导权可以根据移仓时的市场情绪猜测. 理论上如果 IRR 较高(基差很小或为负), 则空头交割意愿更强, 多头可能主导移仓; 反之, IRR 较低甚至为负(基差显著为正), 则多头交割意愿更强, 空头可能主导移仓.


### 波动性交易策略

**萨缪尔森效应(Samuelson Effect)**

期货价格的波动性应随着到期日的临近而上升, 期货近月合约价格比远月合约价格波动得更为剧烈. 

萨缪尔森效应主要是是基于这样一个预期: 尽管短期供需的不平衡会导致商品价格波动性的增加, 但是在长期, 供需的不平衡会逐渐趋于均衡. 

可获得的信息量可以对这一现象做出解释, 交易者能够获得的对远月合约价格产生影响的信息量相对于近月合约来说是比较少的, 而随着到期日的临近, 影响期货标的的价格的信息量会逐渐增加, 从而导致了期货价格的频繁波动. 

如果近月合约的波动性大于远月合约, 则近月合约在牛市中涨的更多, 而在熊市中跌得也更多.

**买入跨期套利(牛市跨期套利)**

若市场处于震荡牛市, 近月合约涨幅大于远月合约, 做多跨期价差获利.

**卖出跨期套利(熊市跨期套利)**

若市场处于震荡熊市, 近月合约的跌幅大于远月合约, 做空跨期价差获利.

## 实证分析

下面主要以 10 年期国债期货为例进行相关的实证分析.

```python
# coding=utf-8
import time
import sys
import os
import datetime as dt
import warnings
warnings.filterwarnings("ignore")
sys.path.append("C:\\Users\\hushuntai\\svn\\python")

from IPython.display import display
import numpy as np
import pandas as pd
import statsmodels.api as sm
from matplotlib.pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
#%matplotlib notebook
import matplotlib.dates as mdate
from tqdm import tqdm_notebook

import QuantStudio.api as QS
WDB = QS.FactorDB.WindDB2();
```

```python
# --------参数设置---------
StartDT = dt.datetime(2017,1,1)
EndDT = dt.datetime(2017,12,31,23,59,59,999999)
RecentMonthID = "T00.CFE"# 近月合约
FarMonthID = "T01.CFE"# 远月合约

# --------准备数据--------
WDB.connect()
FT = WDB.getTable("中国国债期货交易日行情")
DTs = FT.getDateTime(start_dt=StartDT, end_dt=EndDT)
Price = FT.readData(factor_names=["结算价(元)"], ids=[RecentMonthID, FarMonthID], dts=DTs).iloc[0]
WDB.disconnect()

Price = np.log(Price)# 取对数价格

# 绘制价格图像
Fig, Axes = plt.subplots(1, 1, figsize=(16, 8))
Axes.xaxis_date()
Axes.xaxis.set_major_formatter(mdate.DateFormatter('%Y-%m-%d'))
Axes.plot(Price.index, Price.values[:, 0], color='r', label=Price.columns[0])
Axes.plot(Price.index, Price.values[:, 1], color='b', label=Price.columns[1])
Axes.legend(loc='upper left')
Axes.set_title("Price");
```

```python
Spread = Price[RecentMonthID] - Price[FarMonthID]

# --------平稳性检验--------
# ADF 检验
SpreadADFTest = sm.tsa.stattools.adfuller(Spread.values, regression='c', autolag="BIC")
print(("Spread ADF Test Statistic : %f" % SpreadADFTest[0]), 
      ("pvalue : %f" % SpreadADFTest[1]), 
      ("usedlag : %d" % SpreadADFTest[2]), sep="\n")

# Hurst 指数
SpreadHurstExp, _ = QS.Tools.Math.genHurstExp(Spread.values)
print("Spread Hurst Exponent : %f" % SpreadHurstExp)

# ACF, PACF
Fig, Axes = plt.subplots(2, 3, figsize=(32, 16))
Axes[0, 0].plot(Spread)
Axes[0, 0].set_title("Spread")
sm.graphics.tsa.plot_acf(Spread, lags=60, ax=Axes[0, 1])
sm.graphics.tsa.plot_pacf(Spread, lags=60, ax=Axes[0, 2])
Axes[1, 0].plot(Spread.diff())
Axes[1, 0].set_title("Diff. of Spread")
sm.graphics.tsa.plot_acf(Spread.diff().iloc[1:], lags=60, ax=Axes[1, 1])
sm.graphics.tsa.plot_pacf(Spread.diff().iloc[1:], lags=60, ax=Axes[1, 2]);
```

```python
WDB.connect()
RP = WDB.getTable("货币市场日行情").readData(factor_names=["加权平均利率"], ids=["DR007.IB"], dts=DTs).iloc[0,:,0]# 银行间 7 天质押式回购利率
WDB.disconnect()

CointJohansenTest = sm.tsa.var.vecm.coint_johansen(np.c_[Spread.values, RP.values], det_order=0, k_ar_diff=1)
for i in range(CointJohansenTest.lr1.shape[0]):
    print(("Spread-RP Cointegration Johansen Test Trace Statistic(r<=%d): %f" % (i, CointJohansenTest.lr1[i])), 
          ("Critical Values (90%%, 95%%, 99%%) for Trace Statistic(r<=%d): %s" % (i, str(CointJohansenTest.cvt[i]))), 
          ("Maximum Eigenvalue Statistic Statistic(r<=%d): %f" % (i, CointJohansenTest.lr2[i])),
          ("Critical Values (90%%, 95%%, 99%%) for Maximum Eigenvalue Statistic(r<=%d): %s" % (i, str(CointJohansenTest.cvm[i]))), sep="\n")
```

```python
# --------平稳性检验--------
# ADF 检验
RPADFTest = sm.tsa.stattools.adfuller(RP.values, regression='c', autolag="BIC")
print(("RP ADF Test Statistic : %f" % RPADFTest[0]), 
      ("pvalue : %f" % RPADFTest[1]), 
      ("usedlag : %d" % RPADFTest[2]), sep="\n")

# Hurst 指数
RPHurstExp, _ = QS.Tools.Math.genHurstExp(RP.values)
print("RP Hurst Exponent : %f" % RPHurstExp)

# ACF, PACF
Fig, Axes = plt.subplots(2, 3, figsize=(32, 16))
Axes[0, 0].plot(RP)
Axes[0, 0].set_title("RP")
sm.graphics.tsa.plot_acf(RP, lags=60, ax=Axes[0, 1])
sm.graphics.tsa.plot_pacf(RP, lags=60, ax=Axes[0, 2])
Axes[1, 0].plot(RP.diff())
Axes[1, 0].set_title("Diff. of RP")
sm.graphics.tsa.plot_acf(RP.diff().iloc[1:], lags=60, ax=Axes[1, 1])
sm.graphics.tsa.plot_pacf(RP.diff().iloc[1:], lags=60, ax=Axes[1, 2]);
```

```python
# Johansen Test 求 Spread
SpreadWoRP = pd.Series(np.sum(CointJohansenTest.evec[:, 0]*np.c_[Spread.values, RP.values], axis=1), index=Spread.index)

# ACF, PACF
Fig, Axes = plt.subplots(1, 3, figsize=(32, 8))
Axes[0].plot(SpreadWoRP)
Axes[0].set_title("Spread Without RP")
sm.graphics.tsa.plot_acf(SpreadWoRP, lags=60, ax=Axes[-2])
sm.graphics.tsa.plot_pacf(SpreadWoRP, lags=60, ax=Axes[-1]);

# --------平稳性检验--------
# ADF 检验
SpreadWoRPADFTest = sm.tsa.stattools.adfuller(SpreadWoRP.values, regression='c', autolag="BIC")
print(("ADF Test Statistic of Spread without RP: %f" % SpreadWoRPADFTest[0]), 
      ("pvalue : %f" % SpreadWoRPADFTest[1]), 
      ("usedlag : %d" % SpreadWoRPADFTest[2]), sep="\n")

# Hurst 指数
SpreadWoRPHurstExp, _ = QS.Tools.Math.genHurstExp(SpreadWoRP.values)
print("Hurst Exponent of Spread without RP : %f" % SpreadWoRPHurstExp)

# --------自相关性检验--------
# Durbin-Watson 检验
SpreadWoRPDWTest = sm.stats.stattools.durbin_watson(SpreadWoRP.values)
print("Durbin-Watson Test Statistic of Spread without RP : %f" % SpreadWoRPDWTest)
# Ljung-Box Q 检验
SpreadWoRPQTest = sm.stats.diagnostic.acorr_ljungbox(SpreadWoRP.values, lags=None)
SpreadWoRPQTest = pd.DataFrame(np.array(SpreadWoRPQTest).T, columns=["Ljung-Box Q Statistic", "Ljung-Box pvalue"],
                               index=pd.Index(np.arange(1, SpreadWoRPQTest[0].shape[0]+1),name="lag"))
print("Ljung-Box Q Test of Spread without RP", SpreadWoRPQTest, sep="\n")
```

```python
# 对 Spread 序列建立 ARMA 模型
# 确定 ARMA 模型的阶数
# ARMAOrder = sm.tsa.stattools.arma_order_select_ic(SpreadWoRP.values, max_ar=10, max_ma=3, ic="bic", trend='c')
# ARMAOrder = ARMAOrder.bic_min_order
# print(ARMAOrder)
# 构建 ARMA 模型
ARMAModel = sm.tsa.ARMA(SpreadWoRP.values, order=(2, 0))
ARMARslt = ARMAModel.fit(trend="c", disp=False)
print(ARMARslt.summary())

Fig, Axes = plt.subplots(1, 2, figsize=(25, 8))
Axes[0].xaxis_date()
Axes[0].xaxis.set_major_formatter(mdate.DateFormatter('%Y-%m-%d'))
Axes[0].set_title("ARMA Model for Spread without RP")
Axes[0].plot(SpreadWoRP.index, SpreadWoRP.values, label="Spread without RP 实际值")
Axes[0].plot(SpreadWoRP.index, ARMARslt.fittedvalues, label="Spread without RP 预测值")
Axes[0].legend()
Axes[1].xaxis_date()
Axes[1].xaxis.set_major_formatter(mdate.DateFormatter('%Y-%m-%d'))
Axes[1].plot(SpreadWoRP.index, ARMARslt.resid)
Axes[1].set_title("Residuals of ARMA Model for Spread without RP");
```

```python
# 样本内策略
Signal = np.r_[ARMARslt.fittedvalues[1:], ARMARslt.forecast(1)[0]] - (RP.values * CointJohansenTest.evec[:, 0][1]) - (Spread * CointJohansenTest.evec[:, 0][0])
NumUnitsLong = pd.Series(np.nan, index=SpreadWoRP.index)
NumUnitsLong.iloc[0] = 0
NumUnitsLong[Signal > 0] = 1
NumUnitsLong[Signal <= 0] = 0
NumUnitsLong.fillna(method="pad", inplace=True)
NumUnitsShort = pd.Series(np.nan, index=SpreadWoRP.index)
NumUnitsShort.iloc[0] = 0
NumUnitsShort[Signal < 0] = -1
NumUnitsShort[Signal >= 0] = 0
NumUnitsShort.fillna(method="pad", inplace=True)
NumUnits = NumUnitsLong + NumUnitsShort
Positions = pd.DataFrame(np.repeat(NumUnits.values.reshape((NumUnits.shape[0], 1)), Price.shape[1], axis=1) * CointJohansenTest.evec[:, 0][0] * np.array([1, -1]), index=Price.index, columns=Price.columns)
# Positions = np.repeat(NumUnits.values.reshape((NumUnits.shape[0], 1)), Price.shape[1], axis=1) * CointJohansenTest.evec[:, 0][0] * np.array([1, -1]) * Price
PNL = (Positions.shift(1) * (Price.iloc[:, 1:] - Price.iloc[:, 1:].shift(1)) / Price.iloc[:, 1:].shift(1)).sum(axis=1)
Return = PNL / Positions.shift(1).abs().sum(axis=1)
Return[~np.isfinite(Return)] = 0.0
CumReturn = (1+Return).cumprod() - 1

# 展示策略表现
Fig, Axes = plt.subplots(1, 2, figsize=(25, 8))
Axes[0].xaxis_date()
Axes[0].xaxis.set_major_formatter(mdate.DateFormatter('%Y-%m-%d'))
Axes[0].plot(CumReturn.index, CumReturn.values, color='r')
Axes[0].set_title("Cumulative Compounded Return")
Axes[1].bar(PNL.index, PNL.values)
Axes[1].set_title("P&L")
plt.show()

DTs = Return[pd.notnull(Return)].index
nYear = (DTs[-1] - DTs[0]).days/365
print("APR: %.2f%%, Sharpe Ratio: %.2f" % (((1+Return).prod()**(1/nYear)-1) * 100, np.sqrt(len(DTs)/nYear)*Return.mean()/Return.std()))
```
