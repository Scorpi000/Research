# 理论模型

铁矿石指存在有利用价值的、含有铁元素或铁化合物的矿石。铁矿石用途单一，98% 作为钢铁生产原材料使用。

## 分类

经济可用的铁矿石主要有：
* 赤铁矿：主要成分 ${\rm Fe_2O_3}$。
* 磁铁矿：主要成分 ${\rm Fe_3O_4}$，是 ${\rm FeO}$ 和 ${\rm Fe_2O_3}$ 的复合物。
* 菱铁矿：主要成分 ${\rm FeCO_3}$。
* 褐铁矿：含有氢氧化铁的矿石，是针铁矿（${\rm HFeO_2}$）和鳞铁矿（${\rm FeO(OH)}$）两种不同结构矿石的统称，主要成分 ${\rm 2mFe_2O_3\cdot nH_2O}$。

按照矿石的酸碱度分为：
* 酸性矿石：适用于酸性渣熔炼的高炉（为了提高高炉利用系数和降低焦耗）
* 碱性矿石：适用于碱性渣熔炼的高炉（为了更好的脱硫）
国内高炉倾向于碱性矿，但大部分进口粉矿和国产精粉为酸性矿，在后期熟料加工时需添加碱性溶剂。

根据物理形态不同分为：
* 原矿：指采出未经选矿或其他加工的矿石，由于大部分铁矿石原矿自身易碎，因此在开采过程中即可得到大颗粒组成的块矿和小颗粒组成的粉矿。
* 块矿：直径在 6~30mm，由铁矿石破碎过程中产生的较大颗粒组成，是可以直接入炉的高品位矿，又被称为“高炉富矿”。富矿多为含铁量在 70% 以上的磁铁矿和赤铁矿。
* 粉矿：直径在 6mm 以下，一般是直接开采所得或是其后经过粉碎和筛分加工处理。需人工造块后成为人工富矿（熟矿）后才能投入高炉。品位低于 40% 的矿石，需要经过细磨、选矿达到较高品位才能使用，选矿之后的铁矿石即为“精矿”（也叫精粉）。其中，粉矿是生产烧结矿的主要原料，精矿是生产球团矿的主要原料。

![铁矿石分类](.\images\铁矿石分类.PNG)

世界铁矿石中粉矿占有相当大比例，中国进口中有 75% 以上是粉矿。中国钢铁企业主要以烧结矿作为高炉炼铁原料，提炼 1 吨生铁约需要 1.63 吨品位为 63.5% 的铁矿石。基于现货市场的需求，大商所铁矿石期货交易标的物为粉矿。

根据产地及品位、杂质含量及烧结性能的不同，市场上有多个铁矿石品种，如 PB 粉、杨迪粉等。其中，澳大利亚的麦克粉、纽曼粉和 PB 粉是三大主流矿种，被合称为 MNP。由于进口铁矿石来源不一，矿石的品位、物理、化学性能也不尽相同，为达到矿石质量指标要求，确保入炉原料的可靠性和经济性，钢企需要对各地矿粉的成矿性能做进一步的研究分析，并且需要对各种矿按比例进行搭配，使之混合均匀。目前，国内铁矿配矿业务发展基本上已经形成以钢铁企业原料场混匀配矿为主、港口贸易商的混合配矿为辅的两大类型。

## 生产

### 产地

**巴西**

巴西主要矿山分布在铁四角地区（以伊塔比拉为中心的地区）和北部的帕拉州卡拉加斯地区。巴西铁矿以赤铁矿为主，品位大多在 55%~67.5%，属于高品位矿石， 并且杂质少，具有高铁、中硅、低铝的特点，是目前大型钢厂首选原料之一。巴西铁矿产品以高品位的富矿粉和天然块矿为主，此外，巴西也是国际球团矿市场的主要供应地区。

巴西有两个主要矿石港口，分别是图巴朗和瓜伊巴港，均位于大西洋西南岸、 巴西东南部。图巴朗是巴西最大的铁矿石输出港口之一， 也是世界三大散货出口港，南侧和北侧分别可以停靠 15 万吨级和 25 万吨级船，瓜伊巴港是巴西联合矿业公司的自有港口，仅承运铁矿石物资，矿石装卸能力为 5000 吨/小时。

**澳大利亚**

澳大利亚已开采的铁矿石多集中于西澳的皮尔巴拉地区，但是皮尔巴拉矿大部分具有较严重的热爆裂性；此外南部和东部地区的铁矿石资源也很丰富，整体品位较高，多在 56%~62% 之间。澳大利亚矿的化学性能不及巴西矿和加拿大矿，铁矿产品以富矿粉和天然块矿为主，球团矿数量较少。

澳大利亚铁矿区的铁路和港口条件较好，目前有位于西澳的三个港口，分别是黑德兰港（必和必拓用）、丹皮尔港（哈默利斯用）和兰伯特角港（罗布河公司用），与之配套的有三条专用铁路。西澳的三个港口均可接卸 20 万吨的超大型散货船，并且由于西澳距我国沿海港口的平均距离在 3100 海里左右，远低于巴西到中国平均 8700 海里的距离，相较而言，澳大利亚的港口条件更好。

**中国**

中国铁矿石来源主要包括国产矿和进口原矿，国产矿来自钢铁企业的自有矿山和独立的矿山企业，进口矿主要来自澳大利亚、巴西和印度等国。

中国国内铁矿石贫矿多、杂质多，平均品位约 30%，低于世界平均品位，资源质量差，并且多为地下矿，开采难度大、成本高。铁矿资源分布分散，缺乏整合，业内多中小型矿山，产业集中度较低。国内尚没有铁矿石企业能够占有显著的市场份额或对整个产业的发展产生实质性的影响，多数铁矿石企业难以形成规模效应。国内铁矿石企业主要分为钢铁集团下属的铁矿石企业和独立的铁矿石企业（含综合性矿业企业下属铁矿石生产主体）。钢铁集团下属的铁矿石企业生产的铁矿石产品一般主要满足集团内部炼铁所需，很少对外销售，较为优质的包括河钢集团下属的司家营铁矿；独立的铁矿石企业则主要是向钢铁企业供应铁矿石产品。总体上看，国内独立铁矿石企业较为分散，且单个企业市场占有率非常有限，基本不构成直接竞争。

**主要矿山**

巴西和澳大利亚是最主要的铁矿石供应国，在这两国相应诞生了四大铁矿石巨头——包括传统的三大巨头巴西淡水河谷、澳大利亚的必和必拓和力拓以及后起之秀 FMG。巴西淡水河谷公司是全球最大的铁矿石生产商和出口商，拥有挺博佩贝铁矿、卡潘尼马铁矿、卡拉加斯铁矿等矿区。力拓矿业集团总部设在英国，澳洲总部在墨尔本，是全球第二和澳大利亚第一的铁矿石生产商。澳大利亚必和必拓公司是全球第三和澳大利亚第二的铁矿石生产商，主要拥有纽曼（Mt.Newman）、扬迪（Yandi）等铁矿。澳大利亚福特斯克金属集团 FMG 成立于 2003 年，2008 年 5 月第一船矿发往中国。FMG 在西澳地区一共规划了奇切斯特、所罗门、西部三个产区。从 2009 年开始， FMG 带来的产量增量开始影响全球铁矿石市场的供需平衡，并且 FMG 为扩大市场份额，在同品质铁矿石价格的基础上进行贴水销售。目前，FMG 是全球第四大铁矿石生产商、澳大利亚第三大矿业公司。

### 工艺流程

**采矿**

采矿分为露天开采及地下开采。 根据矿床赋存状态不同，开采方式可分为露采和地采，露采技术稳定、成熟，地采技术则相对复杂，分为崩落法、空场法和充填法等。露天开采情况下，使用钻机和炸药对矿床分台阶进行穿孔和爆破，原矿采出后装车运输送往选矿厂，分别进行块矿、粉矿加工或选别铁精粉，岩土则运往废石场排弃。

露天采矿流程:
![露天采矿流程](.\images\露天采矿流程.PNG)

**选矿**

选矿流程主要分为富矿选矿和贫矿选矿。选矿技术可分为磁选、浮选/重选等技术，黑色金属（铁、锰、铬）主要采用磁选或浮选。富矿选矿流程和贫矿选矿流程不同，后者流程更长。富矿选矿过程主要依靠破碎筛选和磁选，选矿方式简单，铁矿石原矿通过三段一闭路破碎筛分、水洗流程，得到成品规格块矿和粉矿，对于水洗过程中的溢流矿通过弱磁、强磁选获得铁精粉产品。贫矿选矿主要采取磁选、浮选/重选工艺，铁矿石原矿破碎工艺采用三段一闭路，选别流程采用三段一闭路，选别流程采用磨矿-分级-弱磁-强磁-反浮选/重选，最终通过浓缩和过滤流程生产出铁精粉。

## 运输

全球开采的铁矿石大致可以分为两类，一类是直接用于自有的钢厂，如鞍钢、河钢均有铁矿，此类铁矿多采取铁路运输，另一类是用于贸易，多走海运。

由于海运距离较长，因此从经济型角度考虑，大吨位散货船比小吨位散货船在单位吨矿石运费上更便宜。随着国内港口泊位吨位的改善，40 万吨级 VLOC 可直接靠泊中国港口，因此铁矿石运输将向能靠泊 40 万吨级散货船的大港集中。从港口看，可接卸 VLOC 的港口大多分布在环渤海地区，包括大连港、曹妃甸、青岛港和宁波-舟山港，共 7 个泊位。部分长三角地区港口虽然也有较大的铁矿石接卸量，但长江入海口附近直接用于铁矿石接卸的码头却并不多。铁矿石运输一般先运送至北方港口，后经转运至长江沿线港口。

![国内铁矿石港口体系](.\images\国内铁矿石港口体系.PNG)


从海运路线看，澳大利亚的铁矿石生产集中在西澳，因此澳大利亚至中国的海运通常从西澳港口出发；而巴西至中国的路线则从巴西东海岸港口出发，绕过非洲好望角，途经马来西亚到达中国。从运距和海运时间看，澳大利亚的西澳是离中国最近的港口。巴西距离远东市场运距较远，因此在与澳洲矿山的竞争中处于劣势。

## 现货价格

铁矿石以第三方现货指数为依据定价，国际上比较有影响力的铁矿石指数有：
* 环球钢讯的 TSI 指数
* 金属导报的 MBIO 指数
* 普氏能源资讯的 Platts 指数

## 价格影响因素

黑色产业链的上游主要是铁矿石、煤炭、有色金属、电力等，中游即为钢铁行业，包括长材、板材和特钢。这三个品种对应的下游也不尽相同，长材主要用于基建和房地产，板材主要用于汽车、家电和造船等，特钢则用于特殊制造，比如机械、汽车、军工和核电等对钢材有特殊要求的行业。

一般而言，黑色产业链的传导是由下游到上游，即下游需求影响钢铁产量，进而影响对于上游原料的需求。而需求的变化主要源于外生变量，比如 GDP、经济景气、周期的因素。

### 宏观经济

宏观经济的健康快速发展，对铁矿石市场具有很强的支撑和拉动作用。 宏观经济主要是通过影响下游产业的需求，进而影响铁矿石市场变化，宏观经济表现是铁矿石市场需求的晴雨表，对其价格变动有重要影响，当宏观经济运行良好，建筑业、汽车制造业等相关行业对钢材的需求较为强劲，相应会带动铁矿石的需求， 支撑其价格在高位运行。

### 政策环境

铁矿石是国际大宗贸易商品，其价格受各种政策因素影响，如产地国的进出口政策，进口国关税政策、以及消费国的钢铁产业发展政策等均会对铁矿石价格造成影响。

### 供给因素

**进口矿**

中国铁矿石高度依赖进口，进口来源主要有澳大利亚、巴西、南非、印度等。反映进口矿对铁矿石价格的影响的供给因子有：
* 海外港口铁矿石发货量
* 铁矿石到货量北方港口合计
* 铁矿石进口量
* 巴西发货量
* 澳洲发货量
* 大矿山产量
* 矿山开工率
* 运费
* 汇率

**国产矿**

反映国产矿对铁矿石价格的影响因子有：
* 国内矿山铁精粉产量

### 库存因素
库存相当于供给和需求博弈后的蓄水池，对分析商品价格至关重要。

库存因子：
* 矿山库存
* 港口库存
* 钢厂库存

### 需求因素

由于铁矿石用途单一， 98% 作为钢铁生产原材料使用，从而铁矿石的价格与下游钢铁行业息息相关。

需求因子：
* 钢材价格
* 钢材利润
* 钢材产量
* 钢厂开工率
* 疏港量

### 替代品

**废钢**

## 数据预处理

**频率调整**

基本面数据更新频率存在差异，从而需要将所有因子统一到同一频率中，而统一到同一频率有以下两种方式：
* 统一成低频数据：将所有因子都统一变成低频因子，例如尽管我们有周频的库存，依然可以用每个月最后一周或者本月所有周的平均值作为该月度库存因子。
* 统一成高频数据：将所有的因子统一成高频因子，即尽管无法获得日频的库存数据，但可以通过一些手段将其映射到高频时间点上，如根据信息获得时间，将两次因子更新时间之前填充为前一次更新所得数据的方式获得高频数据便是常见的一种形式。

传统的统一频率方法以第一种为主，这种方法优点是数据规整，便于在低频基础上实现资产价格的长期预测，调仓频率固定，从而可以有效控制换手率。但是其
缺点也非常明显，首先由于基本面数据更新并不总是及时且未必一致，通常下个月的月中才能发布上个月的数据，从而很难做到各个基本面因子能够很好对齐；
其次很难做到及时性，如新的数据发布后，如果没到我们调仓的低频时点，其信息就无法及时准确地反映到模型中去，尤其是一些重要的行情数据通常能够体现
出市场的变化，因此有必要将其充分考虑到整个模型框架中。鉴于此，将所有因子统一成高频因子，因子两次更新期间的数据用前一次数据代替。

**季节调整**

由于产业链以及下游需求通常有季节性特点，基本面因子很容易表现出较强的季节性，如钢材的月度产量一般来说冬季的四个月为淡季，而 3 月份至 10 月份为生产相对活跃的阶段，这可能与下游需求有关。粗钢月度产量由于具有季节性，所以很难达到平稳，而将粗钢月度产量进行季节性调整后，其季节性减弱，同时使得数据更倾向于平稳过程。

**平稳性处理**

有些数据需要环比处理，以保证数据平稳性和量纲统一。

## 指标释义

### 波罗的海干散货指数(Baltic Dry Index, BDI)

波罗的海干散货指数是航运业的经济指标，它包含了航运业的干散货交易量的转变。BDI 波罗的海指数是由几条主要航线的即期运费（Spot Rate）加权计算而成，反映的是即期市场的行情。BDI 指数一向是散装原物料的运费指数，散装船运以运输钢材、纸浆、谷物、煤、矿砂、磷矿石、铝矾土等民生物资及工业原料为主。因此，散装航运业营运状况与全球经济景气荣枯、原物料行情高低息息相关。故波罗的海指数可视为经济领先指标。

### 普氏指数

普氏价格指数数据是普氏能源资讯（Platts）制定的、通过电话问询等方式，向矿商、钢厂及钢铁交易商采集数据，其中会选择 30 家至 40 家“最为活跃的企业”进行询价，其估价的主要依据是当天最高的买方询价和最低的卖方报价，而不管实际交易是否发生。2010 年，普氏价格指数被世界三大矿山选为铁矿石定价依据。

### TSI 钢铁指数

TSI 钢铁指数基于实际交易，发布全球各地的每周主要钢材, 废钢和铁矿石产品的价格信息，是业内领先的独立提供商。TSI 钢铁指数是 SBB 环球钢讯子公司，于 2006 年 4 月 25 日正式成立。
