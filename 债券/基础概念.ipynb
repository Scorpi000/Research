# 债券

## 定义

债券(Bond), 又称固定收益证券(Fixed-Income Securities).


## 分类

* 利率债: 市场习惯将国债, 央票(央行发行的债券, 主要目的是调节流动性, 目前已停发), 地方政府债, 政策性金融债等"不存在违约风险"的债券统称为利率债
* 信用债: 其他存在违约风险的债券统称为信用债

* 政策性金融债: 官方将国家开发银行定位为开发性金融机构, 中国进出口银行和中国农业发展银行定位为政策性金融机构, 市场习惯沿用之前的说法, 将以上三家银行统称为政策性银行, 其发行的债券一般简称为国开债, 进出债, 农发债 


## 发行

**发行价格**

债券原始投资者购入债券时应支付的市场价格, 分为:
* 平价发行：当债券票面利率等于市场利率时，债券发行价格等于面值，此时债券是平价发行;
* 折价发行：指债券发行价格小于面值。当债券票面利率低于市场利率时，企业仍以面值发行就不能吸引投资者，故一般要折价发行;
* 溢价发行：指债券发行价格大于面值。当债券票面利率高于市场利率时，企业仍以面值发行就会增加发行成本，故一般要溢价发行.


## 交易

**交易方式**

* **现货交易**: 债券买卖双方对债券的买卖价格均表示满意，在成交后立即办理交割，或在很短的时间内办理交割的一种交易方式.
* **回购交易**: 债券持有一方（出券方）和购券方在达成一笔交易的同时，规定出券方必须在未来某一约定时间以双方约定的价格再从购券方那里购回原先售出的那笔债券，并以商定的利率（价格）支付利息。回购交易又分为:
    - **质押式回购交易**: 正回购方（卖出回购方、资金融入方）在将债券出质给逆回购方（买入返售方、资金融出方）融入资金的同时，双方约定在将来某一指定日期，由正回购方按约定回购利率计算的资金额向逆回购方返回资金，逆回购方向正回购方返回原出质债券的融资行为。
    - **买断式回购交易(开放式回购)**: 债券持有人（正回购方）在将一笔债券卖给债券购买方（逆回购方）的同时，交易双方约定在未来某一日期，再由卖方（正回购方）以约定价格从买方（逆回购方）购回相等数量同种债券的交易行为。
* **远期交易**: 交易双方成交以后，交割和清算按照合约中规定的价格在未来某一特定时间进行的交易。


**报价方式**

在债券交易中, 报价是指 100 元面值债券的价格. 分为:
* **全价(发票价格, Dirty Price, Full Price)**: 即买卖双方实际支付价格, 等于净价 + 应计利息. 
* **净价(除息价格, Clean Price)**: 扣除应付累计利息后的报价, 等于全价 - 应计利息.


**应计利息(累计利息, Accrued Interest)**

短期债券: 通常全年天数定为 360 天, 半年定为 180 天, 利息累计天数分为按实际天数(ACT)计算(ACT/360, ACT/180) 和按每月 30 天计算(30/360, 30/180).

例: 当前日期是 2010 年 3 月 5 日, 债券年息 6%, 面值 100 元, 每半年付息一次的 1 年期债券, 上次付息为 2009 年 12 月 31 日, 则应计利息为:
* ACT/180: 累计天数(算头不算尾) = 31 天(1 月) + 28 天(2 月) + 4 天(3 月) = 63 天, 应计利息 = 100 × 6%/2 × 63/180 = 1.05
* 30/180: 累计天数(算头不算尾) = 30 天(1 月) + 30 天(2 月) + 4 天(3 月) = 64 天, 应计利息 = 100 × 6%/2 × 64/180 = 1.07

中长期附息债券: 全年天数分为实际全年天数或固定 365 天, 累计利息天数也分为实际天数和每月按 30 天计算两种. 我国交易所市场对附息债券的计息规定为: 全年天数统一按 365 天计算, 利息累计天数规则是: 按实际天数计算, 算头不算尾, 闰年 2 月 29 日不计息, 即 ACT/365.

例: 当前日期是 2011 年 3 月 5 日, 债券年息 8%, 面值 100 元, 每年付息一次的债券, 上次付息为 2010 年 12 月 31 日, 则应计利息为, ACT/365: 累计天数(算头不算尾) = 31 天(1 月) + 28 天(2 月) + 4 天(3 月) = 63 天, 应计利息 = 100 × 8% × 63/365 = 1.38

贴现式债券: 我国目前对贴现式债券按照实际天数(算头不算尾)计算应计利息, 闰年 2 月 29 日也计息, 公式为: 应计利息 = (到期总付额 - 发行价格) / 起息日至到期日的天数 × 起息日至结算日的天数

例: 当前日期是 2010 年 3 月 5 日, 债券面值 100 元, 发行价 85 元, 发行日 2008 年 1 月 10 日, 到期日 2011 年 1 月 10 日, 则起息日至到期日为 3 年, 其中 2008 年是闰年, 所以起息日至到期日的天数 = 365 + 366 + 365 = 1096 天, 起息日至结算日的天数 = 365 + 366 + 53 = 784 天, 应计利息 = (100 - 85) / 1096 × 784 = 10.73

## 债券市场

目前，我国债券市场形成了银行间债券市场、交易所市场和商业银行柜台市场三个基本子市场在内的统一分层的市场体系，在中央国债登记结算有限公司（简称“中央结算公司”）实行集中统一托管，又根据参与主体层次性的不同，相应实行不同的托管结算安排。

从交易场所看，中国债券市场可以分为场外交易市场和场内交易市场。其中，场外交易市场主要指银行间债券市场和商业银行柜台交易市场，场内交易市场指交易所债券交易市场, 包括: 上海证券交易所, 深圳证券交易所, 中国金融期货交易所。

从托管体系看，中央结算公司是中国债券市场的总托管人，直接托管银行间债券市场参与者的债券资产，而中证登公司作为分托管人托管交易所债券市场参与者的债券资产，四大国有商业银行作为二级托管人托管柜台市场参与者的债券资产。银行间债券市场采用实名制一级账户托管体制，托管依托中央结算公司的债券簿记系统进行。

转托管，是指处于不同市场由不同机构分别托管的债券可以通过转托管业务将债券从一个市场转移到另一个市场进行交易。目前，跨市场转托管的债券主要是部分国债和企业债。

### 银行间市场

债券市场的主体场所，在中国债券市场发挥主导作用。记账式国债的大部分、政策性金融债券都在该市场发行并上市交易。这一市场参与者是各类机构投资者，属于大宗交易市场（批发市场），实行双边谈判成交，逐笔结算。2002 年记账式国债开始柜台交易，使个人投资者间接进入了银行间市场。交易方式有: 现券交易、质押式回购、买断式回购、远期交易. 

银行间市场投资者的证券账户直接开立在中央结算公司，实行一级托管；中央结算公司还为这一市场的交易结算提供服务。

中国外汇交易中心提供了一个交易系统, 银行间市场参与者可以在交易系统(俗称交易前台或交易机)录入交易, 其中一方发起交易, 双方确认无误后, 收到交易的一方确认就完成了一笔债券交易.

全国银行间债券市场的交易日为每周一至周五，遇法定节假日调整除外，交易时间为每交易日 9:00-12:00、13:30-16:30. 回购期限最短为 1 天，最长为 1 年。参与者可在此区间内自由选择回购期限，不得展期。债券交易数额最小为债券面额十万元，交易单位为债券面额一万元。收益率、利率、价格保留小数点后 4 位，交易金额、结算金额精确到分。

债券交易采用询价交易和点击成交两种交易方式.


**询价交易**

指交易双方自行协商确定交易价格以及其他交易要素的交易方式，包括报价、格式化询价、确认成交三个交易步骤:
1. **报价**分为：
    * **公开报价**: 指参与者为表明自身交易意向而面向市场作出的、不可直接确认成交的报价。公开报价又分为:
        - **意向报价**: 指参与者为表明自身对资金或债券的供给或需求而面向市场作出的公开报价。
        - **双向报价**: 指经中国人民银行批准在银行间债券市场开展双边报价业务的参与者在进行现券买卖公开报价时，在中国人民银行核定的债券买卖价差范围内连续报出该券种的买卖实价，并可同时报出该券种的买卖数量、清算速度等交易要素。进行双边报价的参与者有义务在报价或合理范围内与对手方达成交易。
    * **对话报价**: 指参与者为达成交易而直接向交易对手方作出的、对手方确认即可成交的报价。
2. **格式化询价**是指参与者必须按照交易系统规定的格式内容填报自己的交易意向。未按规定所作的报价为无效报价。
3. **确认成交**须经过“对话报价——确认”的过程，即一方发送的对话报价，由对手方确认后成交，交易系统及时反馈成交。


**点击成交交易**

指报价方发出具名或匿名的要约报价，受价方点击该报价后成交或由限价报价直接与之匹配成交的交易方式。点击成交交易方式下，最低交易量为券面总额 100 万元，交易量最小变动单位为券面总额 10 万元, 报价方式包括:
* **做市报价**: 是指报价方就某一券种同时报出买入和卖出价格及数量的报价。做市商和尝试做市机构可对其设定的做市券种进行双边报价。
* **点击成交报价**: 是指报价方就某一券种报出买入或卖出价格及数量的报价.
* **限价报价**: 是指报价方发出的单边买入或卖出报价，该报价不向市场公开，可自动与符合成交条件的点击成交报价成交。


**现货交易**

按净价交易, 全价结算.

应计利息为上一付息日（或起息日）至结算日之间累计的按百元面值计算的债券发行人应付给债券持有人的利息. 2007 年中央国债登记结算有限责任公司发出"关于调整中央债券综合业务系统债券应计利息计算公式的通知", 应计利息日计数基准由 “实际天数 / 365” 调整为 “实际天数 / 实际天数”, 即债券应计利息天数按当期的实际天数计算(算头不算尾)，付息区间天数按实际天数计算(算头不算尾), 闰年 2 月 29 日计算利息。具体如下:
* 付息债券(包括固定利率债和浮动利率债、普通证券和资产支持证券): 记 $AI$ 为每百元面值债券的应计利息; $C$ 为每百元面值年利息, 对浮动利率债券, $C$ 根据当前付息期的票面利率确定; $t$: 起息日或上一付息日至结算日的实际天数, 含 2 月 29 日; $TY$ 为本付息周期所在计息年度的实际天数(从起息日起计算的结算日所属的整年度, 即债券本身的完整计息年度), 含 2 月 29 日; $TS$ 为本付息周期的实际天数, 含 2 月 29 日; $f$ 为一年内的付息次数, $m$ 为百元面值当前剩余的本金值, 则:
    - 付息周期等于一年: $AI = C\times \frac{t}{TS}\times\frac{m}{100}$
    - 付息周期小于一年, 按实际天数付息: $AI = C\times \frac{t}{TY}\times\frac{m}{100}$
    - 付息周期小于一年, 按平均值付息: $AI = \frac{C}{f}\times \frac{t}{TS}\times\frac{m}{100}$
* 利随本清债券: 记 $AI$ 为每百元面值债券的应计利息; $C$ 为每百元面值年利息; $K$: 债券起息日至结算日的整年数; $t$: 起息日或上一理论付息日至结算日的实际天数(从起息日起, 一年为一个理论付息期), 含 2 月 29 日; $TS$ 为本理论付息期的实际天数, 含 2 月 29 日, 则: $AI = K\times C + C\times\frac{t}{TS}$
* 贴现和零息债券: 记 $AI$ 为每百元面值债券的应计利息; $M$ 为每百元面值到期兑付额, 均为 100; $P_d$: 债券发行价; $t$: 起息日至结算日的实际天数, 含 2 月 29 日; $T$ 起息日至到期兑付日的实际天数, 含 2 月 29 日, 则: $AI = (M-P_d)\times\frac{t}{T}$

清算速度指成交日 T 与结算日之间的工作日天数 N，有 T+0 和 T+1 两种, 参与者双方自行决定债券结算和资金清算速度。银行间债券交易中 “T+0” 指参与者于债券交易成交日进行债券结算和资金清算；“T+1” 指参与者于债券交易成交日之后的第一个营业日进行债券结算和资金清算。

结算日指债券交割和资金支付的日期。结算日 = 成交日 + 清算速度。

结算方式包括券款对付、见款付券和见券付款。



### 交易所市场

另一重要部分，它由除银行以外的各类社会投资者参与, 属于集中撮合交易的零售市场，实行净额结算。交易方式有: 现货交易、质押式回购. 中国金融期货交易所开展国债期货交易.

交易所市场实行两级托管体制，其中，中央结算公司为一级托管人，负责为交易所开立代理总账户，中国证券登记结算公司为债券二级托管人，记录交易所投资者账户，中央结算公司与交易所投资者没有直接的权责关系。交易所交易结算由中证登负责。


**现货交易**
* 国债、地方政府债券、企业债券、公司债券、分离交易的可转换公司债券采用净价交易, 全价结算的方式, 可转换公司债券采用全价交易, 全价结算的方式。
* T+0 交易, 当日买入的债券当日可以卖出. 
* 以每百元债券(1 张)价格进行报价.
* 申报价格最小变动单位上交所为 0.01 元, 深交所为 0.001 元; 
* 交易单位为手, 人民币 1000 元面值债券(10 张)为 1 手. 申报数量为 1 手或其整数倍, 单笔申报最大数量不超过 10 万手(100 万张). 
* 债券现券竞价交易不实行价格涨跌幅限制. 债券上市首日开盘集合竞价的有效竞价范围为发行价的上下 30%，连续竞价、收盘集合竞价的有效竞价范围为最近成交价的上下 10%; 非上市首日开盘集合竞价的有效竞价范围为前收盘价的上下 10%; 连续竞价、收盘集合竞价的有效竞价范围为最近成交价的上下 10%.
* 开盘价为当日该债券集合竞价中产生的价格, 集合竞价不能产生开盘价的，连续竞价中的第一笔成交价为开盘价。
* 收盘价为当日该债券最后一笔成交前一分钟所有成交价的加权平均价（含最后一笔成交）。当日无成交的，以前一交易日的收盘价为当日收盘价。
* 兑付兑息于兑息登记日进行兑息权益资金清算，或于兑付登记日后第 2 个工作日进行兑付权益资金清算, 并于兑付兑息资金发放日(清算日后下一工作日)将权益资金划付至托管该国债的结算参与人。
* 应计利息为上一付息日（或起息日）至结算日之间累计的按百元面值计算的债券发行人应付给债券持有人的利息. 具体如下:
    - 付息债券: 记 $AI$ 为每百元面值债券的应计利息; $C$ 为每百元面值年利息, 对浮动利率债券, $C$ 根据当前付息期的票面利率确定; $t$: 起息日或上一付息日至结算日的实际天数, 不含 2 月 29 日; 则: $AI = C\times \frac{t}{365}$
    - 贴现和零息债券: 记 $AI$ 为每百元面值债券的应计利息; $M$ 为每百元面值到期兑付额; $P_d$: 债券发行价; $t$: 起息日至结算日的实际天数, 含 2 月 29 日; $T$ 起息日至到期兑付日的实际天数, 含 2 月 29 日, 则: $AI = (M-P_d)\times\frac{t}{T}$

注意: 当票面利率不能被 365 天整除时, 计算机系统按每百元利息额的精度(小数点后保留 8 位)计算, 交割单所列应计利息额按 4 舍 5 入原则以元为单位保留 2 位小数列示. 国债交易计息原则是"算头不算尾", 即起息日当天计算利息, 到期日当天不计算利息; 交易日挂牌显示的每百元应计利息额是包括交易日当日在内的应计利息额, 如果国债持有到期, 则应计利息额自起息日至到期日(不包括到期日当日)的应计利息.


**回购交易**

指将债券质押的同时，将相应债券以标准券折算比率计算出的标准券数量为融资额度而进行的质押融资，交易双方约定在回购期满后返还资金和解除质押的交易。其中，质押债券取得资金的交易方为“融资方”；作为其对手的交易方为“融券方”.
* 标准券: 由不同债券品种按相应折算比率折算形成的，用以确定可通过质押式回购交易进行融资的额度
* 标准券折算比率: 指各债券现券品种所能折成的标准券金额与债券面值之比。
* 债券回购交易申报中，融资方按“买入”予以申报，融券方按“卖出”予以申报.
* 融资方应当在回购申报前，通过交易所向登记结算机构申报提交相应的可用于回购质押的债券品种作为质押券。
* 当日购买的债券，当日可用于质押券申报，当日通过大宗交易方式买入的债券，次一交易日可申报作为质押券；当日提交的质押券，在确认质押成功后，当日可用于相应的债券回购交易。
* 计价单位为每百元资金到期年收益.
* 申报单位为手，1000 元标准券为 1 手, 申报数量为 100 手(10 万元)或其整数倍，单笔申报最大数量不超过 10 万手(100 万张).
* 申报价格最小变动单位上交所为 0.005 元, 深交所为 0.001 元。
* 债券回购交易设 1 天、2 天、3 天、4 天、7 天、14 天、28 天、91 天和 182 天等回购期限。债券回购交易的期限自成交次日起按自然日计算。到期日为非交易日的，顺延至下一个交易日。
* 债券回购交易实行“一次成交、两次结算”。债券回购交易初次结算的结算价格为 100 元。回购到期二次结算的结算价格为购回价，回购到期日，根据购回价公式计算应进行交割的资金和质押券数量。上交所购回价计算公式为：购回价 = 100元 ＋ 每百元资金到期年收益 × 100元 × 回购天数 / 360。深交所购回价计算公式为: 购回价 = 100 元 + 每百元资金到期年收益 × 回购天数 / 365
* 回购到期日，融资方可以将相应的质押券申报转回原证券账户，也可以申报继续用于债券回购交易。当日申报转回的债券，当日可以卖出。
* 跟踪债券指数的交易型开放式指数基金比照债券进行回购交易，申报单位为份，申报数量为 100 份或其整数倍，单笔申报最大数量不超过 10 万份。

### 商业银行柜台市场

银行间市场的延伸，也属于零售市场。柜台市场实行两级托管体制，其中，中央结算公司为一级托管人，负责为承办银行开立债券自营账户和代理总账户，承办银行为债券二级托管人，中央结算公司与柜台投资者没有直接的权责关系。与交易所市场不同的是，承办银行日终需将余额变动数据传给中央结算公司，同时中央结算公司为柜台投资人提供余额查询服务，成为保护投资者权益的重要途径。交易方式为现券交易.

## 定价模型

**贴现现金流估值法(Discounted Cash Flow, DCF)**

债券估值的基本原理就是现金流贴现: 把现金流用适当的贴现率进行贴现并求和, 便可得到债券的理论价格.
$$
PV = \sum\limits_{i=1}^{N}\frac{C_i}{(1+y_{t_i})^{t_i}}
$$
其中: $P$ 为债券的理论价格; $N$ 为现金流发生的次数; $t_i$ 为第 i 次现金流距离当前时刻的以年为单位的时间; $C_i$ 为第 i 次现金流金额; $y_{t_i}$ 为年化贴现率.


**贴现率**

债券的贴现率是投资者对该债券要求的最低回报率, 也称必要回报率:

债券贴现率 = 名义无风险收益率 + 风险溢价 = (真实无风险收益率 + 预期通货膨胀率) + 风险溢价
1. 真实无风险收益率: 真实资本的无风险回报率, 理论上由社会资本平均回报率决定.
2. 预期通货膨胀率: 对未来通货膨胀率的估计值.
3. 风险溢价: 根据各种债券的风险大小而定, 是投资者因承担投资风险而获得的补偿, 主要包括: 信用风险(违约风险), 流动性风险, 汇率风险等.
4. 名义无风险收益率: 真实无风险收益率 + 预期通货膨胀率, 一般用相同期限零息国债的到期收益率来近似, 也称为**即期利率(Spot Rates)** 或 **零利率(Zero Rates)**.


**零息债券定价**
$$
PV = \frac{FV}{(1+y_T)^T}
$$
其中: $PV$ 为债券的理论价格; $T$ 为以年为单位的到期时间; $FV$ 为债券面值; $y_T$ 为年化贴现率.


**附息债券定价**

附息债券可以视为一组零息债券的组合.
$$
PV = \sum\limits_{i=1}^{N}\frac{C}{(1+y_{t_i})^{t_i}} + \frac{FV}{(1+y_T)^T}
$$
其中: $PV$ 为债券的理论价格; $N$ 为付息次数; $t_i$ 为第 i 次付息距离当前时刻的以年为单位的时间; $C$ 为每期支付的利息; $y_{t_i}$ 为年化贴现率; $FV$ 为债券面值; $T$ 为以年为单位的到期时间.


**当期收益率(当前收益率, Current Yield)**

债券年利息收入与当前债券市场价格的比率, 即: 
$$
I = \frac{C}{P}
$$
其中: $I$ 为当期收益率; $P$ 为债券的市场价格; $C$ 为年息票利息.


**到期收益率(内部收益率, Yield to Maturity, YTM)**

使得投资购买债券获得的未来现金流的现值等于债券当前市场价格的贴现率, 本质上, 到期收益率是一系列不同期限即期利率的某种平均值. 即使得下式成立的 $y$: 
$$
P = \sum\limits_{i=1}^{N}\frac{C}{(1+y)^{t_i}} + \frac{FV}{(1+y)^T}
$$
其中: $y$ 为年化到期收益率; $P$ 为债券的市场价格; $N$ 为付息次数; $t_i$ 为第 i 次付息距离当前时刻的以年为单位的时间; $C$ 为每期支付的利息; $FV$ 为债券面值; $T$ 为以年为单位的到期时间.

Excel 函数:

    YIELD(settlement, maturity, rate, pr, redemption, frequency, [basis])

* Settlement: 必需。有价证券的结算日。有价证券结算日是在发行日之后，有价证券卖给购买者的日期。 
* Maturity: 必需。有价证券的到期日。到期日是有价证券有效期截止时的日期, 即债券本金兑付的日期。
* Rate: 必需。有价证券的年息票利率。 
* Pr: 必需。有价证券的价格（按面值为 100 计算）, 即债券的净价 
* Redemption: 必需。面值 100 的有价证券的清偿价值。 
* Frequency: 必需。年付息次数。如果按年支付, Frequency = 1; 按半年期支付，Frequency = 2；按季支付，Frequency = 4。 
* Basis: 可选。要使用的日计数基准类型。0 或省略: US(NASD) 30/360; 1: 实际天数/实际天数; 2: 实际天数/360; 3: 实际天数/365; 4: 欧洲 30/360 


**即期利率(零利率, Spot Rates)**

零息债券到期收益率的简称, 即单笔现金流的到期收益率, 在债券定价公式中, 即期利率就是用来进行现金流贴现的贴现率.


**持有期收益率**

买入债券到卖出债券期间所获得的平均收益率, 与到期收益率的区别仅在于末笔现金流是卖出价格而非债券到期偿还金额. 即使得下式成立的 $y$: 
$$
P = \sum\limits_{i=1}^{N}\frac{C}{(1+y)^{t_i}} + \frac{P_T}{(1+y)^T}
$$
其中: $y$ 为年化持有期收益率; $P$ 为买入债券的市场价格; $N$ 为持有期间的付息次数; $t_i$ 为第 i 次付息距离当前时刻的以年为单位的时间; $C$ 为每期支付的利息; $P_T$ 为卖出债券的市场价格; $T$ 为以年为单位的持有时间.


**收益率曲线(Yield Curve)**

描述债券到期收益率和到期期限之间关系的曲线, 其用来描述利率的**期限结构(Term Structure)**. 形态分为: 
* 正收益曲线(上升收益曲线): 期限结构特征为短期收益率低, 长期收益率高.
* 反转收益曲线(下降收益曲线): 期限结构特征为短期收益率高, 长期收益率低.
* 水平收益曲线: 长短期收益率基本相等.

收益率曲线的特点:
* 短期收益率一般比长期收益率更富有变化性;
* 收益率曲线一般向上倾斜, 通常上升收益率曲线是一种正常状态;
* 当利率整体水平较高时, 收益率曲线会呈现向下倾斜形态


**久期(Duration)**

**麦考利久期(Macaulay Duration)** 指债券的平均到期时间, 即债券投资者收回其全部本金和利息的平均时间. 计算公式:
$$
D_{mac} = \sum\limits_{i=1}^{N}\frac{PV(C_i)}{P}\times t_i = \sum\limits_{i=1}^{N}\frac{C_i/(1+y)^{t_i}}{P}\times t_i
$$
其中: $y$ 为年化到期收益率; $P$ 为债券的市场价格; $N$ 为付息次数; $t_i$ 为发生第 i 次现金流距离当前时刻的以年为单位的时间; $C_i$ 为第 i 次的现金流金额; $PV$ 为现金流的面值.

债券组合的久期等于组合中所有债券的久期按照各个债券在组合中的比重的加权平均.

**修正久期(Modified Duration)**: $D_{mod} = \frac{D_{mac}}{1+y}$

债券价格变化与到期收益率的关系可以表达为:
$$
\frac{\mathrm{d}P}{P} = -D_{mac}\frac{\mathrm{d}y}{1+y} = -D_{mod}\mathrm{d}y
$$


**基点(Basis Point, BP)**

在现实市场交易中, 收益率一般不会迅速发生特别大幅度的变动, 通常在描述收益率变动时, 使用的单位是万分之一或 0.01%, 即一个基点或 1 bp.


**基点价值(Price Value of a Basis Point, PVBP, DV01)**

收益率变动一个基点时, 债券价格变动的绝对值. PVBP = |初始价格 - 收益率变动 1 个基点后的价格|.


**凸性(Convexity)**

到期收益率发生变动而引起的价格变动幅度的变动程度。凸性是对债券价格曲线弯曲程度的一种度量。在利率变化比较大的情况下久期就不能完全描述债券价格对利率变动的敏感性。凸性越大, 债券价格曲线弯曲程度越大, 用修正久期度量债券的利率风险所产生的误差越大。久期描述了价格-收益率曲线的斜率，凸性描述了曲线的弯曲程度。凸性是债券价格对收益率的二阶导数: 
$$
C = \frac{\frac{\mathrm{d}^2P}{\mathrm{d}y^2}}{P} = \frac{1}{(1+y)^2}\sum\limits_{i=1}^{N}\frac{C_i/(1+y)^{t_i}}{P}\times t_i(t_i+1)
$$

债券价格-收益率曲线是一条从左上向右下倾斜, 并且下凸的曲线, 债券价格变化与到期收益率的关系可以表达为:
$$
\frac{\Delta P}{P} \approx -D_{mod}\cdot\Delta y + \frac{1}{2}\cdot C\cdot\Delta y^2
$$


**信用利差(Credit Spread)**

除了信用评级不同, 其余条件全部相同的两种债券收益率的差额, 其用来描述利率的**风险结构(risk structure)**. 另外, 将某一风险债券的预期到期收益率与某一具有相同期限和票面利率的无风险债券的到期收益率之间的差额称为**风险溢价(Risk Premium)**


**马基尔利率定理**

从债券的定价模型可以看到，债券的价格与市场利率呈反向变化，传导链条为：利率-利差-到期收益率-价格。马基尔利率定理是债券定价的经典理论，指由马基尔(B.G.Malkirl)提出的五个债券定价原理:
* 马基尔第一定理: 债券的价格和债券收益率变化相反。如果整体利率水平上升，现有债券价格将会下降。相反，如果整体利率水平下降，现有债券价格将会上升。原因在于整体利率水平的变化影响收益率流的货币时间价值效果。对于某种特定的债券，市场利率越高，投资者所要求的收益率就越高，从而导致债券价格越低，反之亦然。
* 马基尔第二定理: 由利率下降导致的债券价格上升的幅度大于同等幅度的利率上升导致的债券价格下降的幅度；换句话说，对同等幅度的贴现率变动，贴现率下降给投资者带来的资本利得大于贴现率上升给投资者带来的损失。
* 马基尔第三定理: 当市场利率的变化幅度为固定值时，债券的票面利率的大小与债券价格的波动幅度之间成反比关系。也就是说，债券的票面利率越低，债券价格的波动幅度越大，反之亦然，债券的票面利率越高，债券价格的波动幅度越小。
* 马基尔第四定理(V-n 动态分析): 当其他条件不变时，债券的到期时间越长，其价格波动幅度越大，反之，债券的到期时间越短，其价格波动幅度越小。也就是说债券的到期时间与债券的波动幅度之间成正比关系。该定理不但适用于不同债券之间价格波动的比较，亦可解释同一债券距到期满时间长短不同时与其价格波动之间的关系。
* 马基尔第五定理: 债券距到期时间越远，债券价格波动的幅度上升，而且以递减的速度上升；反之，债券距到期时间越临近债券价格波动的幅度下降，而且以递增的速度下降。该定理同样可用于不同债券之间价格波动时的比较，亦可用于分析同一债券的价格波动与其到期时间的关系。


**中债估值**

中债估值是由中央国债登记结算有限责任公司(简称中债登)采用现金流贴现模型计算出来的一套指标, 其中包括价格指标(全价, 净价, 应计利息), 风险指标(久期, 凸性), 流动性指标和推荐指标(可信度). 中债估值在每个工作日的 18:30 左右对外发布, 市场参与者对成交不活跃的债券价格可以参考中债估值, 其是目前银行间债券市场参与者对需要以公允价值计量债券价格的权威参考来源.


债券价格主要受市场利率影响，而影响市场利率的主要因素有：
1. 货币供给量: 由狭义货币 $M_1$(流通现金 $M_0$ + 支票存款) 和广义货币 $M_2$($M_1$ + 储蓄存款) 组成，还会计算更广义的 $M_3$($M_2$ + 其他短期流动资产)。
2. 宏观经济政策: 主要包括货币政策和财政政策，除了本国的宏观经济政策外，全球主要经济体、本国主要对外贸易伙伴、主要投资输入输出地的宏观经济政策都对利率有着重要影响。
3. 汇率: 某国汇率币值上升，在其他因素不变的情况下，该国利率会相对下降。这是因为更多的外资会流入该国市场，降低资金成本, 即降低无风险利率，反之亦然。

## 国债

### 定义

国债又称公债, 是中央政府为筹集财政资金而发行的一种政府债券. 它以国家信用为基础, 由中央政府通过法定程序和途径向投资者出具, 承诺在一定时间支付利息和到期偿还本金的债权债务凭证. 由于国债的发行主体是国家, 所以在一国债券市场中它具有最高的信用度, 往往被公认为是安全的投资工具.

国债的基本功能包括财政功能和金融功能:
1. 弥补财政赤字、筹集建设资金、弥补国际收支的不足;
2. 形成市场基准利率, 成为其他金融工具定价的基础;
3. 作为机构投资者融资的工具之一, 可调节短期资金的余缺;
4. 作为公开市场操作工具, 调节资金供求和货币流通量.


### 分类

**债券形式**
* 储蓄国债: 政府面向个人投资者发行、以吸收个人储蓄资金为目的, 满足长期储蓄性投资需求, 不可流通且记名的国债品种. 在持有期内, 持券人如遇特殊情况需要提取现金, 可以到购买网点提前兑取. 按照记录债权形式的不同又分为凭证式国债和电子式储蓄国债. 凭证式国债以"凭证式国债收款凭证"记录债权, 电子储蓄国债以电子方式记录债权. 通常，人们按照习惯将传统凭证式国债称为“凭证式国债”，而将 2006 年 6 月底首发的电子凭证式国债称为“储蓄国债”。
* 记账式国债: 以电子记账形式记录债权, 由财政部面向全社会各类投资者发行, 可以记名、挂失、上市和流通转让的国债品种.

**付息方式**
* 贴现国债: 在票面上不规定利率, 往往以低于债券面值的价格发行, 到期按面值支付本息的债券.
* 附息国债: 利息一般按年或半年支付, 到期还本并支付最后一期利息.

**票面利率**
* 固定利率国债: 指票面利率在发行时确定, 在国债的整个存续期内保持不变.
* 浮动利率国债: 指国债付息的利率一般为相同期限的市场利率加上一个固定利差来确定, 在我国国债市场, 浮动利率国债的利率一般参照一年定存利率确定.

**发行期限**
* 短期国债: 1 年以内.
* 中期国债: 1 年以上, 10 年以下.
* 长期国债: 10 年以上.


### 国债市场

国债市场是国债发行和交易的场所, 一般由一级发行市场、二级交易市场和相应的风险管理市场组成. 其中, 二级市场根据交易场所不同又可分为场外市场和场内市场.

场内市场以沪、深交易所为主体, 参与者为非银行金融机构、企业与个人投资者, 是债券的"零售"市场, 债券托管在中国证券登记结算中心, 采用集中撮合的方式进行交易, 通常单笔交易的金额较小.

场外市场以银行间债券市场为主, 包括商业银行等金融机构参与的场外现券、回购交易以及人民银行的公开市场操作. 银行间市场是债券的"批发"市场, 债券托管在中央国债登记结算有限责任公司(简称"中债登"), 采取柜台交易的方式运作, 机构间通过相互询价的方式完成交易, 通常单笔交易的金额较大

同时, 场外市场还有主要供企业和个人投资者参与的商业银行柜台市场

风险管理市场主要是指利率衍生品市场.

### 国债发行

**发行模式**
1. 以市场化的招标方式向国债一级承销商发行可上市国债;
2. 以承销方式向承销商, 比如商业银行和财政部门所属国债经营机构, 出售不上市的储蓄国债;
3. 以定向招募方式向社会保障机构和保险公司出售定向国债.

我国记账式国债采用面向国债承销团成员公开招标的方式发行. 承销团成员包括从事国债承销业务的商业银行、证券公司、保险公司和信托投资公司等金融机构.


### 定价机制

银行间债券市场以交易双方一对一询价方式进行交易, 同时还建立了完善的做市商制度, 这对于二级市场的价格形成、流动性提高有着重要的作用.

交易所债券市场采用"竞价交易、撮合成交"方式进行交易, 也可通过交易所的固定收益平台进行交易.




## 利率市场分析

利率变动的影响因素:
* 经济基本面因素: “经济增长 + 通货膨胀”
* 资金面因素: “资金供应 + 债券供应”(前者是硬性, 后者更具备弹性)
* 政策面因素: “货币政策 + 财政政策”
* 技术面因素: “估值 + 心理”

影响因素的相互关系:
* 基本面要素为"锚"、资金面如“水”、政策面如“波涛”
* 基本面最领先、政策面其后、资金面再后
* 政策面因素作为中介桥梁，对资金面具有直接影响，对于经济基本面双因素中的“经济增长”的影响敏感度强于“通货膨胀”因素，因此政策基本面因素具有双重属性，特别在分析经济增长中.
